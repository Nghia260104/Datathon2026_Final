```python
# =============================================================================
# DATATHON 2026 — ONE-DAY USER DEEP DIVE — OOM-SAFE MAPREDUCE VERSION
#
# Goal:
#   So sánh one-day login users vs multi-day users vs first-3-consecutive-day users
#   bằng metric normalized per-user / user-rate, tránh bias do cohort size khác nhau.
#
# Why this version is safer:
#   1. Không GROUP BY trực tiếp toàn bộ fact_user_events trong 1 query lớn.
#   2. Đọc Parquet theo batch bằng pyarrow.dataset.
#   3. Pre-aggregate sớm ở cấp user_id × date.
#   4. Chia bucket theo hash(user_id), reduce từng bucket nhỏ.
#   5. Không dùng COUNT(DISTINCT ...) trên bảng event lớn.
#
# Output:
#   /kaggle/working/one_day_user_deep_dive_mapreduce/
# =============================================================================

import os
import gc
import sys
import glob
import shutil
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyarrow as pa
import pyarrow.dataset as ds
import pyarrow.parquet as pq

# =============================================================================
# 0. CONFIG
# =============================================================================

BASE_PATH_2 = "/kaggle/input/datasets/kimkhanhnguyen214/datathon2026-2"
EVENTS_PATH = os.path.join(BASE_PATH_2, "fact_user_events")
EVENTS_GLOB = os.path.join(EVENTS_PATH, "*.parquet").replace("\\", "/")

OUTPUT_DIR = "/kaggle/working/one_day_user_deep_dive_mapreduce"
PARTIAL_DIR = os.path.join(OUTPUT_DIR, "01_partial_user_day")
DAILY_BUCKET_DIR = os.path.join(OUTPUT_DIR, "02_daily_bucket")
LIFECYCLE_BUCKET_DIR = os.path.join(OUTPUT_DIR, "03_lifecycle_bucket")
SUMMARY_BUCKET_DIR = os.path.join(OUTPUT_DIR, "04_summary_bucket")
PLOT_DIR = os.path.join(OUTPUT_DIR, "plots")

for d in [OUTPUT_DIR, PARTIAL_DIR, DAILY_BUCKET_DIR, LIFECYCLE_BUCKET_DIR, SUMMARY_BUCKET_DIR, PLOT_DIR]:
    os.makedirs(d, exist_ok=True)

# Nếu muốn chạy lại từ đầu hoàn toàn thì set True.
# Nếu False và đã có output trung gian thì code sẽ skip bước đã có.
FORCE_REBUILD = False

# Batch càng nhỏ càng ít RAM, nhưng chạy lâu hơn.
BATCH_SIZE = 1_000_000

# Bucket càng nhiều càng ít RAM ở bước reduce, nhưng tạo nhiều file hơn.
N_BUCKETS = 32

DUCKDB_MEMORY_LIMIT = "4GB"
DUCKDB_THREADS = 1

STRICT_CONTACT_EVENTS = [
    "view_phone",
    "contact_chat",
    "contact_zalo",
    "contact_sms",
]

POSITIVE_EVENTS_WITH_OTHER = [
    "view_phone",
    "contact_chat",
    "contact_zalo",
    "contact_sms",
    "other_interaction",
]

BROWSE_EVENTS = [
    "pageview",
    "ad_view",
]

STRICT_SQL = ", ".join([f"'{x}'" for x in STRICT_CONTACT_EVENTS])
POSITIVE_SQL = ", ".join([f"'{x}'" for x in POSITIVE_EVENTS_WITH_OTHER])
BROWSE_SQL = ", ".join([f"'{x}'" for x in BROWSE_EVENTS])

print("EVENTS_PATH exists:", os.path.exists(EVENTS_PATH), EVENTS_PATH)
print("OUTPUT_DIR:", OUTPUT_DIR)

# =============================================================================
# 1. SETUP DUCKDB
# =============================================================================

try:
    import duckdb
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "duckdb"])
    import duckdb

try:
    from IPython.display import display
except Exception:
    def display(x):
        print(x)

DUCKDB_PATH = os.path.join(OUTPUT_DIR, "one_day_mapreduce.duckdb")
TMP_DIR = os.path.join(OUTPUT_DIR, "duckdb_tmp")
os.makedirs(TMP_DIR, exist_ok=True)

con = duckdb.connect(DUCKDB_PATH)
con.execute(f"PRAGMA threads={DUCKDB_THREADS}")
con.execute(f"PRAGMA memory_limit='{DUCKDB_MEMORY_LIMIT}'")
con.execute(f"PRAGMA temp_directory='{TMP_DIR}'")
con.execute("PRAGMA preserve_insertion_order=false")

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

def save_df(df, filename, show_head=20):
    csv_path = os.path.join(OUTPUT_DIR, f"{filename}.csv")
    parquet_path = os.path.join(OUTPUT_DIR, f"{filename}.parquet")
    df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    try:
        df.to_parquet(parquet_path, index=False)
    except Exception as e:
        print("[WARN] parquet save failed:", filename, e)
    print(f"\n[SAVE] {csv_path}")
    display(df.head(show_head))
    return df

def safe_rmtree(path):
    if os.path.exists(path):
        shutil.rmtree(path)
    os.makedirs(path, exist_ok=True)

def parquet_exists(path_or_glob):
    return len(glob.glob(path_or_glob)) > 0

def save_plot(filename):
    path = os.path.join(PLOT_DIR, filename)
    plt.savefig(path, dpi=180, bbox_inches="tight")
    plt.show()
    print("[SAVE PLOT]", path)

# =============================================================================
# 2. MAP STEP — READ fact_user_events BY BATCH, AGG USER × DATE, WRITE PARTIALS
# =============================================================================

print("\n[1/7] MAP STEP: batch aggregate fact_user_events -> partial user-day parquet")

if FORCE_REBUILD:
    print("[FORCE_REBUILD] Removing old intermediate folders...")
    safe_rmtree(PARTIAL_DIR)
    safe_rmtree(DAILY_BUCKET_DIR)
    safe_rmtree(LIFECYCLE_BUCKET_DIR)
    safe_rmtree(SUMMARY_BUCKET_DIR)
    safe_rmtree(PLOT_DIR)

partial_check = os.path.join(PARTIAL_DIR, "bucket=*", "*.parquet")

if parquet_exists(partial_check) and not FORCE_REBUILD:
    print("[SKIP] Partial files already exist.")
else:
    events_ds = ds.dataset(EVENTS_PATH, format="parquet")

    cols = [
        "is_login",
        "user_id",
        "date",
        "event_type",
        "is_contact",
    ]

    batch_idx = 0
    written_files = 0

    for batch in events_ds.to_batches(columns=cols, batch_size=BATCH_SIZE):
        df = batch.to_pandas()

        if len(df) == 0:
            continue

        # Keep login only
        is_login = (
            df["is_login"]
            .astype("string")
            .str.lower()
            .str.strip()
            .eq("login")
        )

        df = df.loc[is_login].copy()

        if len(df) == 0:
            del batch
            gc.collect()
            continue

        df["user_id"] = df["user_id"].astype("string")
        df = df[df["user_id"].notna()]

        # Date
        dt = pd.to_datetime(df["date"], errors="coerce")
        df = df[dt.notna()].copy()
        df["active_date"] = dt[dt.notna()].dt.strftime("%Y-%m-%d")

        if len(df) == 0:
            del batch
            gc.collect()
            continue

        # Clean event type
        df["event_type_clean"] = (
            df["event_type"]
            .astype("string")
            .str.lower()
            .str.strip()
            .fillna("unknown")
        )

        is_contact_num = pd.to_numeric(df["is_contact"], errors="coerce").fillna(0).astype("int8")
        df["contact_flag_row"] = (is_contact_num == 1).astype("int16")

        df["browse_row"] = df["event_type_clean"].isin(BROWSE_EVENTS).astype("int16")
        df["strict_contact_row"] = df["event_type_clean"].isin(STRICT_CONTACT_EVENTS).astype("int16")
        df["positive_including_other_row"] = df["event_type_clean"].isin(POSITIVE_EVENTS_WITH_OTHER).astype("int16")
        df["other_interaction_row"] = (df["event_type_clean"] == "other_interaction").astype("int16")

        # Aggregate inside this batch only
        grp = (
            df
            .groupby(["user_id", "active_date"], sort=False, observed=True)
            .agg(
                event_rows=("event_type_clean", "size"),
                browse_event_rows=("browse_row", "sum"),
                other_interaction_rows=("other_interaction_row", "sum"),
                strict_contact_rows=("strict_contact_row", "sum"),
                positive_including_other_rows=("positive_including_other_row", "sum"),
                contact_flag_rows=("contact_flag_row", "sum"),
            )
            .reset_index()
        )

        # Hash bucket by user_id so the same user always goes to same bucket.
        grp["bucket"] = (
            pd.util.hash_pandas_object(grp["user_id"], index=False).astype("uint64") % N_BUCKETS
        ).astype("int16")

        # Write partitioned partial files
        for b, sub in grp.groupby("bucket", sort=False):
            bucket_dir = os.path.join(PARTIAL_DIR, f"bucket={int(b):02d}")
            os.makedirs(bucket_dir, exist_ok=True)

            out_path = os.path.join(bucket_dir, f"part_{batch_idx:05d}.parquet")
            sub = sub.drop(columns=["bucket"])

            table = pa.Table.from_pandas(sub, preserve_index=False)
            pq.write_table(table, out_path, compression="snappy")
            written_files += 1

        batch_idx += 1

        if batch_idx % 10 == 0:
            print(f"  processed batches={batch_idx:,} | written partial files={written_files:,}")

        del df, grp, batch
        gc.collect()

    print(f"[MAP DONE] batches={batch_idx:,} | written partial files={written_files:,}")

# =============================================================================
# 3. REDUCE STEP — REDUCE PARTIALS INSIDE EACH BUCKET TO EXACT USER × DATE
# =============================================================================

print("\n[2/7] REDUCE STEP: reduce partials by bucket -> daily bucket parquet")

for b in range(N_BUCKETS):
    out_path = os.path.join(DAILY_BUCKET_DIR, f"login_user_daily_bucket_{b:02d}.parquet")
    bucket_glob = os.path.join(PARTIAL_DIR, f"bucket={b:02d}", "*.parquet").replace("\\", "/")

    if os.path.exists(out_path) and not FORCE_REBUILD:
        print(f"[SKIP] daily bucket {b:02d}")
        continue

    if not parquet_exists(bucket_glob):
        print(f"[EMPTY] bucket {b:02d}")
        continue

    print(f"  reducing bucket {b:02d}...")

    con.execute(f"""
    COPY (
        SELECT
            CAST(user_id AS VARCHAR) AS user_id,
            CAST(active_date AS DATE) AS active_date,

            SUM(event_rows)::BIGINT AS event_rows,
            SUM(browse_event_rows)::BIGINT AS browse_event_rows,
            SUM(other_interaction_rows)::BIGINT AS other_interaction_rows,
            SUM(strict_contact_rows)::BIGINT AS strict_contact_rows,
            SUM(positive_including_other_rows)::BIGINT AS positive_including_other_rows,
            SUM(contact_flag_rows)::BIGINT AS contact_flag_rows

        FROM read_parquet('{bucket_glob}')
        GROUP BY
            CAST(user_id AS VARCHAR),
            CAST(active_date AS DATE)
    )
    TO '{out_path.replace("\\", "/")}' (FORMAT PARQUET, COMPRESSION SNAPPY)
    """)

daily_glob = os.path.join(DAILY_BUCKET_DIR, "*.parquet").replace("\\", "/")

date_bounds = con.execute(f"""
SELECT
    MIN(active_date) AS min_date,
    MAX(active_date) AS max_date,
    COUNT(*) AS user_day_rows,
    COUNT(DISTINCT user_id) AS login_users
FROM read_parquet('{daily_glob}')
""").df()

save_df(date_bounds, "00_date_bounds")
MAX_DATE = str(date_bounds["max_date"].iloc[0])
print("MAX_DATE:", MAX_DATE)

# =============================================================================
# 4. BUILD LIFECYCLE PER BUCKET
# =============================================================================

print("\n[3/7] Build lifecycle per bucket")

for b in range(N_BUCKETS):
    daily_path = os.path.join(DAILY_BUCKET_DIR, f"login_user_daily_bucket_{b:02d}.parquet").replace("\\", "/")
    out_path = os.path.join(LIFECYCLE_BUCKET_DIR, f"login_user_lifecycle_bucket_{b:02d}.parquet").replace("\\", "/")

    if os.path.exists(out_path) and not FORCE_REBUILD:
        print(f"[SKIP] lifecycle bucket {b:02d}")
        continue

    if not os.path.exists(daily_path):
        continue

    print(f"  lifecycle bucket {b:02d}...")

    con.execute(f"""
    COPY (
        WITH ordered AS (
            SELECT
                user_id,
                active_date,

                event_rows,
                browse_event_rows,
                other_interaction_rows,
                strict_contact_rows,
                positive_including_other_rows,
                contact_flag_rows,

                LAG(active_date) OVER (
                    PARTITION BY user_id
                    ORDER BY active_date
                ) AS prev_active_date
            FROM read_parquet('{daily_path}')
        ),

        with_gap AS (
            SELECT
                *,
                CASE
                    WHEN prev_active_date IS NULL THEN NULL
                    ELSE DATE_DIFF('day', prev_active_date, active_date) - 1
                END AS inactive_gap_days
            FROM ordered
        ),

        agg AS (
            SELECT
                user_id,
                COUNT(*)::BIGINT AS active_days,
                MIN(active_date) AS first_active_date,
                MAX(active_date) AS last_active_date,
                DATE_DIFF('day', MIN(active_date), MAX(active_date)) + 1 AS observed_span_days,

                COALESCE(MAX(inactive_gap_days), 0) AS max_inactive_gap_days,

                SUM(event_rows)::BIGINT AS total_event_rows,
                SUM(browse_event_rows)::BIGINT AS total_browse_event_rows,
                SUM(other_interaction_rows)::BIGINT AS total_other_interaction_rows,
                SUM(strict_contact_rows)::BIGINT AS total_strict_contact_rows,
                SUM(positive_including_other_rows)::BIGINT AS total_positive_including_other_rows,
                SUM(contact_flag_rows)::BIGINT AS total_contact_flag_rows
            FROM with_gap
            GROUP BY user_id
        )

        SELECT
            *,
            CASE
                WHEN active_days = 1 THEN '01_one_day_only'
                WHEN active_days BETWEEN 2 AND 3 THEN '02_2_3_days'
                WHEN active_days BETWEEN 4 AND 7 THEN '03_4_7_days'
                WHEN active_days BETWEEN 8 AND 14 THEN '04_8_14_days'
                WHEN active_days BETWEEN 15 AND 30 THEN '05_15_30_days'
                WHEN active_days BETWEEN 31 AND 60 THEN '06_31_60_days'
                ELSE '07_61plus_days'
            END AS active_days_bucket,

            CASE
                WHEN active_days = 1 THEN 1
                WHEN active_days BETWEEN 2 AND 3 THEN 2
                WHEN active_days BETWEEN 4 AND 7 THEN 3
                WHEN active_days BETWEEN 8 AND 14 THEN 4
                WHEN active_days BETWEEN 15 AND 30 THEN 5
                WHEN active_days BETWEEN 31 AND 60 THEN 6
                ELSE 7
            END AS active_days_bucket_sort

        FROM agg
    )
    TO '{out_path}' (FORMAT PARQUET, COMPRESSION SNAPPY)
    """)

lifecycle_glob = os.path.join(LIFECYCLE_BUCKET_DIR, "*.parquet").replace("\\", "/")

# =============================================================================
# 5. SUMMARY PER BUCKET
# =============================================================================

print("\n[4/7] Build summary tables per bucket")

def copy_query_to_parquet(query, path):
    con.execute(f"""
    COPY (
        {query}
    )
    TO '{path.replace("\\", "/")}' (FORMAT PARQUET, COMPRESSION SNAPPY)
    """)

for b in range(N_BUCKETS):
    daily_path = os.path.join(DAILY_BUCKET_DIR, f"login_user_daily_bucket_{b:02d}.parquet").replace("\\", "/")
    lifecycle_path = os.path.join(LIFECYCLE_BUCKET_DIR, f"login_user_lifecycle_bucket_{b:02d}.parquet").replace("\\", "/")

    if not os.path.exists(daily_path) or not os.path.exists(lifecycle_path):
        continue

    print(f"  summary bucket {b:02d}...")

    # -------------------------------------------------------------------------
    # 5.1 Active-day bucket contribution
    # -------------------------------------------------------------------------
    out_path = os.path.join(SUMMARY_BUCKET_DIR, f"active_bucket_agg_{b:02d}.parquet")

    if not os.path.exists(out_path) or FORCE_REBUILD:
        copy_query_to_parquet(f"""
        SELECT
            active_days_bucket_sort,
            active_days_bucket,

            COUNT(*)::BIGINT AS users,
            SUM(total_event_rows)::BIGINT AS event_rows,
            SUM(total_browse_event_rows)::BIGINT AS browse_event_rows,
            SUM(total_positive_including_other_rows)::BIGINT AS positive_including_other_rows,
            SUM(total_strict_contact_rows)::BIGINT AS strict_contact_rows,
            SUM(total_contact_flag_rows)::BIGINT AS contact_flag_rows,

            SUM(CASE WHEN total_positive_including_other_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_positive,
            SUM(CASE WHEN total_strict_contact_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_strict_contact,
            SUM(CASE WHEN total_contact_flag_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_contact_flag

        FROM read_parquet('{lifecycle_path}')
        GROUP BY active_days_bucket_sort, active_days_bucket
        """, out_path)

    # -------------------------------------------------------------------------
    # 5.2 Cohort users
    # -------------------------------------------------------------------------
    con.execute(f"""
    CREATE OR REPLACE TEMP TABLE d AS
    SELECT * FROM read_parquet('{daily_path}')
    """)

    con.execute(f"""
    CREATE OR REPLACE TEMP TABLE l AS
    SELECT * FROM read_parquet('{lifecycle_path}')
    """)

    con.execute(f"""
    CREATE OR REPLACE TEMP TABLE cohort_users AS
    WITH eligible AS (
        SELECT *
        FROM l
        WHERE first_active_date <= CAST(DATE '{MAX_DATE}' - INTERVAL 2 DAY AS DATE)
    ),

    first3_consecutive AS (
        SELECT
            e.user_id,
            e.first_active_date AS anchor_date
        FROM eligible e
        INNER JOIN d d1
            ON d1.user_id = e.user_id
           AND d1.active_date = CAST(e.first_active_date + INTERVAL 1 DAY AS DATE)
        INNER JOIN d d2
            ON d2.user_id = e.user_id
           AND d2.active_date = CAST(e.first_active_date + INTERVAL 2 DAY AS DATE)
        WHERE e.active_days >= 3
    )

    SELECT
        'one_day_only' AS cohort,
        user_id,
        first_active_date AS anchor_date
    FROM eligible
    WHERE active_days = 1

    UNION ALL

    SELECT
        'multi_day_2plus' AS cohort,
        user_id,
        first_active_date AS anchor_date
    FROM eligible
    WHERE active_days >= 2

    UNION ALL

    SELECT
        'first3_consecutive' AS cohort,
        user_id,
        anchor_date
    FROM first3_consecutive
    """)

    con.execute("""
    CREATE OR REPLACE TEMP TABLE cohort_window AS
    WITH offsets(day_index) AS (
        VALUES (0), (1), (2)
    )

    SELECT
        c.cohort,
        c.user_id,
        c.anchor_date,
        o.day_index,
        CAST(c.anchor_date + CAST(o.day_index AS INTEGER) AS DATE) AS calendar_date,

        COALESCE(d.event_rows, 0) AS event_rows,
        COALESCE(d.browse_event_rows, 0) AS browse_event_rows,
        COALESCE(d.other_interaction_rows, 0) AS other_interaction_rows,
        COALESCE(d.strict_contact_rows, 0) AS strict_contact_rows,
        COALESCE(d.positive_including_other_rows, 0) AS positive_including_other_rows,
        COALESCE(d.contact_flag_rows, 0) AS contact_flag_rows

    FROM cohort_users c
    CROSS JOIN offsets o
    LEFT JOIN d
        ON d.user_id = c.user_id
       AND d.active_date = CAST(c.anchor_date + CAST(o.day_index AS INTEGER) AS DATE)
    """)

    # Cohort size
    out_path = os.path.join(SUMMARY_BUCKET_DIR, f"cohort_size_{b:02d}.parquet")
    if not os.path.exists(out_path) or FORCE_REBUILD:
        copy_query_to_parquet("""
        SELECT cohort, COUNT(*)::BIGINT AS users
        FROM cohort_users
        GROUP BY cohort
        """, out_path)

    # Day 0 raw summary
    out_path = os.path.join(SUMMARY_BUCKET_DIR, f"day0_raw_{b:02d}.parquet")
    if not os.path.exists(out_path) or FORCE_REBUILD:
        copy_query_to_parquet("""
        SELECT
            cohort,
            COUNT(*)::BIGINT AS users,

            SUM(event_rows)::BIGINT AS sum_event_rows_day0,
            SUM(browse_event_rows)::BIGINT AS sum_browse_event_rows_day0,
            SUM(other_interaction_rows)::BIGINT AS sum_other_interaction_rows_day0,
            SUM(positive_including_other_rows)::BIGINT AS sum_positive_rows_day0,
            SUM(strict_contact_rows)::BIGINT AS sum_strict_contact_rows_day0,
            SUM(contact_flag_rows)::BIGINT AS sum_contact_flag_rows_day0,

            SUM(CASE WHEN positive_including_other_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_positive_day0,
            SUM(CASE WHEN strict_contact_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_strict_contact_day0,
            SUM(CASE WHEN contact_flag_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_contact_flag_day0

        FROM cohort_window
        WHERE day_index = 0
        GROUP BY cohort
        """, out_path)

    # First 3 calendar day line
    out_path = os.path.join(SUMMARY_BUCKET_DIR, f"first3_line_raw_{b:02d}.parquet")
    if not os.path.exists(out_path) or FORCE_REBUILD:
        copy_query_to_parquet("""
        SELECT
            cohort,
            day_index,

            COUNT(*)::BIGINT AS users,
            SUM(CASE WHEN event_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS active_users,

            SUM(event_rows)::BIGINT AS sum_event_rows,
            SUM(browse_event_rows)::BIGINT AS sum_browse_event_rows,
            SUM(positive_including_other_rows)::BIGINT AS sum_positive_rows,
            SUM(strict_contact_rows)::BIGINT AS sum_strict_contact_rows,
            SUM(contact_flag_rows)::BIGINT AS sum_contact_flag_rows,

            SUM(CASE WHEN positive_including_other_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_positive,
            SUM(CASE WHEN strict_contact_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_strict_contact,
            SUM(CASE WHEN contact_flag_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_contact_flag

        FROM cohort_window
        GROUP BY cohort, day_index
        """, out_path)

    # First 3 cumulative
    out_path = os.path.join(SUMMARY_BUCKET_DIR, f"first3_cum_raw_{b:02d}.parquet")
    if not os.path.exists(out_path) or FORCE_REBUILD:
        copy_query_to_parquet("""
        WITH user3 AS (
            SELECT
                cohort,
                user_id,

                SUM(CASE WHEN event_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS active_days_in_first3,

                SUM(event_rows)::BIGINT AS event_rows_3d,
                SUM(browse_event_rows)::BIGINT AS browse_event_rows_3d,
                SUM(positive_including_other_rows)::BIGINT AS positive_rows_3d,
                SUM(strict_contact_rows)::BIGINT AS strict_contact_rows_3d,
                SUM(contact_flag_rows)::BIGINT AS contact_flag_rows_3d

            FROM cohort_window
            GROUP BY cohort, user_id
        )

        SELECT
            cohort,
            COUNT(*)::BIGINT AS users,

            SUM(active_days_in_first3)::BIGINT AS sum_active_days_in_first3,
            SUM(CASE WHEN active_days_in_first3 >= 2 THEN 1 ELSE 0 END)::BIGINT AS users_active_2plus_days,
            SUM(CASE WHEN active_days_in_first3 = 3 THEN 1 ELSE 0 END)::BIGINT AS users_active_3_days,

            SUM(event_rows_3d)::BIGINT AS sum_event_rows_3d,
            SUM(browse_event_rows_3d)::BIGINT AS sum_browse_event_rows_3d,
            SUM(positive_rows_3d)::BIGINT AS sum_positive_rows_3d,
            SUM(strict_contact_rows_3d)::BIGINT AS sum_strict_contact_rows_3d,
            SUM(contact_flag_rows_3d)::BIGINT AS sum_contact_flag_rows_3d,

            SUM(CASE WHEN positive_rows_3d > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_positive_3d,
            SUM(CASE WHEN strict_contact_rows_3d > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_strict_contact_3d,
            SUM(CASE WHEN contact_flag_rows_3d > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_contact_flag_3d

        FROM user3
        GROUP BY cohort
        """, out_path)

    # One-day date quality
    out_path = os.path.join(SUMMARY_BUCKET_DIR, f"one_day_date_raw_{b:02d}.parquet")
    if not os.path.exists(out_path) or FORCE_REBUILD:
        copy_query_to_parquet(f"""
        SELECT
            d.active_date,

            COUNT(*)::BIGINT AS one_day_users,

            SUM(d.event_rows)::BIGINT AS sum_event_rows,
            SUM(d.browse_event_rows)::BIGINT AS sum_browse_event_rows,
            SUM(d.positive_including_other_rows)::BIGINT AS sum_positive_rows,
            SUM(d.strict_contact_rows)::BIGINT AS sum_strict_contact_rows,
            SUM(d.contact_flag_rows)::BIGINT AS sum_contact_flag_rows,

            SUM(CASE WHEN d.positive_including_other_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_positive,
            SUM(CASE WHEN d.strict_contact_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_strict_contact,
            SUM(CASE WHEN d.contact_flag_rows > 0 THEN 1 ELSE 0 END)::BIGINT AS users_with_contact_flag

        FROM d
        INNER JOIN l
            ON d.user_id = l.user_id
           AND d.active_date = l.first_active_date
        WHERE l.active_days = 1
          AND l.first_active_date <= CAST(DATE '{MAX_DATE}' - INTERVAL 2 DAY AS DATE)
        GROUP BY d.active_date
        """, out_path)

    con.execute("DROP TABLE IF EXISTS d")
    con.execute("DROP TABLE IF EXISTS l")
    con.execute("DROP TABLE IF EXISTS cohort_users")
    con.execute("DROP TABLE IF EXISTS cohort_window")
    gc.collect()

# =============================================================================
# 6. COMBINE SUMMARY FILES
# =============================================================================

print("\n[5/7] Combine summaries")

def read_summary_glob(pattern):
    files = glob.glob(os.path.join(SUMMARY_BUCKET_DIR, pattern))
    if len(files) == 0:
        return pd.DataFrame()
    return pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)

cohort_order = ["one_day_only", "multi_day_2plus", "first3_consecutive"]

# 6.1 Active-day bucket contribution
active_raw = read_summary_glob("active_bucket_agg_*.parquet")

active_bucket = (
    active_raw
    .groupby(["active_days_bucket_sort", "active_days_bucket"], as_index=False)
    .sum(numeric_only=True)
    .sort_values("active_days_bucket_sort")
)

for col in ["users", "event_rows", "browse_event_rows", "positive_including_other_rows", "strict_contact_rows", "contact_flag_rows"]:
    total = active_bucket[col].sum()
    active_bucket[col.replace("_rows", "").replace("users", "user") + "_share"] = active_bucket[col] / total if total else 0

active_bucket["positive_user_rate"] = active_bucket["users_with_positive"] / active_bucket["users"].replace(0, np.nan)
active_bucket["strict_contact_user_rate"] = active_bucket["users_with_strict_contact"] / active_bucket["users"].replace(0, np.nan)
active_bucket["contact_flag_user_rate"] = active_bucket["users_with_contact_flag"] / active_bucket["users"].replace(0, np.nan)

save_df(active_bucket, "01_active_day_bucket_contribution")

# 6.2 Cohort size
cohort_size_raw = read_summary_glob("cohort_size_*.parquet")
cohort_size = (
    cohort_size_raw
    .groupby("cohort", as_index=False)["users"]
    .sum()
)

cohort_size["cohort"] = pd.Categorical(cohort_size["cohort"], categories=cohort_order, ordered=True)
cohort_size = cohort_size.sort_values("cohort")
save_df(cohort_size, "02_cohort_size")

# 6.3 Day 0 summary
day0_raw = read_summary_glob("day0_raw_*.parquet")
day0 = (
    day0_raw
    .groupby("cohort", as_index=False)
    .sum(numeric_only=True)
)

day0["avg_event_rows_day0"] = day0["sum_event_rows_day0"] / day0["users"].replace(0, np.nan)
day0["avg_browse_event_rows_day0"] = day0["sum_browse_event_rows_day0"] / day0["users"].replace(0, np.nan)
day0["avg_positive_rows_day0"] = day0["sum_positive_rows_day0"] / day0["users"].replace(0, np.nan)
day0["avg_strict_contact_rows_day0"] = day0["sum_strict_contact_rows_day0"] / day0["users"].replace(0, np.nan)
day0["avg_contact_flag_rows_day0"] = day0["sum_contact_flag_rows_day0"] / day0["users"].replace(0, np.nan)

day0["positive_user_rate_day0"] = day0["users_with_positive_day0"] / day0["users"].replace(0, np.nan)
day0["strict_contact_user_rate_day0"] = day0["users_with_strict_contact_day0"] / day0["users"].replace(0, np.nan)
day0["contact_flag_user_rate_day0"] = day0["users_with_contact_flag_day0"] / day0["users"].replace(0, np.nan)

day0["strict_contact_per_100_users_day0"] = day0["sum_strict_contact_rows_day0"] * 100.0 / day0["users"].replace(0, np.nan)
day0["positive_per_100_users_day0"] = day0["sum_positive_rows_day0"] * 100.0 / day0["users"].replace(0, np.nan)

day0["cohort"] = pd.Categorical(day0["cohort"], categories=cohort_order, ordered=True)
day0 = day0.sort_values("cohort")

save_df(day0, "03_cohort_day0_summary")

# 6.4 First 3 day line
line_raw = read_summary_glob("first3_line_raw_*.parquet")
line = (
    line_raw
    .groupby(["cohort", "day_index"], as_index=False)
    .sum(numeric_only=True)
)

line["active_user_rate"] = line["active_users"] / line["users"].replace(0, np.nan)
line["positive_user_rate"] = line["users_with_positive"] / line["users"].replace(0, np.nan)
line["strict_contact_user_rate"] = line["users_with_strict_contact"] / line["users"].replace(0, np.nan)
line["contact_flag_user_rate"] = line["users_with_contact_flag"] / line["users"].replace(0, np.nan)

line["avg_event_rows_per_user"] = line["sum_event_rows"] / line["users"].replace(0, np.nan)
line["avg_browse_rows_per_user"] = line["sum_browse_event_rows"] / line["users"].replace(0, np.nan)
line["avg_positive_rows_per_user"] = line["sum_positive_rows"] / line["users"].replace(0, np.nan)
line["avg_strict_contact_rows_per_user"] = line["sum_strict_contact_rows"] / line["users"].replace(0, np.nan)

line["cohort"] = pd.Categorical(line["cohort"], categories=cohort_order, ordered=True)
line = line.sort_values(["cohort", "day_index"])

save_df(line, "04_cohort_first3_calendar_day_line")

# 6.5 First 3 cumulative
cum_raw = read_summary_glob("first3_cum_raw_*.parquet")
cum = (
    cum_raw
    .groupby("cohort", as_index=False)
    .sum(numeric_only=True)
)

cum["avg_active_days_in_first3"] = cum["sum_active_days_in_first3"] / cum["users"].replace(0, np.nan)
cum["active_2plus_days_rate"] = cum["users_active_2plus_days"] / cum["users"].replace(0, np.nan)
cum["active_3_days_rate"] = cum["users_active_3_days"] / cum["users"].replace(0, np.nan)

cum["avg_event_rows_3d"] = cum["sum_event_rows_3d"] / cum["users"].replace(0, np.nan)
cum["avg_browse_event_rows_3d"] = cum["sum_browse_event_rows_3d"] / cum["users"].replace(0, np.nan)
cum["avg_positive_rows_3d"] = cum["sum_positive_rows_3d"] / cum["users"].replace(0, np.nan)
cum["avg_strict_contact_rows_3d"] = cum["sum_strict_contact_rows_3d"] / cum["users"].replace(0, np.nan)

cum["positive_user_rate_3d"] = cum["users_with_positive_3d"] / cum["users"].replace(0, np.nan)
cum["strict_contact_user_rate_3d"] = cum["users_with_strict_contact_3d"] / cum["users"].replace(0, np.nan)
cum["contact_flag_user_rate_3d"] = cum["users_with_contact_flag_3d"] / cum["users"].replace(0, np.nan)

cum["strict_contact_per_100_users_3d"] = cum["sum_strict_contact_rows_3d"] * 100.0 / cum["users"].replace(0, np.nan)
cum["positive_per_100_users_3d"] = cum["sum_positive_rows_3d"] * 100.0 / cum["users"].replace(0, np.nan)

cum["cohort"] = pd.Categorical(cum["cohort"], categories=cohort_order, ordered=True)
cum = cum.sort_values("cohort")

save_df(cum, "05_cohort_first3_cumulative_summary")

# 6.6 One-day quality over time
one_date_raw = read_summary_glob("one_day_date_raw_*.parquet")
one_date = (
    one_date_raw
    .groupby("active_date", as_index=False)
    .sum(numeric_only=True)
    .sort_values("active_date")
)

one_date["avg_event_rows"] = one_date["sum_event_rows"] / one_date["one_day_users"].replace(0, np.nan)
one_date["avg_browse_event_rows"] = one_date["sum_browse_event_rows"] / one_date["one_day_users"].replace(0, np.nan)
one_date["avg_positive_rows"] = one_date["sum_positive_rows"] / one_date["one_day_users"].replace(0, np.nan)
one_date["avg_strict_contact_rows"] = one_date["sum_strict_contact_rows"] / one_date["one_day_users"].replace(0, np.nan)

one_date["positive_user_rate"] = one_date["users_with_positive"] / one_date["one_day_users"].replace(0, np.nan)
one_date["strict_contact_user_rate"] = one_date["users_with_strict_contact"] / one_date["one_day_users"].replace(0, np.nan)
one_date["contact_flag_user_rate"] = one_date["users_with_contact_flag"] / one_date["one_day_users"].replace(0, np.nan)

save_df(one_date, "06_one_day_users_by_active_date_quality")

# =============================================================================
# 7. PLOTS — NORMALIZED COMPARISON
# =============================================================================

print("\n[6/7] Plot charts")

def add_bar_labels(ax, fmt="{:.1f}", fontsize=8):
    for container in ax.containers:
        labels = []
        for v in container.datavalues:
            labels.append("" if pd.isna(v) else fmt.format(v))
        ax.bar_label(container, labels=labels, fontsize=fontsize, padding=2)

# -----------------------------------------------------------------------------
# Chart 1: Volume contribution, not quality
# -----------------------------------------------------------------------------

df = active_bucket.copy()
df["user_share_pct"] = df["user_share"] * 100
df["event_share_pct"] = df["event_share"] * 100
df["positive_including_other_share_pct"] = df["positive_including_other_share"] * 100
df["strict_contact_share_pct"] = df["strict_contact_share"] * 100
df["contact_flag_share_pct"] = df["contact_flag_share"] * 100

x = np.arange(len(df))
width = 0.18

fig, ax = plt.subplots(figsize=(16, 6))
ax.bar(x - 2*width, df["user_share_pct"], width, label="User share")
ax.bar(x - width, df["event_share_pct"], width, label="Event share")
ax.bar(x, df["positive_including_other_share_pct"], width, label="Positive share incl. other")
ax.bar(x + width, df["strict_contact_share_pct"], width, label="Strict contact share")
ax.bar(x + 2*width, df["contact_flag_share_pct"], width, label="Contact flag share")

ax.set_title("Volume contribution by active-days bucket — NOT per-user quality", fontsize=14, fontweight="bold")
ax.set_ylabel("Share of total login users / events (%)")
ax.set_xlabel("Active-days bucket")
ax.set_xticks(x)
ax.set_xticklabels(df["active_days_bucket"], rotation=25, ha="right")
ax.legend(ncols=2)
ax.grid(axis="y", alpha=0.25)
plt.tight_layout()
save_plot("01_volume_contribution_by_active_days_bucket.png")

# -----------------------------------------------------------------------------
# Chart 2: Day-0 demand user rates
# -----------------------------------------------------------------------------

df = day0.copy()
df["positive_user_rate_day0_pct"] = df["positive_user_rate_day0"] * 100
df["strict_contact_user_rate_day0_pct"] = df["strict_contact_user_rate_day0"] * 100
df["contact_flag_user_rate_day0_pct"] = df["contact_flag_user_rate_day0"] * 100

x = np.arange(len(df))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - width, df["positive_user_rate_day0_pct"], width, label="Has positive incl. other")
ax.bar(x, df["strict_contact_user_rate_day0_pct"], width, label="Has strict contact")
ax.bar(x + width, df["contact_flag_user_rate_day0_pct"], width, label="Has contact_flag")

ax.set_title("Day-0 demand signal — normalized by cohort size", fontsize=14, fontweight="bold")
ax.set_ylabel("% users in cohort")
ax.set_xlabel("Cohort")
ax.set_xticks(x)
ax.set_xticklabels(df["cohort"].astype(str), rotation=15, ha="right")
ax.legend()
ax.grid(axis="y", alpha=0.25)
add_bar_labels(ax, fmt="{:.1f}")
plt.tight_layout()
save_plot("02_day0_demand_signal_user_rates.png")

# -----------------------------------------------------------------------------
# Chart 3: Day-0 per-user activity intensity
# -----------------------------------------------------------------------------

df = day0.copy()

metrics = [
    ("avg_event_rows_day0", "Events/user"),
    ("avg_browse_event_rows_day0", "Browse/user"),
    ("avg_positive_rows_day0", "Positive/user"),
    ("avg_strict_contact_rows_day0", "Strict contact/user"),
]

x = np.arange(len(df))
width = 0.20

fig, ax = plt.subplots(figsize=(13, 6))

for i, (col, label) in enumerate(metrics):
    ax.bar(x + (i - 1.5) * width, df[col], width, label=label)

ax.set_title("Day-0 activity intensity — mean per user", fontsize=14, fontweight="bold")
ax.set_ylabel("Average rows per user on first active day")
ax.set_xlabel("Cohort")
ax.set_xticks(x)
ax.set_xticklabels(df["cohort"].astype(str), rotation=15, ha="right")
ax.legend(ncols=2)
ax.grid(axis="y", alpha=0.25)
plt.tight_layout()
save_plot("03_day0_activity_intensity_mean_per_user.png")

# -----------------------------------------------------------------------------
# Chart 4: First 3 calendar days line
# -----------------------------------------------------------------------------

df = line.copy()
df["active_user_rate_pct"] = df["active_user_rate"] * 100
df["positive_user_rate_pct"] = df["positive_user_rate"] * 100
df["strict_contact_user_rate_pct"] = df["strict_contact_user_rate"] * 100
df["contact_flag_user_rate_pct"] = df["contact_flag_user_rate"] * 100

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("First 3 calendar days after first activity — normalized comparison", fontsize=15, fontweight="bold")

line_specs = [
    ("active_user_rate_pct", "Active user rate (%)", axes[0, 0]),
    ("positive_user_rate_pct", "Positive user rate (%)", axes[0, 1]),
    ("strict_contact_user_rate_pct", "Strict contact user rate (%)", axes[1, 0]),
    ("avg_event_rows_per_user", "Avg event rows per user", axes[1, 1]),
]

for metric, title, ax in line_specs:
    for cohort in cohort_order:
        sub = df[df["cohort"].astype(str) == cohort]
        if len(sub) > 0:
            ax.plot(sub["day_index"], sub[metric], marker="o", linewidth=2, label=cohort)

    ax.set_title(title, fontweight="bold")
    ax.set_xlabel("Calendar day index from first active day")
    ax.set_xticks([0, 1, 2])
    ax.set_xticklabels(["Day 0", "Day 1", "Day 2"])
    ax.grid(alpha=0.25)
    ax.legend()

plt.tight_layout()
save_plot("04_first3_calendar_days_normalized_behavior.png")

# -----------------------------------------------------------------------------
# Chart 5: First-3-day cumulative rates
# -----------------------------------------------------------------------------

df = cum.copy()
df["positive_user_rate_3d_pct"] = df["positive_user_rate_3d"] * 100
df["strict_contact_user_rate_3d_pct"] = df["strict_contact_user_rate_3d"] * 100
df["active_2plus_days_rate_pct"] = df["active_2plus_days_rate"] * 100
df["active_3_days_rate_pct"] = df["active_3_days_rate"] * 100

x = np.arange(len(df))
width = 0.20

fig, ax = plt.subplots(figsize=(13, 6))
ax.bar(x - 1.5*width, df["positive_user_rate_3d_pct"], width, label="Has positive in first 3d")
ax.bar(x - 0.5*width, df["strict_contact_user_rate_3d_pct"], width, label="Has strict contact in first 3d")
ax.bar(x + 0.5*width, df["active_2plus_days_rate_pct"], width, label="Active >=2 days")
ax.bar(x + 1.5*width, df["active_3_days_rate_pct"], width, label="Active all 3 days")

ax.set_title("First-3-day cumulative user rates", fontsize=14, fontweight="bold")
ax.set_ylabel("% users in cohort")
ax.set_xlabel("Cohort")
ax.set_xticks(x)
ax.set_xticklabels(df["cohort"].astype(str), rotation=15, ha="right")
ax.legend(ncols=2)
ax.grid(axis="y", alpha=0.25)
add_bar_labels(ax, fmt="{:.1f}")
plt.tight_layout()
save_plot("05_first3_cumulative_user_rates.png")

# -----------------------------------------------------------------------------
# Chart 6: One-day user quality over time
# -----------------------------------------------------------------------------

df = one_date.copy()
df["active_date"] = pd.to_datetime(df["active_date"])
df["positive_user_rate_pct"] = df["positive_user_rate"] * 100
df["strict_contact_user_rate_pct"] = df["strict_contact_user_rate"] * 100

fig, ax1 = plt.subplots(figsize=(16, 6))

ax1.plot(df["active_date"], df["one_day_users"], marker="o", linewidth=2, label="One-day users")
ax1.set_ylabel("One-day users")
ax1.set_xlabel("Only active date")
ax1.tick_params(axis="x", rotation=45)
ax1.grid(alpha=0.25)

ax2 = ax1.twinx()
ax2.plot(df["active_date"], df["positive_user_rate_pct"], marker="s", linestyle="--", linewidth=2, label="Positive user rate (%)")
ax2.plot(df["active_date"], df["strict_contact_user_rate_pct"], marker="^", linestyle="--", linewidth=2, label="Strict contact user rate (%)")
ax2.set_ylabel("User rate (%)")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

plt.title("One-day users over time: volume vs demand quality", fontsize=14, fontweight="bold")
plt.tight_layout()
save_plot("06_one_day_users_over_time_quality.png")

# =============================================================================
# 8. QUICK READ
# =============================================================================

print("\n[7/7] Quick interpretation")

one = day0[day0["cohort"].astype(str) == "one_day_only"]
three = day0[day0["cohort"].astype(str) == "first3_consecutive"]

if len(one) > 0 and len(three) > 0:
    one_strict = float(one["strict_contact_user_rate_day0"].iloc[0])
    three_strict = float(three["strict_contact_user_rate_day0"].iloc[0])

    one_pos = float(one["positive_user_rate_day0"].iloc[0])
    three_pos = float(three["positive_user_rate_day0"].iloc[0])

    one_events = float(one["avg_event_rows_day0"].iloc[0])
    three_events = float(three["avg_event_rows_day0"].iloc[0])

    print(f"one_day_only strict contact user rate       : {one_strict:.2%}")
    print(f"first3_consecutive strict contact user rate : {three_strict:.2%}")
    print(f"one_day_only positive user rate             : {one_pos:.2%}")
    print(f"first3_consecutive positive user rate       : {three_pos:.2%}")
    print(f"one_day_only avg event rows day0            : {one_events:.2f}")
    print(f"first3_consecutive avg event rows day0      : {three_events:.2f}")

    print("\nSuggested interpretation:")
    if one_strict >= 0.5 * three_strict:
        print("- One-day users có strict-contact demand khá meaningful so với nhóm active 3 ngày liên tục.")
        print("- Có thể không phải traffic rác; họ có thể tìm/contact xong rồi rời đi, hoặc bị drop sau contact.")
    else:
        print("- One-day users có strict-contact demand thấp hơn nhiều so với nhóm active 3 ngày liên tục.")
        print("- Một phần lớn one-day volume có thể là low-intent browsing hoặc traffic yếu.")

    if one_events >= 0.5 * three_events:
        print("- Cường độ browse ngày đầu của one-day users cũng không quá thấp.")
    else:
        print("- Cường độ browse ngày đầu của one-day users thấp hơn rõ rệt.")

print("\nDONE.")
print("Output folder:", OUTPUT_DIR)
print("Plots folder:", PLOT_DIR)

con.close()
```

```python
# ============================================================
# EARLY REMOVAL + CHAT QUALITY + SELLER REPUTATION EDA
#
# Goal:
# - Gỡ sớm là do chốt được, thất vọng, hay ghost/lead-bait?
# - Agent vs Private khác nhau thế nào?
# - Seller nào có lịch sử ghost/lead-bait nhiều để có thể phạt trong RecSys?
#
# Memory-safe design:
# - dim_listing: read once
# - fact_listing_snapshot: aggregate to item-level
# - fact_post_contact_interactions:
#   + scan batch nhỏ
#   + filter only valid item_id
#   + aggregate each batch to user-item partial parquet
#   + DuckDB final aggregate out-of-core
#
# IMPORTANT:
# - chat_turn_count chỉ là proxy cho độ sâu hội thoại.
# - Không có message direction nên KHÔNG tính được true seller reply rate.
# ============================================================

import os
import gc
import sys
import shutil
import subprocess
import numpy as np
import pandas as pd
import pyarrow.dataset as ds

# Optional plotting
import matplotlib.pyplot as plt

# ============================================================
# 0. CONFIG
# ============================================================

BASE_PATH = "/kaggle/input/datasets/kimkhanhnguyen214/datathon2026-1"

DIM_LISTING_PATH = f"{BASE_PATH}/dim_listing"
SNAPSHOT_PATH = f"{BASE_PATH}/fact_listing_snapshot"
INTERACTION_PATH = f"{BASE_PATH}/fact_post_contact_interactions"

CACHE_DIR = "/kaggle/working/early_removal_chat_quality_eda_cache"
os.makedirs(CACHE_DIR, exist_ok=True)

# Nếu chạy lại từ đầu thì True.
# Sau khi chạy thành công 1 lần, có thể set False để đọc cache.
FORCE_REBUILD = True

# Batch size
BATCH_SIZE_DIM = 1_000_000
BATCH_SIZE_SNAPSHOT = 1_000_000
BATCH_SIZE_INTERACTION = 300_000   # nếu vẫn crash RAM thì giảm 100_000

# DuckDB memory limit
DUCKDB_MEMORY_LIMIT = "3GB"

# ============================================================
# Analysis filters
# ============================================================

# Chỉ phân tích các tin có expected lifetime hợp lý.
# Mấy expected_lifetime_days quá dài dễ là special/outlier.
MIN_EXPECTED_LIFETIME_DAYS = 7
MAX_EXPECTED_LIFETIME_DAYS = 60
MAIN_PACKAGE_DAYS = [10, 15, 30, 60]

# Early removal logic
EARLY_REMOVED_DAYS = 7
NORMAL_EXPIRED_TOLERANCE_DAYS = 2
VERY_EARLY_LIFETIME_RATIO = 0.30

# Chat timing windows
EARLY_CHAT_WINDOW_3D = 3
EARLY_CHAT_WINDOW_7D = 7
LAST_CHAT_WINDOW_5D = 5

# Conversation depth
DEEP_CHAT_TURNS = 4
SHALLOW_CHAT_TURNS = 1

# Classification thresholds
REAL_FAST_CLOSE_MAX_DAYS_TO_FIRST_CHAT = 3
REAL_FAST_CLOSE_MIN_AVG_TURNS = 3
REAL_FAST_CLOSE_MIN_DEEP_RATE = 0.30

ABANDONED_MAX_CHAT_USERS = 1
ABANDONED_MAX_TOTAL_CONTACTS = 1

GHOST_MIN_CHAT_USERS = 5
GHOST_MAX_AVG_TURNS = 2
GHOST_MIN_SHALLOW_RATE = 0.60

ENDGAME_GHOST_MIN_LAST5_CHAT_USERS = 2
ENDGAME_GHOST_MAX_LAST5_AVG_TURNS = 2

SELLER_REPUTATION_MIN_LISTINGS = 5

CATEGORY_MAP = {
    1010: "1010 - Phòng trọ / căn hộ thuê",
    1020: "1020 - Căn hộ / chung cư",
    1030: "1030 - Nhà ở",
    1040: "1040 - Đất / BĐS thương mại",
    1050: "1050 - Dự án / căn hộ mới",
}

# ============================================================
# 1. HELPERS
# ============================================================

def clean_text(x, unknown="Unknown"):
    if pd.isna(x):
        return unknown
    x = str(x).strip()
    if x == "":
        return unknown
    return x


def clean_city(x):
    if pd.isna(x):
        return "Unknown"

    x = str(x).strip()

    if x == "":
        return "Unknown"

    xl = x.lower()

    if (
        "hồ chí minh" in xl
        or "ho chi minh" in xl
        or "tp.hcm" in xl
        or "tp hcm" in xl
        or xl == "hcm"
    ):
        return "Tp Hồ Chí Minh"

    return x


def clean_district(x):
    if pd.isna(x):
        return np.nan

    x = str(x).strip()

    if x == "":
        return np.nan

    xl = x.lower()

    if xl in {"unknown", "unknown district", "none", "null", "nan"}:
        return np.nan

    return x


def clean_status(x):
    if pd.isna(x):
        return "unknown"
    return str(x).strip().lower()


def clean_category(x):
    if pd.isna(x):
        return "Unknown Category"

    try:
        x = int(x)
        return CATEGORY_MAP.get(x, str(x))
    except Exception:
        return "Unknown Category"


def bool_to_int_series(s):
    if pd.api.types.is_bool_dtype(s):
        return s.fillna(False).astype(int)

    if pd.api.types.is_numeric_dtype(s):
        return pd.to_numeric(s, errors="coerce").fillna(0).astype(int)

    return (
        s.astype(str)
        .str.strip()
        .str.lower()
        .isin(["true", "1", "yes", "y", "t"])
        .astype(int)
    )


def save_csv(df, filename):
    path = os.path.join(CACHE_DIR, filename)
    df.to_csv(path, index=False, encoding="utf-8-sig")
    print(f"[SAVE] {path} | shape={df.shape}")
    return path


def display_top(df, n=20, title=None):
    if title:
        print("\n" + "=" * 120)
        print(title)
        print("=" * 120)

    if df is None or len(df) == 0:
        print("Không tìm thấy.")
    else:
        display(df.head(n))


def count_true(mask):
    return int(mask.fillna(False).astype(bool).sum())


def gini(array):
    arr = np.asarray(array, dtype=float)
    arr = arr[~np.isnan(arr)]
    arr = arr[arr >= 0]

    if len(arr) == 0:
        return np.nan

    if np.sum(arr) == 0:
        return 0.0

    arr = np.sort(arr)
    n = len(arr)
    index = np.arange(1, n + 1)

    return (2 * np.sum(index * arr) / (n * np.sum(arr))) - ((n + 1) / n)


# Wider display
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 120)
pd.set_option("display.width", 220)

# ============================================================
# 2. PATHS + CLEAN CACHE
# ============================================================

dim_cache_path = os.path.join(CACHE_DIR, "dim_listing_clean.parquet")
snapshot_item_path = os.path.join(CACHE_DIR, "snapshot_item_summary.parquet")
item_lifecycle_path = os.path.join(CACHE_DIR, "item_lifecycle_with_dim.parquet")

partial_ui_dir = os.path.join(CACHE_DIR, "partial_user_item_interactions")
user_item_path = os.path.join(CACHE_DIR, "user_item_interaction_summary.parquet")
item_chat_path = os.path.join(CACHE_DIR, "item_chat_quality_summary.parquet")

item_behavior_path = os.path.join(CACHE_DIR, "item_behavior_features.parquet")

if FORCE_REBUILD:
    for p in [
        dim_cache_path,
        snapshot_item_path,
        item_lifecycle_path,
        user_item_path,
        item_chat_path,
        item_behavior_path,
    ]:
        if os.path.exists(p):
            os.remove(p)

    if os.path.exists(partial_ui_dir):
        shutil.rmtree(partial_ui_dir)

os.makedirs(partial_ui_dir, exist_ok=True)

# ============================================================
# 3. READ dim_listing
# ============================================================

if os.path.exists(dim_cache_path):
    print("[SKIP] dim_listing cache exists.")
    dim = pd.read_parquet(dim_cache_path)

else:
    print("[BUILD] reading dim_listing...")

    dim_ds = ds.dataset(DIM_LISTING_PATH, format="parquet")
    available_cols = set(dim_ds.schema.names)

    wanted_cols = [
        "item_id",
        "seller_id",
        "category",
        "title",
        "seller_type",
        "ad_type",
        "ad_status",
        "area_sqm",
        "bedrooms",
        "bathrooms",
        "floors",
        "width_m",
        "direction",
        "legal_status",
        "house_type",
        "furnishing",
        "city_name",
        "district_name",
        "ward_name",
        "project_id",
        "price_bucket",
        "images_count",
        "posted_date",
        "expected_expired_date",
    ]

    cols = [c for c in wanted_cols if c in available_cols]

    if "item_id" not in cols:
        raise ValueError("dim_listing thiếu item_id.")

    parts = []

    for i, batch in enumerate(
        dim_ds.to_batches(columns=cols, batch_size=BATCH_SIZE_DIM),
        start=1
    ):
        df = batch.to_pandas()

        for c in wanted_cols:
            if c not in df.columns:
                df[c] = np.nan

        df = df[df["item_id"].notna()].copy()

        df["item_id"] = df["item_id"].astype(str)
        df["seller_id"] = df["seller_id"].astype(str)

        df["posted_date"] = pd.to_datetime(df["posted_date"], errors="coerce")
        df["expected_expired_date"] = pd.to_datetime(df["expected_expired_date"], errors="coerce")

        df["seller_type"] = df["seller_type"].map(lambda x: clean_text(x, "unknown").lower())
        df["ad_type"] = df["ad_type"].map(lambda x: clean_text(x, "unknown").lower())
        df["ad_status_clean"] = df["ad_status"].map(clean_status)

        df["city_name"] = df["city_name"].map(clean_city)
        df["district_name"] = df["district_name"].map(clean_district)
        df["ward_name"] = df["ward_name"].map(lambda x: clean_text(x, "Unknown Ward"))

        df["category_name"] = df["category"].map(clean_category)

        df["is_deleted"] = df["ad_status_clean"].eq("deleted")
        df["is_agent"] = df["seller_type"].eq("agent")
        df["is_private"] = df["seller_type"].eq("private")

        df["expected_lifetime_days"] = (
            df["expected_expired_date"] - df["posted_date"]
        ).dt.days

        df["has_valid_expected_lifetime"] = (
            df["posted_date"].notna()
            & df["expected_expired_date"].notna()
            & df["expected_lifetime_days"].notna()
            & (df["expected_lifetime_days"] >= 0)
        )

        df["is_main_package_lifetime"] = df["expected_lifetime_days"].isin(MAIN_PACKAGE_DAYS)

        df["is_valid_lifetime_for_analysis"] = (
            df["has_valid_expected_lifetime"]
            & df["expected_lifetime_days"].between(
                MIN_EXPECTED_LIFETIME_DAYS,
                MAX_EXPECTED_LIFETIME_DAYS
            )
        )

        df = df.drop_duplicates("item_id")

        keep_cols = wanted_cols + [
            "ad_status_clean",
            "category_name",
            "is_deleted",
            "is_agent",
            "is_private",
            "expected_lifetime_days",
            "has_valid_expected_lifetime",
            "is_main_package_lifetime",
            "is_valid_lifetime_for_analysis",
        ]

        parts.append(df[keep_cols])

        if i % 10 == 0:
            print("dim batches:", i)

        del df
        gc.collect()

    dim = pd.concat(parts, ignore_index=True).drop_duplicates("item_id")
    dim.to_parquet(dim_cache_path, index=False)

print("dim shape:", dim.shape)
display(dim.head())

dim_dq = pd.DataFrame([{
    "dim_items": dim["item_id"].nunique(),
    "valid_expected_lifetime_items": count_true(dim["has_valid_expected_lifetime"]),
    "valid_lifetime_for_analysis_items": count_true(dim["is_valid_lifetime_for_analysis"]),
    "main_package_lifetime_items": count_true(dim["is_main_package_lifetime"]),
    "expected_lifetime_gt100_items": int((dim["expected_lifetime_days"] > 100).sum()),
}])

display_top(dim_dq, title="dim_listing DQ")
save_csv(dim_dq, "dim_listing_dq.csv")

# ============================================================
# 4. SCAN fact_listing_snapshot -> item-level lifecycle
# ============================================================

if os.path.exists(snapshot_item_path):
    print("[SKIP] snapshot item summary exists.")
    snapshot_item = pd.read_parquet(snapshot_item_path)

else:
    print("[BUILD] scanning fact_listing_snapshot...")

    snap_ds = ds.dataset(SNAPSHOT_PATH, format="parquet")
    available_cols = set(snap_ds.schema.names)

    wanted_cols = [
        "item_id",
        "date",
        "views_24h",
        "contacts_24h",
        "listing_age_days",
    ]

    cols = [c for c in wanted_cols if c in available_cols]

    if "item_id" not in cols:
        raise ValueError("fact_listing_snapshot thiếu item_id.")

    partials = []

    dq = {
        "total_rows": 0,
        "missing_item_id_rows": 0,
        "missing_date_rows": 0,
        "missing_age_rows": 0,
        "negative_age_rows": 0,
        "valid_rows": 0,
        "missing_views_rows": 0,
        "missing_contacts_rows": 0,
    }

    for i, batch in enumerate(
        snap_ds.to_batches(columns=cols, batch_size=BATCH_SIZE_SNAPSHOT),
        start=1
    ):
        df = batch.to_pandas()

        for c in wanted_cols:
            if c not in df.columns:
                df[c] = np.nan

        dq["total_rows"] += len(df)
        dq["missing_item_id_rows"] += int(df["item_id"].isna().sum())
        dq["missing_date_rows"] += int(df["date"].isna().sum())

        df = df[df["item_id"].notna()].copy()

        df["item_id"] = df["item_id"].astype(str)
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
        df["views_24h"] = pd.to_numeric(df["views_24h"], errors="coerce")
        df["contacts_24h"] = pd.to_numeric(df["contacts_24h"], errors="coerce")
        df["listing_age_days"] = pd.to_numeric(df["listing_age_days"], errors="coerce")

        dq["missing_age_rows"] += int(df["listing_age_days"].isna().sum())
        dq["negative_age_rows"] += int((df["listing_age_days"] < 0).sum())
        dq["missing_views_rows"] += int(df["views_24h"].isna().sum())
        dq["missing_contacts_rows"] += int(df["contacts_24h"].isna().sum())

        # Chỉ bỏ age âm hoặc missing age. Không bỏ NaN views/contact.
        df = df[
            df["date"].notna()
            & df["listing_age_days"].notna()
            & (df["listing_age_days"] >= 0)
        ].copy()

        dq["valid_rows"] += len(df)

        if len(df) == 0:
            del df
            gc.collect()
            continue

        df["views_clean"] = df["views_24h"].fillna(0)
        df["contacts_clean"] = df["contacts_24h"].fillna(0)
        df["has_contact"] = (df["contacts_clean"] > 0).astype(int)

        df["lead_age"] = np.where(df["has_contact"] == 1, df["listing_age_days"], np.nan)
        df["lead_date"] = df["date"].where(df["has_contact"] == 1, pd.NaT)

        agg = (
            df.groupby("item_id", sort=False, dropna=False)
            .agg(
                snapshot_rows=("item_id", "size"),
                snapshot_days=("date", "nunique"),
                first_snapshot_date=("date", "min"),
                last_snapshot_date=("date", "max"),
                min_listing_age_days=("listing_age_days", "min"),
                max_listing_age_days=("listing_age_days", "max"),
                total_views=("views_clean", "sum"),
                total_contacts=("contacts_clean", "sum"),
                contact_active_days=("has_contact", "sum"),
                first_contact_date_snapshot=("lead_date", "min"),
                last_contact_date_snapshot=("lead_date", "max"),
                first_contact_age_snapshot=("lead_age", "min"),
                last_contact_age_snapshot=("lead_age", "max"),
                missing_views_rows=("views_24h", lambda x: x.isna().sum()),
                missing_contacts_rows=("contacts_24h", lambda x: x.isna().sum()),
            )
            .reset_index()
        )

        partials.append(agg)

        if i % 10 == 0:
            print("snapshot batches:", i)

        del df, agg
        gc.collect()

    if len(partials) == 0:
        raise ValueError("Không có valid rows trong fact_listing_snapshot.")

    print("[COMBINE] snapshot item summary...")

    all_partials = pd.concat(partials, ignore_index=True)

    snapshot_item = (
        all_partials.groupby("item_id", sort=False, dropna=False)
        .agg(
            snapshot_rows=("snapshot_rows", "sum"),
            snapshot_days=("snapshot_days", "sum"),
            first_snapshot_date=("first_snapshot_date", "min"),
            last_snapshot_date=("last_snapshot_date", "max"),
            min_listing_age_days=("min_listing_age_days", "min"),
            max_listing_age_days=("max_listing_age_days", "max"),
            total_views=("total_views", "sum"),
            total_contacts=("total_contacts", "sum"),
            contact_active_days=("contact_active_days", "sum"),
            first_contact_date_snapshot=("first_contact_date_snapshot", "min"),
            last_contact_date_snapshot=("last_contact_date_snapshot", "max"),
            first_contact_age_snapshot=("first_contact_age_snapshot", "min"),
            last_contact_age_snapshot=("last_contact_age_snapshot", "max"),
            missing_views_rows=("missing_views_rows", "sum"),
            missing_contacts_rows=("missing_contacts_rows", "sum"),
        )
        .reset_index()
    )

    snapshot_item["snapshot_cr"] = (
        snapshot_item["total_contacts"]
        / snapshot_item["total_views"].replace(0, np.nan)
    )

    snapshot_item["contacts_per_1000_views"] = snapshot_item["snapshot_cr"] * 1000

    snapshot_item.to_parquet(snapshot_item_path, index=False)

    dq_df = pd.DataFrame([dq])
    display_top(dq_df, title="Snapshot DQ")
    save_csv(dq_df, "snapshot_dq.csv")

    del all_partials, partials
    gc.collect()

print("snapshot_item shape:", snapshot_item.shape)
display(snapshot_item.head())

# ============================================================
# 5. JOIN lifecycle with dim_listing
# ============================================================

if os.path.exists(item_lifecycle_path):
    print("[SKIP] item lifecycle exists.")
    item_lifecycle = pd.read_parquet(item_lifecycle_path)

else:
    print("[BUILD] joining snapshot summary with dim_listing...")

    item_lifecycle = snapshot_item.merge(
        dim,
        on="item_id",
        how="left",
        indicator=True
    )

    item_lifecycle["is_found_in_dim"] = item_lifecycle["_merge"].eq("both")

    item_lifecycle["observed_lifetime_days"] = (
        item_lifecycle["last_snapshot_date"] - item_lifecycle["posted_date"]
    ).dt.days + 1

    item_lifecycle["days_removed_before_expected"] = (
        item_lifecycle["expected_expired_date"] - item_lifecycle["last_snapshot_date"]
    ).dt.days

    item_lifecycle["days_after_expected"] = (
        item_lifecycle["last_snapshot_date"] - item_lifecycle["expected_expired_date"]
    ).dt.days

    item_lifecycle["lifetime_ratio"] = (
        item_lifecycle["observed_lifetime_days"]
        / item_lifecycle["expected_lifetime_days"].replace(0, np.nan)
    )

    item_lifecycle["early_removed"] = (
        item_lifecycle["days_removed_before_expected"] >= EARLY_REMOVED_DAYS
    )

    item_lifecycle["very_early_removed"] = (
        item_lifecycle["lifetime_ratio"] <= VERY_EARLY_LIFETIME_RATIO
    )

    item_lifecycle["normal_expired"] = (
        item_lifecycle["days_removed_before_expected"].abs()
        <= NORMAL_EXPIRED_TOLERANCE_DAYS
    )

    item_lifecycle["still_seen_after_expected"] = (
        item_lifecycle["days_removed_before_expected"]
        < -NORMAL_EXPIRED_TOLERANCE_DAYS
    )

    item_lifecycle["valid_for_behavior_analysis"] = (
        item_lifecycle["is_found_in_dim"]
        & item_lifecycle["is_valid_lifetime_for_analysis"].fillna(False)
        & item_lifecycle["posted_date"].notna()
        & item_lifecycle["expected_expired_date"].notna()
        & item_lifecycle["last_snapshot_date"].notna()
    )

    item_lifecycle.to_parquet(item_lifecycle_path, index=False)

print("item_lifecycle shape:", item_lifecycle.shape)
display(item_lifecycle.head())

valid_mask = item_lifecycle["valid_for_behavior_analysis"].fillna(False).astype(bool)

early_removed_mask = (
    valid_mask
    & item_lifecycle["early_removed"].fillna(False).astype(bool)
)

normal_expired_mask = (
    valid_mask
    & item_lifecycle["normal_expired"].fillna(False).astype(bool)
)

still_seen_after_expected_mask = (
    valid_mask
    & item_lifecycle["still_seen_after_expected"].fillna(False).astype(bool)
)

lifecycle_summary = pd.DataFrame([{
    "items_in_snapshot": snapshot_item["item_id"].nunique(),
    "items_found_in_dim": count_true(item_lifecycle["is_found_in_dim"]),
    "valid_for_behavior_analysis": count_true(item_lifecycle["valid_for_behavior_analysis"]),

    "early_removed_items": count_true(early_removed_mask),
    "normal_expired_items": count_true(normal_expired_mask),
    "still_seen_after_expected_items": count_true(still_seen_after_expected_mask),

    "early_removed_rate_among_valid": count_true(early_removed_mask) / max(count_true(valid_mask), 1),
    "normal_expired_rate_among_valid": count_true(normal_expired_mask) / max(count_true(valid_mask), 1),
    "still_seen_after_expected_rate_among_valid": count_true(still_seen_after_expected_mask) / max(count_true(valid_mask), 1),
}])

display_top(lifecycle_summary, title="Lifecycle summary")
save_csv(lifecycle_summary, "lifecycle_summary.csv")

# ============================================================
# 6. MEMORY-SAFE SCAN fact_post_contact_interactions
# ============================================================

try:
    import duckdb
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "duckdb"])
    import duckdb

if os.path.exists(user_item_path) and os.path.exists(item_chat_path):
    print("[SKIP] user_item and item_chat already exist.")
    item_chat = pd.read_parquet(item_chat_path)

else:
    print("[BUILD] scanning fact_post_contact_interactions memory-safe...")

    # Keep only valid items to reduce interaction rows.
    item_time_map = item_lifecycle[
        [
            "item_id",
            "posted_date",
            "last_snapshot_date",
            "expected_expired_date",
            "expected_lifetime_days",
            "valid_for_behavior_analysis",
        ]
    ].copy()

    item_time_map = item_time_map[
        item_time_map["valid_for_behavior_analysis"].fillna(False)
    ].copy()

    item_time_map["item_id"] = item_time_map["item_id"].astype(str)

    valid_item_set = set(item_time_map["item_id"].unique())

    print("Valid items for interaction scan:", len(valid_item_set))

    inter_ds = ds.dataset(INTERACTION_PATH, format="parquet")
    available_cols = set(inter_ds.schema.names)

    wanted_cols = [
        "user_id",
        "item_id",
        "date",
        "adview_count",
        "lead_count",
        "chat_message_count",
        "chat_turn_count",
        "chat_lead",
        "purchased",
        "category",
    ]

    cols = [c for c in wanted_cols if c in available_cols]

    if "user_id" not in cols:
        raise ValueError("fact_post_contact_interactions thiếu user_id.")
    if "item_id" not in cols:
        raise ValueError("fact_post_contact_interactions thiếu item_id.")
    if "date" not in cols:
        raise ValueError("fact_post_contact_interactions thiếu date.")

    numeric_cols = [
        "adview_count",
        "lead_count",
        "chat_message_count",
        "chat_turn_count",
        "chat_lead",
    ]

    dq = {
        "total_rows": 0,
        "rows_after_valid_item_filter": 0,
        "valid_rows": 0,
        "partial_files": 0,
        "missing_user_id_rows": 0,
        "missing_item_id_rows": 0,
        "missing_date_rows": 0,
    }

    partial_id = 0

    for i, batch in enumerate(
        inter_ds.to_batches(columns=cols, batch_size=BATCH_SIZE_INTERACTION),
        start=1
    ):
        df = batch.to_pandas()

        for c in wanted_cols:
            if c not in df.columns:
                df[c] = np.nan

        dq["total_rows"] += len(df)
        dq["missing_user_id_rows"] += int(df["user_id"].isna().sum())
        dq["missing_item_id_rows"] += int(df["item_id"].isna().sum())
        dq["missing_date_rows"] += int(df["date"].isna().sum())

        df = df[
            df["user_id"].notna()
            & df["item_id"].notna()
            & df["date"].notna()
        ].copy()

        if len(df) == 0:
            del df
            gc.collect()
            continue

        df["user_id"] = df["user_id"].astype(str)
        df["item_id"] = df["item_id"].astype(str)

        # Critical memory reduction.
        df = df[df["item_id"].isin(valid_item_set)].copy()

        dq["rows_after_valid_item_filter"] += len(df)

        if len(df) == 0:
            del df
            gc.collect()
            continue

        df["date"] = pd.to_datetime(df["date"], errors="coerce")

        df = df[
            df["date"].notna()
            & (df["user_id"].str.strip() != "")
            & (df["item_id"].str.strip() != "")
        ].copy()

        dq["valid_rows"] += len(df)

        if len(df) == 0:
            del df
            gc.collect()
            continue

        for c in numeric_cols:
            df[c] = pd.to_numeric(df[c], errors="coerce")
            df[c + "_clean"] = df[c].fillna(0)

        df["purchased_clean"] = bool_to_int_series(df["purchased"])

        df["has_chat_signal"] = (
            (df["chat_message_count_clean"] > 0)
            | (df["chat_turn_count_clean"] > 0)
            | (df["chat_lead_clean"] > 0)
        ).astype(int)

        df["has_lead_signal"] = (
            (df["lead_count_clean"] > 0)
            | (df["chat_lead_clean"] > 0)
        ).astype(int)

        # Join only after filtering.
        df = df.merge(
            item_time_map,
            on="item_id",
            how="left"
        )

        df["days_since_posted"] = (
            df["date"] - df["posted_date"]
        ).dt.days

        df["days_to_removal_at_row"] = (
            df["last_snapshot_date"] - df["date"]
        ).dt.days

        df["is_early_3d_row"] = (
            df["days_since_posted"].notna()
            & (df["days_since_posted"] >= 0)
            & (df["days_since_posted"] < EARLY_CHAT_WINDOW_3D)
        )

        df["is_early_7d_row"] = (
            df["days_since_posted"].notna()
            & (df["days_since_posted"] >= 0)
            & (df["days_since_posted"] < EARLY_CHAT_WINDOW_7D)
        )

        df["is_last_5d_row"] = (
            df["days_to_removal_at_row"].notna()
            & (df["days_to_removal_at_row"] >= 0)
            & (df["days_to_removal_at_row"] < LAST_CHAT_WINDOW_5D)
        )

        df["chat_date"] = df["date"].where(df["has_chat_signal"] == 1, pd.NaT)

        df["early3_chat_flag"] = (
            (df["has_chat_signal"] == 1)
            & df["is_early_3d_row"]
        ).astype(int)

        df["early7_chat_flag"] = (
            (df["has_chat_signal"] == 1)
            & df["is_early_7d_row"]
        ).astype(int)

        df["last5_chat_flag"] = (
            (df["has_chat_signal"] == 1)
            & df["is_last_5d_row"]
        ).astype(int)

        df["early3_chat_turns"] = np.where(
            df["early3_chat_flag"] == 1,
            df["chat_turn_count_clean"],
            0
        )

        df["early7_chat_turns"] = np.where(
            df["early7_chat_flag"] == 1,
            df["chat_turn_count_clean"],
            0
        )

        df["last5_chat_turns"] = np.where(
            df["last5_chat_flag"] == 1,
            df["chat_turn_count_clean"],
            0
        )

        df["early3_chat_messages"] = np.where(
            df["early3_chat_flag"] == 1,
            df["chat_message_count_clean"],
            0
        )

        df["early7_chat_messages"] = np.where(
            df["early7_chat_flag"] == 1,
            df["chat_message_count_clean"],
            0
        )

        df["last5_chat_messages"] = np.where(
            df["last5_chat_flag"] == 1,
            df["chat_message_count_clean"],
            0
        )

        # Batch-level user-item aggregate.
        ui = (
            df.groupby(["item_id", "user_id"], sort=False, dropna=False)
            .agg(
                ui_first_date=("date", "min"),
                ui_last_date=("date", "max"),
                ui_rows=("user_id", "size"),

                ui_adviews=("adview_count_clean", "sum"),
                ui_leads=("lead_count_clean", "sum"),
                ui_chat_messages=("chat_message_count_clean", "sum"),
                ui_chat_turns=("chat_turn_count_clean", "sum"),
                ui_chat_leads=("chat_lead_clean", "sum"),

                ui_has_chat=("has_chat_signal", "max"),
                ui_has_lead=("has_lead_signal", "max"),
                ui_purchased=("purchased_clean", "max"),

                ui_first_chat_date=("chat_date", "min"),
                ui_last_chat_date=("chat_date", "max"),

                ui_has_early3_chat=("early3_chat_flag", "max"),
                ui_has_early7_chat=("early7_chat_flag", "max"),
                ui_has_last5_chat=("last5_chat_flag", "max"),

                ui_early3_chat_turns=("early3_chat_turns", "sum"),
                ui_early7_chat_turns=("early7_chat_turns", "sum"),
                ui_last5_chat_turns=("last5_chat_turns", "sum"),

                ui_early3_chat_messages=("early3_chat_messages", "sum"),
                ui_early7_chat_messages=("early7_chat_messages", "sum"),
                ui_last5_chat_messages=("last5_chat_messages", "sum"),
            )
            .reset_index()
        )

        partial_id += 1
        part_path = os.path.join(partial_ui_dir, f"user_item_part_{partial_id:05d}.parquet")
        ui.to_parquet(part_path, index=False)

        dq["partial_files"] = partial_id

        if i % 10 == 0:
            print(
                "interaction batches:", i,
                "| raw rows:", dq["total_rows"],
                "| kept rows:", dq["rows_after_valid_item_filter"],
                "| partial files:", partial_id
            )

        del df, ui
        gc.collect()

    dq_df = pd.DataFrame([dq])
    display_top(dq_df, title="Interaction memory-safe DQ")
    save_csv(dq_df, "interaction_memory_safe_dq.csv")

    if partial_id == 0:
        raise ValueError(
            "Không tạo được partial user-item file nào. "
            "Có thể valid_item_set không match item_id trong fact_post_contact_interactions."
        )

    # ========================================================
    # DuckDB aggregate partial parquet files out-of-core
    # ========================================================

    print("[DUCKDB] Final user-item aggregate out-of-core...")

    duckdb_path = os.path.join(CACHE_DIR, "duckdb_work.duckdb")
    duckdb_temp_dir = os.path.join(CACHE_DIR, "duckdb_temp")
    os.makedirs(duckdb_temp_dir, exist_ok=True)

    if FORCE_REBUILD and os.path.exists(duckdb_path):
        os.remove(duckdb_path)

    con = duckdb.connect(database=duckdb_path)
    con.execute("PRAGMA threads=4")
    con.execute(f"PRAGMA memory_limit='{DUCKDB_MEMORY_LIMIT}'")

    try:
        con.execute(f"SET temp_directory='{duckdb_temp_dir}'")
    except Exception:
        pass

    partial_glob = os.path.join(partial_ui_dir, "*.parquet").replace("\\", "/")
    user_item_out = user_item_path.replace("\\", "/")
    item_chat_out = item_chat_path.replace("\\", "/")

    con.execute(f"""
        COPY (
            SELECT
                item_id,
                user_id,

                MIN(ui_first_date) AS ui_first_date,
                MAX(ui_last_date) AS ui_last_date,
                SUM(ui_rows) AS ui_rows,

                SUM(ui_adviews) AS ui_adviews,
                SUM(ui_leads) AS ui_leads,
                SUM(ui_chat_messages) AS ui_chat_messages,
                SUM(ui_chat_turns) AS ui_chat_turns,
                SUM(ui_chat_leads) AS ui_chat_leads,

                MAX(ui_has_chat) AS ui_has_chat,
                MAX(ui_has_lead) AS ui_has_lead,
                MAX(ui_purchased) AS ui_purchased,

                MIN(ui_first_chat_date) AS ui_first_chat_date,
                MAX(ui_last_chat_date) AS ui_last_chat_date,

                MAX(ui_has_early3_chat) AS ui_has_early3_chat,
                MAX(ui_has_early7_chat) AS ui_has_early7_chat,
                MAX(ui_has_last5_chat) AS ui_has_last5_chat,

                SUM(ui_early3_chat_turns) AS ui_early3_chat_turns,
                SUM(ui_early7_chat_turns) AS ui_early7_chat_turns,
                SUM(ui_last5_chat_turns) AS ui_last5_chat_turns,

                SUM(ui_early3_chat_messages) AS ui_early3_chat_messages,
                SUM(ui_early7_chat_messages) AS ui_early7_chat_messages,
                SUM(ui_last5_chat_messages) AS ui_last5_chat_messages

            FROM read_parquet('{partial_glob}')
            GROUP BY item_id, user_id
        )
        TO '{user_item_out}'
        (FORMAT PARQUET)
    """)

    print("[DUCKDB] Build item-level chat quality summary...")

    con.execute(f"""
        COPY (
            WITH ui AS (
                SELECT *
                FROM read_parquet('{user_item_out}')
            ),

            item_chat_base AS (
                SELECT
                    item_id,

                    COUNT(DISTINCT user_id) AS interaction_users,
                    SUM(ui_rows) AS interaction_rows,

                    SUM(ui_adviews) AS total_adviews_interaction,
                    SUM(ui_leads) AS total_leads_interaction,

                    SUM(ui_has_chat) AS chat_users,
                    SUM(ui_has_lead) AS lead_users,
                    SUM(ui_purchased) AS purchased_users,

                    MAX(ui_purchased) AS is_predicted_sold,

                    SUM(ui_chat_messages) AS total_chat_messages,
                    SUM(ui_chat_turns) AS total_chat_turns,
                    SUM(ui_chat_leads) AS total_chat_leads,

                    MIN(ui_first_chat_date) AS first_chat_date,
                    MAX(ui_last_chat_date) AS last_chat_date,

                    SUM(
                        CASE
                            WHEN ui_has_chat = 1 AND ui_chat_turns >= {DEEP_CHAT_TURNS}
                            THEN 1 ELSE 0
                        END
                    ) AS deep_chat_users,

                    SUM(
                        CASE
                            WHEN ui_has_chat = 1 AND ui_chat_turns <= {SHALLOW_CHAT_TURNS}
                            THEN 1 ELSE 0
                        END
                    ) AS shallow_chat_users,

                    SUM(ui_has_early3_chat) AS early3_chat_users,
                    SUM(ui_has_early7_chat) AS early7_chat_users,
                    SUM(ui_has_last5_chat) AS last5_chat_users,

                    SUM(ui_early3_chat_turns) AS early3_chat_turns,
                    SUM(ui_early7_chat_turns) AS early7_chat_turns,
                    SUM(ui_last5_chat_turns) AS last5_chat_turns,

                    SUM(ui_early3_chat_messages) AS early3_chat_messages,
                    SUM(ui_early7_chat_messages) AS early7_chat_messages,
                    SUM(ui_last5_chat_messages) AS last5_chat_messages

                FROM ui
                GROUP BY item_id
            )

            SELECT
                *,

                total_chat_turns / NULLIF(chat_users, 0) AS avg_turns_per_chat_user,
                total_chat_messages / NULLIF(chat_users, 0) AS avg_messages_per_chat_user,
                total_chat_messages / NULLIF(total_chat_turns, 0) AS messages_per_turn,

                deep_chat_users / NULLIF(chat_users, 0) AS deep_chat_user_rate,
                shallow_chat_users / NULLIF(chat_users, 0) AS shallow_chat_user_rate,

                early3_chat_turns / NULLIF(early3_chat_users, 0) AS early3_avg_turns_per_chat_user,
                early7_chat_turns / NULLIF(early7_chat_users, 0) AS early7_avg_turns_per_chat_user,
                last5_chat_turns / NULLIF(last5_chat_users, 0) AS last5_avg_turns_per_chat_user,

                lead_users / NULLIF(interaction_users, 0) AS lead_user_rate

            FROM item_chat_base
        )
        TO '{item_chat_out}'
        (FORMAT PARQUET)
    """)

    con.close()

    item_chat = pd.read_parquet(item_chat_path)

print("item_chat shape:", item_chat.shape)
display(item_chat.head())

# ============================================================
# 7. JOIN behavior features + classify
# ============================================================

if os.path.exists(item_behavior_path) and not FORCE_REBUILD:
    print("[SKIP] item behavior features exists.")
    item_behavior = pd.read_parquet(item_behavior_path)

else:
    print("[BUILD] item behavior features and classification...")

    item_behavior = item_lifecycle.merge(
        item_chat,
        on="item_id",
        how="left"
    )

    count_cols = [
        "interaction_users",
        "interaction_rows",
        "total_adviews_interaction",
        "total_leads_interaction",
        "chat_users",
        "lead_users",
        "purchased_users",
        "is_predicted_sold",
        "total_chat_messages",
        "total_chat_turns",
        "total_chat_leads",
        "deep_chat_users",
        "shallow_chat_users",
        "early3_chat_users",
        "early7_chat_users",
        "last5_chat_users",
        "early3_chat_turns",
        "early7_chat_turns",
        "last5_chat_turns",
        "early3_chat_messages",
        "early7_chat_messages",
        "last5_chat_messages",
    ]

    for c in count_cols:
        if c in item_behavior.columns:
            item_behavior[c] = item_behavior[c].fillna(0)

    item_behavior["is_predicted_sold"] = (
        item_behavior["is_predicted_sold"]
        .fillna(0)
        .astype(int)
    )

    item_behavior["days_to_first_chat"] = (
        item_behavior["first_chat_date"] - item_behavior["posted_date"]
    ).dt.days

    item_behavior["days_from_last_chat_to_removal"] = (
        item_behavior["last_snapshot_date"] - item_behavior["last_chat_date"]
    ).dt.days

    item_behavior["has_any_chat"] = item_behavior["chat_users"] > 0

    item_behavior["has_deep_chat_signal"] = (
        (item_behavior["avg_turns_per_chat_user"].fillna(0) >= REAL_FAST_CLOSE_MIN_AVG_TURNS)
        | (item_behavior["deep_chat_user_rate"].fillna(0) >= REAL_FAST_CLOSE_MIN_DEEP_RATE)
        | (item_behavior["total_chat_turns"].fillna(0) >= DEEP_CHAT_TURNS)
    )

    # Strict: có purchased pseudo-label.
    item_behavior["is_real_fast_close_strict"] = (
        item_behavior["valid_for_behavior_analysis"].fillna(False)
        & item_behavior["early_removed"].fillna(False)
        & item_behavior["days_to_first_chat"].notna()
        & (item_behavior["days_to_first_chat"] <= REAL_FAST_CLOSE_MAX_DAYS_TO_FIRST_CHAT)
        & item_behavior["has_deep_chat_signal"].fillna(False)
        & (item_behavior["is_predicted_sold"] == 1)
    )

    # Proxy: không bắt buộc purchased=True, nhưng phải có chat sâu + lead/contact.
    item_behavior["is_real_fast_close_proxy"] = (
        item_behavior["valid_for_behavior_analysis"].fillna(False)
        & item_behavior["early_removed"].fillna(False)
        & item_behavior["days_to_first_chat"].notna()
        & (item_behavior["days_to_first_chat"] <= REAL_FAST_CLOSE_MAX_DAYS_TO_FIRST_CHAT)
        & item_behavior["has_deep_chat_signal"].fillna(False)
        & (
            (item_behavior["is_predicted_sold"] == 1)
            | (item_behavior["total_contacts"].fillna(0) > 0)
            | (item_behavior["total_leads_interaction"].fillna(0) > 0)
        )
    )

    item_behavior["is_abandoned_disappointed"] = (
        item_behavior["valid_for_behavior_analysis"].fillna(False)
        & item_behavior["early_removed"].fillna(False)
        & (item_behavior["chat_users"].fillna(0) <= ABANDONED_MAX_CHAT_USERS)
        & (item_behavior["total_contacts"].fillna(0) <= ABANDONED_MAX_TOTAL_CONTACTS)
        & (item_behavior["total_leads_interaction"].fillna(0) <= ABANDONED_MAX_TOTAL_CONTACTS)
        & (item_behavior["is_predicted_sold"] == 0)
    )

    item_behavior["is_lead_bait_ghosting"] = (
        item_behavior["valid_for_behavior_analysis"].fillna(False)
        & (item_behavior["chat_users"].fillna(0) >= GHOST_MIN_CHAT_USERS)
        & (item_behavior["avg_turns_per_chat_user"].fillna(0) <= GHOST_MAX_AVG_TURNS)
        & (item_behavior["shallow_chat_user_rate"].fillna(0) >= GHOST_MIN_SHALLOW_RATE)
        & (item_behavior["is_predicted_sold"] == 0)
    )

    item_behavior["is_endgame_ghosting"] = (
        item_behavior["valid_for_behavior_analysis"].fillna(False)
        & (item_behavior["last5_chat_users"].fillna(0) >= ENDGAME_GHOST_MIN_LAST5_CHAT_USERS)
        & (item_behavior["last5_avg_turns_per_chat_user"].fillna(0) <= ENDGAME_GHOST_MAX_LAST5_AVG_TURNS)
        & (item_behavior["is_predicted_sold"] == 0)
    )

    item_behavior["item_behavior_bucket"] = np.select(
        [
            item_behavior["is_real_fast_close_strict"],
            item_behavior["is_real_fast_close_proxy"],
            item_behavior["is_lead_bait_ghosting"],
            item_behavior["is_endgame_ghosting"],
            item_behavior["is_abandoned_disappointed"],
            item_behavior["normal_expired"].fillna(False),
            item_behavior["still_seen_after_expected"].fillna(False),
        ],
        [
            "real_fast_close_strict",
            "real_fast_close_proxy",
            "lead_bait_ghosting",
            "endgame_ghosting",
            "abandoned_disappointed",
            "normal_expired",
            "still_seen_after_expected",
        ],
        default="unclear_other"
    )

    item_behavior.to_parquet(item_behavior_path, index=False)

print("item_behavior shape:", item_behavior.shape)
display(item_behavior.head())

# ============================================================
# 8. EXPORT ITEM-LEVEL EVIDENCE
# ============================================================

analysis_items = item_behavior[
    item_behavior["valid_for_behavior_analysis"].fillna(False)
].copy()

preferred_cols = [
    "item_id",
    "seller_id",
    "seller_type",
    "ad_type",
    "category",
    "category_name",
    "title",
    "city_name",
    "district_name",
    "ward_name",
    "price_bucket",
    "images_count",
    "area_sqm",
    "bedrooms",
    "bathrooms",
    "ad_status_clean",

    "posted_date",
    "expected_expired_date",
    "expected_lifetime_days",
    "is_main_package_lifetime",
    "is_valid_lifetime_for_analysis",

    "first_snapshot_date",
    "last_snapshot_date",
    "observed_lifetime_days",
    "days_removed_before_expected",
    "days_after_expected",
    "lifetime_ratio",
    "early_removed",
    "very_early_removed",
    "normal_expired",
    "still_seen_after_expected",

    "total_views",
    "total_contacts",
    "snapshot_cr",
    "contacts_per_1000_views",

    "interaction_users",
    "chat_users",
    "lead_users",
    "purchased_users",
    "is_predicted_sold",

    "total_adviews_interaction",
    "total_leads_interaction",
    "total_chat_messages",
    "total_chat_turns",
    "total_chat_leads",

    "first_chat_date",
    "last_chat_date",
    "days_to_first_chat",
    "days_from_last_chat_to_removal",

    "avg_turns_per_chat_user",
    "avg_messages_per_chat_user",
    "messages_per_turn",
    "deep_chat_users",
    "shallow_chat_users",
    "deep_chat_user_rate",
    "shallow_chat_user_rate",

    "early3_chat_users",
    "early7_chat_users",
    "last5_chat_users",
    "early3_avg_turns_per_chat_user",
    "early7_avg_turns_per_chat_user",
    "last5_avg_turns_per_chat_user",

    "has_deep_chat_signal",
    "is_real_fast_close_strict",
    "is_real_fast_close_proxy",
    "is_abandoned_disappointed",
    "is_lead_bait_ghosting",
    "is_endgame_ghosting",
    "item_behavior_bucket",
]

existing_cols = [c for c in preferred_cols if c in analysis_items.columns]

item_behavior_export = analysis_items[existing_cols].copy()
save_csv(item_behavior_export, "item_behavior_features.csv")

early_removed_items = analysis_items[
    analysis_items["early_removed"].fillna(False)
].copy()

real_fast_close_candidates = analysis_items[
    analysis_items["item_behavior_bucket"].isin([
        "real_fast_close_strict",
        "real_fast_close_proxy",
    ])
].copy()

abandoned_candidates = analysis_items[
    analysis_items["item_behavior_bucket"].eq("abandoned_disappointed")
].copy()

lead_bait_candidates = analysis_items[
    analysis_items["item_behavior_bucket"].eq("lead_bait_ghosting")
].copy()

endgame_ghosting_candidates = analysis_items[
    analysis_items["item_behavior_bucket"].eq("endgame_ghosting")
].copy()

save_csv(
    early_removed_items[existing_cols],
    "early_removed_diagnosis.csv"
)

save_csv(
    real_fast_close_candidates[existing_cols].sort_values(
        ["is_real_fast_close_strict", "total_chat_turns", "chat_users"],
        ascending=[False, False, False]
    ),
    "real_fast_close_candidates.csv"
)

save_csv(
    abandoned_candidates[existing_cols].sort_values(
        ["days_removed_before_expected", "total_views"],
        ascending=[False, True]
    ),
    "abandoned_disappointed_candidates.csv"
)

save_csv(
    lead_bait_candidates[existing_cols].sort_values(
        ["chat_users", "shallow_chat_user_rate", "total_views"],
        ascending=[False, False, False]
    ),
    "lead_bait_ghosting_candidates.csv"
)

save_csv(
    endgame_ghosting_candidates[existing_cols].sort_values(
        ["last5_chat_users", "last5_avg_turns_per_chat_user"],
        ascending=[False, True]
    ),
    "endgame_ghosting_candidates.csv"
)

display_top(real_fast_close_candidates[existing_cols], 20, "Real fast close candidates")
display_top(abandoned_candidates[existing_cols], 20, "Abandoned / disappointed candidates")
display_top(lead_bait_candidates[existing_cols], 20, "Lead-bait / ghosting candidates")
display_top(endgame_ghosting_candidates[existing_cols], 20, "End-game ghosting candidates")

# ============================================================
# 9. BUCKET SUMMARY
# ============================================================

bucket_summary = (
    analysis_items.groupby("item_behavior_bucket", dropna=False)
    .agg(
        items=("item_id", "nunique"),
        sellers=("seller_id", "nunique"),

        total_views=("total_views", "sum"),
        total_contacts=("total_contacts", "sum"),
        total_chat_users=("chat_users", "sum"),
        total_chat_turns=("total_chat_turns", "sum"),

        median_expected_lifetime_days=("expected_lifetime_days", "median"),
        median_observed_lifetime_days=("observed_lifetime_days", "median"),
        median_days_removed_before_expected=("days_removed_before_expected", "median"),
        median_lifetime_ratio=("lifetime_ratio", "median"),

        median_chat_users=("chat_users", "median"),
        median_avg_turns_per_chat_user=("avg_turns_per_chat_user", "median"),
        median_deep_chat_user_rate=("deep_chat_user_rate", "median"),
        median_shallow_chat_user_rate=("shallow_chat_user_rate", "median"),

        predicted_sold_rate=("is_predicted_sold", "mean"),
        early_removed_rate=("early_removed", "mean"),
    )
    .reset_index()
)

bucket_summary["item_share"] = bucket_summary["items"] / bucket_summary["items"].sum()
bucket_summary["snapshot_cr"] = (
    bucket_summary["total_contacts"]
    / bucket_summary["total_views"].replace(0, np.nan)
)
bucket_summary["contacts_per_1000_views"] = bucket_summary["snapshot_cr"] * 1000

bucket_summary = bucket_summary.sort_values("items", ascending=False)

display_top(bucket_summary, 20, "Behavior bucket summary")
save_csv(bucket_summary, "item_behavior_bucket_summary.csv")

# ============================================================
# 10. SELLER_TYPE SUMMARY: agent vs private
# ============================================================

seller_type_behavior = (
    analysis_items.groupby(["seller_type", "item_behavior_bucket"], dropna=False)
    .agg(
        items=("item_id", "nunique"),
        sellers=("seller_id", "nunique"),

        total_views=("total_views", "sum"),
        total_contacts=("total_contacts", "sum"),
        total_chat_users=("chat_users", "sum"),
        total_chat_turns=("total_chat_turns", "sum"),

        median_observed_lifetime_days=("observed_lifetime_days", "median"),
        median_days_removed_before_expected=("days_removed_before_expected", "median"),
        median_lifetime_ratio=("lifetime_ratio", "median"),

        median_chat_users=("chat_users", "median"),
        median_avg_turns_per_chat_user=("avg_turns_per_chat_user", "median"),
        median_deep_chat_user_rate=("deep_chat_user_rate", "median"),
        median_shallow_chat_user_rate=("shallow_chat_user_rate", "median"),

        predicted_sold_rate=("is_predicted_sold", "mean"),
    )
    .reset_index()
)

seller_type_behavior["item_share_within_seller_type"] = (
    seller_type_behavior["items"]
    / seller_type_behavior.groupby("seller_type")["items"].transform("sum")
)

seller_type_behavior["snapshot_cr"] = (
    seller_type_behavior["total_contacts"]
    / seller_type_behavior["total_views"].replace(0, np.nan)
)

seller_type_behavior["contacts_per_1000_views"] = seller_type_behavior["snapshot_cr"] * 1000

seller_type_behavior = seller_type_behavior.sort_values(
    ["seller_type", "items"],
    ascending=[True, False]
)

display_top(seller_type_behavior, 80, "Seller type x behavior bucket")
save_csv(seller_type_behavior, "seller_type_behavior_summary.csv")

seller_type_overview = (
    analysis_items.groupby("seller_type", dropna=False)
    .agg(
        items=("item_id", "nunique"),
        sellers=("seller_id", "nunique"),

        early_removed_rate=("early_removed", "mean"),
        normal_expired_rate=("normal_expired", "mean"),

        real_fast_close_strict_rate=("is_real_fast_close_strict", "mean"),
        real_fast_close_proxy_rate=("is_real_fast_close_proxy", "mean"),
        abandoned_rate=("is_abandoned_disappointed", "mean"),
        lead_bait_ghosting_rate=("is_lead_bait_ghosting", "mean"),
        endgame_ghosting_rate=("is_endgame_ghosting", "mean"),

        total_views=("total_views", "sum"),
        total_contacts=("total_contacts", "sum"),
        total_chat_users=("chat_users", "sum"),
        total_chat_turns=("total_chat_turns", "sum"),

        median_chat_users=("chat_users", "median"),
        median_avg_turns_per_chat_user=("avg_turns_per_chat_user", "median"),
        median_deep_chat_user_rate=("deep_chat_user_rate", "median"),
        median_shallow_chat_user_rate=("shallow_chat_user_rate", "median"),

        predicted_sold_rate=("is_predicted_sold", "mean"),
    )
    .reset_index()
)

seller_type_overview["item_share"] = seller_type_overview["items"] / seller_type_overview["items"].sum()
seller_type_overview["traffic_share"] = seller_type_overview["total_views"] / seller_type_overview["total_views"].sum()
seller_type_overview["contact_share"] = seller_type_overview["total_contacts"] / seller_type_overview["total_contacts"].sum()

seller_type_overview["snapshot_cr"] = (
    seller_type_overview["total_contacts"]
    / seller_type_overview["total_views"].replace(0, np.nan)
)

seller_type_overview["contacts_per_1000_views"] = seller_type_overview["snapshot_cr"] * 1000

seller_type_overview["avg_turns_per_chat_user_global"] = (
    seller_type_overview["total_chat_turns"]
    / seller_type_overview["total_chat_users"].replace(0, np.nan)
)

seller_type_overview = seller_type_overview.sort_values("items", ascending=False)

display_top(seller_type_overview, 20, "Seller type overview")
save_csv(seller_type_overview, "seller_type_overview.csv")

# ============================================================
# 11. SELLER-LEVEL REPUTATION
# ============================================================

seller_reputation = (
    analysis_items.groupby(["seller_id", "seller_type"], dropna=False)
    .agg(
        listings=("item_id", "nunique"),

        early_removed_rate=("early_removed", "mean"),
        normal_expired_rate=("normal_expired", "mean"),
        still_seen_after_expected_rate=("still_seen_after_expected", "mean"),

        real_fast_close_strict_rate=("is_real_fast_close_strict", "mean"),
        real_fast_close_proxy_rate=("is_real_fast_close_proxy", "mean"),
        abandoned_rate=("is_abandoned_disappointed", "mean"),
        lead_bait_ghosting_rate=("is_lead_bait_ghosting", "mean"),
        endgame_ghosting_rate=("is_endgame_ghosting", "mean"),

        total_views=("total_views", "sum"),
        total_contacts=("total_contacts", "sum"),
        total_chat_users=("chat_users", "sum"),
        total_chat_turns=("total_chat_turns", "sum"),

        median_observed_lifetime_days=("observed_lifetime_days", "median"),
        median_days_removed_before_expected=("days_removed_before_expected", "median"),
        median_lifetime_ratio=("lifetime_ratio", "median"),

        median_chat_users=("chat_users", "median"),
        median_avg_turns_per_chat_user=("avg_turns_per_chat_user", "median"),
        median_deep_chat_user_rate=("deep_chat_user_rate", "median"),
        median_shallow_chat_user_rate=("shallow_chat_user_rate", "median"),

        predicted_sold_rate=("is_predicted_sold", "mean"),
    )
    .reset_index()
)

seller_reputation["snapshot_cr"] = (
    seller_reputation["total_contacts"]
    / seller_reputation["total_views"].replace(0, np.nan)
)

seller_reputation["contacts_per_1000_views"] = seller_reputation["snapshot_cr"] * 1000

seller_reputation["avg_turns_per_chat_user_global"] = (
    seller_reputation["total_chat_turns"]
    / seller_reputation["total_chat_users"].replace(0, np.nan)
)

seller_reputation_min5 = seller_reputation[
    seller_reputation["listings"] >= SELLER_REPUTATION_MIN_LISTINGS
].copy()

seller_reputation_min5 = seller_reputation_min5.sort_values(
    ["lead_bait_ghosting_rate", "endgame_ghosting_rate", "listings"],
    ascending=[False, False, False]
)

display_top(
    seller_reputation_min5,
    50,
    f"Seller reputation, min {SELLER_REPUTATION_MIN_LISTINGS} listings"
)

save_csv(seller_reputation, "seller_reputation_all.csv")
save_csv(seller_reputation_min5, "seller_reputation_min5.csv")

# ============================================================
# 12. CATEGORY + SELLER_TYPE SUMMARY
# ============================================================

category_seller_summary = (
    analysis_items.groupby(
        ["category_name", "seller_type", "item_behavior_bucket"],
        dropna=False
    )
    .agg(
        items=("item_id", "nunique"),
        sellers=("seller_id", "nunique"),

        total_views=("total_views", "sum"),
        total_contacts=("total_contacts", "sum"),
        total_chat_users=("chat_users", "sum"),
        total_chat_turns=("total_chat_turns", "sum"),

        median_chat_users=("chat_users", "median"),
        median_avg_turns_per_chat_user=("avg_turns_per_chat_user", "median"),
        median_shallow_chat_user_rate=("shallow_chat_user_rate", "median"),

        predicted_sold_rate=("is_predicted_sold", "mean"),
    )
    .reset_index()
)

category_seller_summary["item_share_within_category_seller"] = (
    category_seller_summary["items"]
    / category_seller_summary.groupby(["category_name", "seller_type"])["items"].transform("sum")
)

category_seller_summary["snapshot_cr"] = (
    category_seller_summary["total_contacts"]
    / category_seller_summary["total_views"].replace(0, np.nan)
)

category_seller_summary["contacts_per_1000_views"] = category_seller_summary["snapshot_cr"] * 1000

category_seller_summary = category_seller_summary.sort_values(
    ["category_name", "seller_type", "items"],
    ascending=[True, True, False]
)

display_top(category_seller_summary, 100, "Category x seller type x behavior")
save_csv(category_seller_summary, "category_seller_behavior_summary.csv")

# ============================================================
# 13. INTERACTIVE PLOTLY CHARTS
# ============================================================

try:
    import plotly.express as px

    INTERACTIVE_DIR = os.path.join(CACHE_DIR, "interactive_charts")
    os.makedirs(INTERACTIVE_DIR, exist_ok=True)

    def show_and_save(fig, filename):
        fig.update_layout(
            template="plotly_white",
            hovermode="closest",
        )
        fig.show()

        path = os.path.join(INTERACTIVE_DIR, filename)
        fig.write_html(path)
        print("[SAVE HTML]", path)

    fig = px.bar(
        bucket_summary,
        x="item_behavior_bucket",
        y="items",
        title=f"Item behavior buckets<br>Total valid analysis items: {analysis_items['item_id'].nunique():,}",
        hover_data={
            "items": ":,",
            "item_share": ":.2%",
            "sellers": ":,",
            "median_observed_lifetime_days": ":.1f",
            "median_days_removed_before_expected": ":.1f",
            "median_chat_users": ":.1f",
            "median_avg_turns_per_chat_user": ":.2f",
            "median_shallow_chat_user_rate": ":.2%",
            "predicted_sold_rate": ":.2%",
            "contacts_per_1000_views": ":.2f",
        },
    )
    fig.update_layout(xaxis_tickangle=-30, height=600)
    show_and_save(fig, "item_behavior_bucket_summary.html")

    metric_cols = [
        "item_share",
        "traffic_share",
        "contact_share",
        "real_fast_close_proxy_rate",
        "abandoned_rate",
        "lead_bait_ghosting_rate",
        "endgame_ghosting_rate",
    ]

    seller_type_long = seller_type_overview.melt(
        id_vars=["seller_type"],
        value_vars=metric_cols,
        var_name="metric",
        value_name="value",
    )

    fig = px.bar(
        seller_type_long,
        x="seller_type",
        y="value",
        color="metric",
        barmode="group",
        title="Agent vs Private: supply/traffic/contact and behavior rates",
        hover_data={
            "seller_type": True,
            "metric": True,
            "value": ":.2%",
        },
    )
    fig.update_layout(height=600)
    show_and_save(fig, "seller_type_behavior_rates.html")

    fig = px.bar(
        seller_type_behavior,
        x="seller_type",
        y="item_share_within_seller_type",
        color="item_behavior_bucket",
        title="Behavior bucket composition within seller type",
        hover_data={
            "seller_type": True,
            "item_behavior_bucket": True,
            "items": ":,",
            "item_share_within_seller_type": ":.2%",
            "median_avg_turns_per_chat_user": ":.2f",
            "median_shallow_chat_user_rate": ":.2%",
            "predicted_sold_rate": ":.2%",
        },
    )
    fig.update_layout(height=600)
    show_and_save(fig, "seller_type_bucket_composition.html")

    scatter_df = analysis_items[
        analysis_items["chat_users"].fillna(0) > 0
    ].copy()

    if len(scatter_df) > 0:
        fig = px.scatter(
            scatter_df,
            x="chat_users",
            y="avg_turns_per_chat_user",
            color="seller_type",
            symbol="item_behavior_bucket",
            size="total_views",
            hover_data={
                "item_id": True,
                "title": True,
                "seller_type": True,
                "item_behavior_bucket": True,
                "chat_users": ":,",
                "avg_turns_per_chat_user": ":.2f",
                "shallow_chat_user_rate": ":.2%",
                "days_removed_before_expected": ":.1f",
                "is_predicted_sold": True,
                "total_views": ":,",
                "total_contacts": ":,",
            },
            title="Chat users vs conversation depth proxy",
        )
        fig.update_xaxes(type="log")
        fig.update_layout(height=700)
        show_and_save(fig, "chat_users_vs_avg_turns.html")

    print("\nInteractive charts folder:", INTERACTIVE_DIR)

except Exception as e:
    print("[SKIP Plotly charts] Error:", e)

# ============================================================
# 14. FINAL QUICK SUMMARY
# ============================================================

quick_summary = pd.DataFrame([{
    "total_dim_items": dim["item_id"].nunique(),
    "total_snapshot_items": snapshot_item["item_id"].nunique(),
    "valid_analysis_items": analysis_items["item_id"].nunique(),

    "early_removed_items": count_true(analysis_items["early_removed"]),
    "normal_expired_items": count_true(analysis_items["normal_expired"]),
    "still_seen_after_expected_items": count_true(analysis_items["still_seen_after_expected"]),

    "real_fast_close_strict_items": count_true(analysis_items["is_real_fast_close_strict"]),
    "real_fast_close_proxy_items": count_true(analysis_items["is_real_fast_close_proxy"]),
    "abandoned_disappointed_items": count_true(analysis_items["is_abandoned_disappointed"]),
    "lead_bait_ghosting_items": count_true(analysis_items["is_lead_bait_ghosting"]),
    "endgame_ghosting_items": count_true(analysis_items["is_endgame_ghosting"]),

    "agent_items": int((analysis_items["seller_type"] == "agent").sum()),
    "private_items": int((analysis_items["seller_type"] == "private").sum()),
}])

display_top(quick_summary, title="FINAL QUICK SUMMARY")
save_csv(quick_summary, "final_quick_summary.csv")

print("\nDONE.")
print("Output folder:")
print(CACHE_DIR)

print("\nMain files:")
print("1. item_behavior_features.csv")
print("2. early_removed_diagnosis.csv")
print("3. real_fast_close_candidates.csv")
print("4. abandoned_disappointed_candidates.csv")
print("5. lead_bait_ghosting_candidates.csv")
print("6. endgame_ghosting_candidates.csv")
print("7. item_behavior_bucket_summary.csv")
print("8. seller_type_behavior_summary.csv")
print("9. seller_type_overview.csv")
print("10. seller_reputation_min5.csv")
print("11. category_seller_behavior_summary.csv")
print("12. final_quick_summary.csv")
```

```python
import pandas as pd
import plotly.express as px

# 1. Đọc dữ liệu
try:
    df_items = pd.read_csv('/kaggle/input/datasets/kimkhanhnguyen214/item-behavior-feature/item_behavior_features.csv') 
except FileNotFoundError:
    df_items = pd.read_csv('item_behavior_features.csv')

# 2. Lọc tin có chat
chat_df = df_items[df_items["chat_users"].fillna(0) > 0].copy()

# 3. Phân loại "Độ sâu tư vấn" thành 3 nhóm
def categorize_turns(turns):
    if turns <= 1.5:
        return '1. Bơ khách / Trả lời cho có (<= 1.5 vòng)'
    elif turns <= 3.5:
        return '2. Tương tác cơ bản (1.5 - 3.5 vòng)'
    else:
        return '3. Tư vấn sâu có tâm (> 3.5 vòng)'

chat_df['turn_category'] = chat_df['avg_turns_per_chat_user'].apply(categorize_turns)

# 4. Tính toán tỷ trọng % cho Agent và Private
turn_summary = chat_df.groupby(['seller_type', 'turn_category']).size().reset_index(name='count')
turn_summary['percentage'] = turn_summary['count'] / turn_summary.groupby('seller_type')['count'].transform('sum')

# 5. Vẽ biểu đồ 100% Stacked Bar
fig1 = px.bar(
    turn_summary, 
    x="seller_type", 
    y="percentage", 
    color="turn_category",
    title="So sánh Tỷ lệ 'Bơ khách' vs 'Tư vấn sâu' (Agent vs Private)",
    color_discrete_map={
        '1. Bơ khách / Trả lời cho có (<= 1.5 vòng)': '#EF553B', # Đỏ cảnh báo
        '2. Tương tác cơ bản (1.5 - 3.5 vòng)': '#FECB52',       # Vàng
        '3. Tư vấn sâu có tâm (> 3.5 vòng)': '#00CC96'          # Xanh tích cực
    },
    text=turn_summary['percentage'].apply(lambda x: f'{x:.1%}'),
    labels={'seller_type': 'Loại người bán', 'percentage': 'Tỷ lệ %'}
)

fig1.update_layout(template="plotly_white", height=600, barmode='stack')
fig1.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Đảm bảo bạn đã chạy code trước đó để có dataframe `analysis_items`

def plot_early_chat_quality(df, day_window):
    """
    Hàm vẽ biểu đồ chất lượng trả lời tin nhắn trong N ngày đầu tiên.
    day_window: 3 hoặc 7
    """
    # Lấy đúng tên cột theo khung thời gian
    users_col = f"early{day_window}_chat_users"
    turns_col = f"early{day_window}_avg_turns_per_chat_user"
    
    # 1. Lọc những tin có phát sinh ít nhất 1 khách hỏi trong N ngày đầu
    df_filtered = df[df[users_col].fillna(0) > 0].copy()
    
    if len(df_filtered) == 0:
        print(f"Không có dữ liệu chat nào trong {day_window} ngày đầu.")
        return None

    # 2. Phân loại mức độ tương tác dựa trên số vòng lặp (turns) trung bình
    conditions = [
        df_filtered[turns_col] <= 1.5,
        (df_filtered[turns_col] > 1.5) & (df_filtered[turns_col] <= 3.5),
        df_filtered[turns_col] > 3.5
    ]
    
    # Đặt tên category có đánh số để Plotly tự động sort đúng thứ tự
    choices = [
        "1. Bơ khách / Trả lời cho có (<= 1.5 vòng)",
        "2. Tương tác cơ bản (1.5 - 3.5 vòng)",
        "3. Tư vấn sâu có tâm (> 3.5 vòng)"
    ]
    df_filtered['turn_category'] = np.select(conditions, choices, default="Unknown")

    # 3. Gom nhóm và tính Tỷ lệ % (Normalize về 100%)
    grouped = df_filtered.groupby(['seller_type', 'turn_category']).size().reset_index(name='count')
    
    # Tính tổng số tin đăng theo từng loại người bán (agent/private)
    total_by_seller = grouped.groupby('seller_type')['count'].transform('sum')
    
    # Tính tỷ lệ (dạng thập phân) và text để hiển thị (dạng % có 1 chữ số thập phân)
    grouped['Tỷ lệ'] = grouped['count'] / total_by_seller
    grouped['Tỷ lệ % hiển thị'] = (grouped['Tỷ lệ'] * 100).round(1).astype(str) + '%'

    # 4. Gán màu đồng bộ với hình mẫu
    color_map = {
        "1. Bơ khách / Trả lời cho có (<= 1.5 vòng)": "#EF553B", # Đỏ cam
        "2. Tương tác cơ bản (1.5 - 3.5 vòng)": "#F6C85F",        # Vàng
        "3. Tư vấn sâu có tâm (> 3.5 vòng)": "#00CC96"           # Xanh lá
    }

    # 5. Vẽ Plotly Stacked Bar
    fig = px.bar(
        grouped,
        x="seller_type",
        y="Tỷ lệ",
        color="turn_category",
        text="Tỷ lệ % hiển thị",
        color_discrete_map=color_map,
        category_orders={"turn_category": choices}, # Ép thứ tự hiển thị của Legend
        title=f"So sánh Tỷ lệ 'Bơ khách' vs 'Tư vấn sâu' (Agent vs Private) - <b>{day_window} Ngày Đầu</b>",
        labels={'seller_type': 'Loại người bán', 'turn_category': 'turn_category'}
    )

    # Làm đẹp layout
    fig.update_layout(
        barmode='stack',
        template='plotly_white',
        yaxis=dict(title='Tỷ lệ %', tickformat='.0%'),
        legend_title_text='turn_category'
    )
    
    # Chỉnh cho text hiển thị rõ ràng bên trong bar
    fig.update_traces(textposition='inside', insidetextanchor='middle')

    return fig

# ============================================================
# Gọi hàm để tạo và hiển thị 2 biểu đồ
# ============================================================

# Biểu đồ 3 ngày đầu
fig_3d = plot_early_chat_quality(analysis_items, 3)
if fig_3d:
    fig_3d.show()
    # fig_3d.write_html(os.path.join(INTERACTIVE_DIR, "chat_quality_3d_stacked.html"))

# Biểu đồ 7 ngày đầu
fig_7d = plot_early_chat_quality(analysis_items, 7)
if fig_7d:
    fig_7d.show()
    # fig_7d.write_html(os.path.join(INTERACTIVE_DIR, "chat_quality_7d_stacked.html"))
```
