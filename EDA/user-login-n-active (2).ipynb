```python
# ============================================================
# DATATHON 2026 — EDA TASKS
# 1. Listing spam / duplicate check in dim_listing
# 2. Login user active-day behavior
# 3. One-day login users: appear once then disappear
# 4. fact_user_events traffic/contact contribution by is_login + event_type
# ============================================================

import os
import sys
import subprocess
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# 0. CONFIG
# ============================================================

BASE_PATH_1 = "/kaggle/input/datasets/kimkhanhnguyen214/datathon2026-1"
BASE_PATH_2 = "/kaggle/input/datasets/kimkhanhnguyen214/datathon2026-2"


DIM_LISTING_PATH = os.path.join(BASE_PATH_1, "dim_listing")
EVENTS_PATH = os.path.join(BASE_PATH_2, "fact_user_events")

OUTPUT_DIR = "/kaggle/working/eda_marketplace_user_behavior"
os.makedirs(OUTPUT_DIR, exist_ok=True)

DUCKDB_MEMORY_LIMIT = "6GB"
DUCKDB_THREADS = 4

# Chỉnh ngưỡng tại đây
REGULAR_MAX_OFF_DAYS = 2      # off 1-2 ngày vẫn xem là active khá thường xuyên
LONG_OFF_DAYS = 7             # off >= 7 ngày xem là off dài
TOP_N = 30

# Dùng approx distinct cho task 4 để đỡ nặng.
# Row count vẫn là exact.
USE_APPROX_DISTINCT_FOR_BIG_EVENTS = True

POSITIVE_EVENTS = [
    "view_phone",
    "contact_chat",
    "other_interaction",
    "contact_zalo",
    "contact_sms",
]

POSITIVE_EVENTS_SQL = ", ".join([f"'{x}'" for x in POSITIVE_EVENTS])

print("DIM_LISTING_PATH exists:", os.path.exists(DIM_LISTING_PATH), DIM_LISTING_PATH)
print("EVENTS_PATH exists:", os.path.exists(EVENTS_PATH), EVENTS_PATH)
print("OUTPUT_DIR:", OUTPUT_DIR)

DIM_GLOB = os.path.join(DIM_LISTING_PATH, "*.parquet").replace("\\", "/")
EVENTS_GLOB = os.path.join(EVENTS_PATH, "*.parquet").replace("\\", "/")

# ============================================================
# 1. SETUP DUCKDB
# ============================================================

try:
    import duckdb
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "duckdb"])
    import duckdb

try:
    from IPython.display import display
except Exception:
    def display(x):
        print(x)

duckdb_path = os.path.join(OUTPUT_DIR, "eda_tasks.duckdb")

# Nếu muốn chạy lại sạch từ đầu thì uncomment dòng dưới
# if os.path.exists(duckdb_path):
#     os.remove(duckdb_path)

con = duckdb.connect(duckdb_path)
con.execute(f"PRAGMA threads={DUCKDB_THREADS}")
con.execute(f"PRAGMA memory_limit='{DUCKDB_MEMORY_LIMIT}'")

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

def category_name_sql(col="category"):
    return f"""
    CASE TRY_CAST({col} AS INTEGER)
        WHEN 1010 THEN '1010_room_rental'
        WHEN 1020 THEN '1020_apartment'
        WHEN 1030 THEN '1030_house'
        WHEN 1040 THEN '1040_land_commercial'
        WHEN 1050 THEN '1050_new_project'
        ELSE CONCAT('unknown_', COALESCE(CAST({col} AS VARCHAR), 'NULL'))
    END
    """

def save_df(df, filename, show_head=30):
    csv_path = os.path.join(OUTPUT_DIR, f"{filename}.csv")
    parquet_path = os.path.join(OUTPUT_DIR, f"{filename}.parquet")

    df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    try:
        df.to_parquet(parquet_path, index=False)
    except Exception as e:
        print("[WARN] Cannot save parquet:", filename, e)

    print(f"\n[SAVE] {csv_path}")
    display(df.head(show_head))
    return df

def run_query(filename, query, show_head=30):
    print("\n" + "=" * 90)
    print(filename)
    print("=" * 90)
    df = con.execute(query).df()
    return save_df(df, filename, show_head=show_head)

def plot_bar(df, x_col, y_col, title, filename, top_n=None, rotate=45):
    if df is None or len(df) == 0:
        print("[SKIP PLOT] empty df:", filename)
        return

    plot_df = df.copy()
    if top_n is not None:
        plot_df = plot_df.head(top_n)

    plt.figure(figsize=(12, 5))
    plt.bar(plot_df[x_col].astype(str), plot_df[y_col])
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.xticks(rotation=rotate, ha="right")
    plt.tight_layout()

    path = os.path.join(OUTPUT_DIR, filename)
    plt.savefig(path, dpi=160, bbox_inches="tight")
    plt.show()
    print("[SAVE PLOT]", path)

def plot_line(df, x_col, y_col, title, filename):
    if df is None or len(df) == 0:
        print("[SKIP PLOT] empty df:", filename)
        return

    plot_df = df.copy()
    plot_df[x_col] = pd.to_datetime(plot_df[x_col])

    plt.figure(figsize=(12, 5))
    plt.plot(plot_df[x_col], plot_df[y_col], marker="o")
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()

    path = os.path.join(OUTPUT_DIR, filename)
    plt.savefig(path, dpi=160, bbox_inches="tight")
    plt.show()
    print("[SAVE PLOT]", path)

# ============================================================
# 2. BUILD CLEAN dim_listing TABLE
# ============================================================

print("\nBuilding dim_listing_clean...")

con.execute(f"""
CREATE OR REPLACE TABLE dim_listing_clean AS
WITH raw AS (
    SELECT
        CAST(item_id AS VARCHAR) AS item_id,
        CAST(seller_id AS VARCHAR) AS seller_id,

        TRY_CAST(category AS INTEGER) AS category,
        {category_name_sql("category")} AS category_name,

        LOWER(TRIM(COALESCE(CAST(seller_type AS VARCHAR), 'unknown'))) AS seller_type_clean,
        LOWER(TRIM(COALESCE(CAST(ad_type AS VARCHAR), 'unknown'))) AS ad_type_clean,
        LOWER(TRIM(COALESCE(CAST(ad_status AS VARCHAR), 'unknown'))) AS ad_status_clean,

        COALESCE(CAST(title AS VARCHAR), '') AS title_raw,
        REGEXP_REPLACE(
            LOWER(TRIM(COALESCE(CAST(title AS VARCHAR), ''))),
            '\\s+',
            ' ',
            'g'
        ) AS title_norm,

        LOWER(TRIM(COALESCE(CAST(city_name AS VARCHAR), 'unknown'))) AS city_clean,
        LOWER(TRIM(COALESCE(CAST(district_name AS VARCHAR), 'unknown'))) AS district_clean,
        LOWER(TRIM(COALESCE(CAST(ward_name AS VARCHAR), 'unknown'))) AS ward_clean,

        LOWER(TRIM(COALESCE(CAST(price_bucket AS VARCHAR), 'unknown'))) AS price_bucket_clean,
        LOWER(TRIM(COALESCE(CAST(direction AS VARCHAR), 'unknown'))) AS direction_clean,
        LOWER(TRIM(COALESCE(CAST(legal_status AS VARCHAR), 'unknown'))) AS legal_status_clean,
        LOWER(TRIM(COALESCE(CAST(house_type AS VARCHAR), 'unknown'))) AS house_type_clean,
        LOWER(TRIM(COALESCE(CAST(furnishing AS VARCHAR), 'unknown'))) AS furnishing_clean,

        CAST(project_id AS VARCHAR) AS project_id,

        ROUND(TRY_CAST(area_sqm AS DOUBLE), 1) AS area_sqm_round,
        TRY_CAST(bedrooms AS DOUBLE) AS bedrooms_num,
        TRY_CAST(bathrooms AS DOUBLE) AS bathrooms_num,
        TRY_CAST(floors AS DOUBLE) AS floors_num,
        ROUND(TRY_CAST(width_m AS DOUBLE), 1) AS width_m_round,
        TRY_CAST(images_count AS DOUBLE) AS images_count_num,

        TRY_CAST(posted_date AS DATE) AS posted_date,
        TRY_CAST(expected_expired_date AS DATE) AS expected_expired_date

    FROM read_parquet('{DIM_GLOB}')
),

ranked AS (
    SELECT
        *,
        ROW_NUMBER() OVER (
            PARTITION BY item_id
            ORDER BY posted_date ASC NULLS LAST, expected_expired_date ASC NULLS LAST
        ) AS rn
    FROM raw
    WHERE item_id IS NOT NULL
)

SELECT
    * EXCLUDE (rn),

    -- Duplicate fingerprint 1:
    -- same seller + same content/location/price/basic specs
    MD5(CONCAT_WS(
        '||',
        COALESCE(seller_id, 'NULL'),
        COALESCE(CAST(category AS VARCHAR), 'NULL'),
        COALESCE(title_norm, 'NULL'),
        COALESCE(ad_type_clean, 'NULL'),
        COALESCE(city_clean, 'NULL'),
        COALESCE(district_clean, 'NULL'),
        COALESCE(ward_clean, 'NULL'),
        COALESCE(price_bucket_clean, 'NULL'),
        COALESCE(CAST(area_sqm_round AS VARCHAR), 'NULL'),
        COALESCE(CAST(bedrooms_num AS VARCHAR), 'NULL'),
        COALESCE(CAST(bathrooms_num AS VARCHAR), 'NULL'),
        COALESCE(CAST(floors_num AS VARCHAR), 'NULL'),
        COALESCE(CAST(width_m_round AS VARCHAR), 'NULL'),
        COALESCE(house_type_clean, 'NULL'),
        COALESCE(furnishing_clean, 'NULL')
    )) AS fp_seller_strict,

    -- Duplicate fingerprint 2:
    -- same content/location/price/basic specs, ignore seller
    MD5(CONCAT_WS(
        '||',
        COALESCE(CAST(category AS VARCHAR), 'NULL'),
        COALESCE(title_norm, 'NULL'),
        COALESCE(ad_type_clean, 'NULL'),
        COALESCE(city_clean, 'NULL'),
        COALESCE(district_clean, 'NULL'),
        COALESCE(ward_clean, 'NULL'),
        COALESCE(price_bucket_clean, 'NULL'),
        COALESCE(CAST(area_sqm_round AS VARCHAR), 'NULL'),
        COALESCE(CAST(bedrooms_num AS VARCHAR), 'NULL'),
        COALESCE(CAST(bathrooms_num AS VARCHAR), 'NULL'),
        COALESCE(CAST(floors_num AS VARCHAR), 'NULL'),
        COALESCE(CAST(width_m_round AS VARCHAR), 'NULL'),
        COALESCE(house_type_clean, 'NULL'),
        COALESCE(furnishing_clean, 'NULL')
    )) AS fp_cross_seller,

    -- Duplicate fingerprint 3:
    -- same seller + same title + same category/location only
    -- dùng để bắt spam nhẹ hơn, vì đôi khi specs hơi khác nhau
    MD5(CONCAT_WS(
        '||',
        COALESCE(seller_id, 'NULL'),
        COALESCE(CAST(category AS VARCHAR), 'NULL'),
        COALESCE(title_norm, 'NULL'),
        COALESCE(ad_type_clean, 'NULL'),
        COALESCE(city_clean, 'NULL'),
        COALESCE(district_clean, 'NULL'),
        COALESCE(ward_clean, 'NULL')
    )) AS fp_seller_title_location

FROM ranked
WHERE rn = 1
""")

# ============================================================
# TASK 1 — SPAM / DUPLICATE LISTING CHECK
# ============================================================

# 1A. Primary key quality
dim_pk_summary = run_query(
    "01_dim_listing_primary_key_quality",
    f"""
    WITH raw AS (
        SELECT CAST(item_id AS VARCHAR) AS item_id
        FROM read_parquet('{DIM_GLOB}')
    ),
    item_counts AS (
        SELECT item_id, COUNT(*) AS n_rows
        FROM raw
        GROUP BY item_id
    )

    SELECT
        (SELECT COUNT(*) FROM raw) AS total_rows,
        (SELECT COUNT(DISTINCT item_id) FROM raw) AS distinct_item_id,
        (SELECT COUNT(*) FROM raw) - (SELECT COUNT(DISTINCT item_id) FROM raw) AS duplicated_item_id_rows,
        SUM(CASE WHEN n_rows > 1 THEN 1 ELSE 0 END) AS item_id_with_multiple_rows,
        MAX(n_rows) AS max_rows_per_item_id
    FROM item_counts
    """
)

# 1B. Duplicate summary by category + seller_type
listing_dup_summary = run_query(
    "02_listing_duplicate_by_category_seller_type",
    f"""
    WITH seller_strict AS (
        SELECT
            'same_seller_strict_content' AS duplicate_type,
            seller_type_clean,
            category,
            category_name,
            fp_seller_strict AS fp,
            COUNT(DISTINCT item_id) AS n_items,
            COUNT(DISTINCT seller_id) AS n_sellers
        FROM dim_listing_clean
        GROUP BY seller_type_clean, category, category_name, fp_seller_strict
        HAVING COUNT(DISTINCT item_id) > 1
    ),

    seller_title_location AS (
        SELECT
            'same_seller_title_location' AS duplicate_type,
            seller_type_clean,
            category,
            category_name,
            fp_seller_title_location AS fp,
            COUNT(DISTINCT item_id) AS n_items,
            COUNT(DISTINCT seller_id) AS n_sellers
        FROM dim_listing_clean
        GROUP BY seller_type_clean, category, category_name, fp_seller_title_location
        HAVING COUNT(DISTINCT item_id) > 1
    ),

    cross_seller AS (
        SELECT
            'cross_seller_same_content' AS duplicate_type,
            CASE
                WHEN COUNT(DISTINCT seller_type_clean) = 1 THEN MIN(seller_type_clean)
                ELSE 'mixed_seller_type'
            END AS seller_type_clean,
            category,
            category_name,
            fp_cross_seller AS fp,
            COUNT(DISTINCT item_id) AS n_items,
            COUNT(DISTINCT seller_id) AS n_sellers
        FROM dim_listing_clean
        GROUP BY category, category_name, fp_cross_seller
        HAVING COUNT(DISTINCT item_id) > 1
           AND COUNT(DISTINCT seller_id) > 1
    ),

    all_dup AS (
        SELECT * FROM seller_strict
        UNION ALL
        SELECT * FROM seller_title_location
        UNION ALL
        SELECT * FROM cross_seller
    )

    SELECT
        duplicate_type,
        seller_type_clean,
        category,
        category_name,
        COUNT(*) AS duplicate_groups,
        SUM(n_items) AS listings_inside_duplicate_groups,
        SUM(n_items - 1) AS extra_duplicate_listings,
        AVG(n_items) AS avg_items_per_group,
        MAX(n_items) AS max_items_per_group,
        AVG(n_sellers) AS avg_sellers_per_group,
        MAX(n_sellers) AS max_sellers_per_group
    FROM all_dup
    GROUP BY duplicate_type, seller_type_clean, category, category_name
    ORDER BY extra_duplicate_listings DESC
    """
)

# 1C. Top duplicate groups to inspect manually
top_dup_groups = run_query(
    "03_top_duplicate_listing_groups_to_inspect",
    f"""
    WITH seller_strict AS (
        SELECT
            'same_seller_strict_content' AS duplicate_type,
            seller_type_clean,
            category,
            category_name,
            COUNT(DISTINCT item_id) AS n_items,
            COUNT(DISTINCT seller_id) AS n_sellers,
            ANY_VALUE(seller_id) AS sample_seller_id,
            ANY_VALUE(item_id) AS sample_item_id,
            ANY_VALUE(title_raw) AS sample_title,
            ANY_VALUE(city_clean) AS sample_city,
            ANY_VALUE(district_clean) AS sample_district,
            ANY_VALUE(ward_clean) AS sample_ward,
            ANY_VALUE(price_bucket_clean) AS sample_price_bucket,
            ANY_VALUE(area_sqm_round) AS sample_area_sqm,
            ANY_VALUE(bedrooms_num) AS sample_bedrooms,
            ANY_VALUE(bathrooms_num) AS sample_bathrooms
        FROM dim_listing_clean
        GROUP BY seller_type_clean, category, category_name, fp_seller_strict
        HAVING COUNT(DISTINCT item_id) > 1
    ),

    seller_title_location AS (
        SELECT
            'same_seller_title_location' AS duplicate_type,
            seller_type_clean,
            category,
            category_name,
            COUNT(DISTINCT item_id) AS n_items,
            COUNT(DISTINCT seller_id) AS n_sellers,
            ANY_VALUE(seller_id) AS sample_seller_id,
            ANY_VALUE(item_id) AS sample_item_id,
            ANY_VALUE(title_raw) AS sample_title,
            ANY_VALUE(city_clean) AS sample_city,
            ANY_VALUE(district_clean) AS sample_district,
            ANY_VALUE(ward_clean) AS sample_ward,
            ANY_VALUE(price_bucket_clean) AS sample_price_bucket,
            ANY_VALUE(area_sqm_round) AS sample_area_sqm,
            ANY_VALUE(bedrooms_num) AS sample_bedrooms,
            ANY_VALUE(bathrooms_num) AS sample_bathrooms
        FROM dim_listing_clean
        GROUP BY seller_type_clean, category, category_name, fp_seller_title_location
        HAVING COUNT(DISTINCT item_id) > 1
    ),

    cross_seller AS (
        SELECT
            'cross_seller_same_content' AS duplicate_type,
            CASE
                WHEN COUNT(DISTINCT seller_type_clean) = 1 THEN MIN(seller_type_clean)
                ELSE 'mixed_seller_type'
            END AS seller_type_clean,
            category,
            category_name,
            COUNT(DISTINCT item_id) AS n_items,
            COUNT(DISTINCT seller_id) AS n_sellers,
            ANY_VALUE(seller_id) AS sample_seller_id,
            ANY_VALUE(item_id) AS sample_item_id,
            ANY_VALUE(title_raw) AS sample_title,
            ANY_VALUE(city_clean) AS sample_city,
            ANY_VALUE(district_clean) AS sample_district,
            ANY_VALUE(ward_clean) AS sample_ward,
            ANY_VALUE(price_bucket_clean) AS sample_price_bucket,
            ANY_VALUE(area_sqm_round) AS sample_area_sqm,
            ANY_VALUE(bedrooms_num) AS sample_bedrooms,
            ANY_VALUE(bathrooms_num) AS sample_bathrooms
        FROM dim_listing_clean
        GROUP BY category, category_name, fp_cross_seller
        HAVING COUNT(DISTINCT item_id) > 1
           AND COUNT(DISTINCT seller_id) > 1
    ),

    all_dup AS (
        SELECT * FROM seller_strict
        UNION ALL
        SELECT * FROM seller_title_location
        UNION ALL
        SELECT * FROM cross_seller
    )

    SELECT *
    FROM all_dup
    ORDER BY n_items DESC, n_sellers DESC
    LIMIT 200
    """,
    show_head=50
)

# 1D. Seller volume concentration
con.execute("""
CREATE OR REPLACE TABLE seller_listing_counts AS
SELECT
    seller_id,
    seller_type_clean,
    category,
    category_name,
    COUNT(*) AS listings
FROM dim_listing_clean
WHERE seller_id IS NOT NULL
GROUP BY seller_id, seller_type_clean, category, category_name
""")

seller_volume_summary = run_query(
    "04_seller_listing_volume_distribution",
    """
    SELECT
        seller_type_clean,
        category,
        category_name,
        COUNT(*) AS sellers,
        SUM(listings) AS total_listings,
        AVG(listings) AS avg_listings_per_seller,
        QUANTILE_CONT(listings, 0.50) AS median_listings_per_seller,
        QUANTILE_CONT(listings, 0.90) AS p90_listings_per_seller,
        QUANTILE_CONT(listings, 0.95) AS p95_listings_per_seller,
        QUANTILE_CONT(listings, 0.99) AS p99_listings_per_seller,
        MAX(listings) AS max_listings_per_seller,
        SUM(CASE WHEN listings >= 5 THEN 1 ELSE 0 END) AS sellers_5plus_listings,
        SUM(CASE WHEN listings >= 10 THEN 1 ELSE 0 END) AS sellers_10plus_listings,
        SUM(CASE WHEN listings >= 20 THEN 1 ELSE 0 END) AS sellers_20plus_listings
    FROM seller_listing_counts
    GROUP BY seller_type_clean, category, category_name
    ORDER BY p99_listings_per_seller DESC, total_listings DESC
    """
)

top_sellers = run_query(
    "05_top_sellers_by_listing_count",
    """
    SELECT
        seller_id,
        seller_type_clean,
        category,
        category_name,
        listings
    FROM seller_listing_counts
    ORDER BY listings DESC
    LIMIT 200
    """,
    show_head=50
)

# Chart task 1
tmp = listing_dup_summary.copy()
tmp["label"] = tmp["duplicate_type"] + " | " + tmp["seller_type_clean"] + " | " + tmp["category_name"]
plot_bar(
    tmp.sort_values("extra_duplicate_listings", ascending=False),
    "label",
    "extra_duplicate_listings",
    "Task 1 — Extra duplicate listings by duplicate type / seller type / category",
    "task1_duplicate_listings.png",
    top_n=TOP_N
)

# ============================================================
# TASK 2 — LOGIN USER ACTIVE DAYS / OFF GAP
# ============================================================

LOGIN_DAILY_CACHE = os.path.join(OUTPUT_DIR, "cache_login_user_daily.parquet").replace("\\", "/")
LOGIN_LIFECYCLE_CACHE = os.path.join(OUTPUT_DIR, "cache_login_user_lifecycle.parquet").replace("\\", "/")

print("\nBuilding login_user_daily... This reads fact_user_events once.")

if os.path.exists(LOGIN_DAILY_CACHE):
    print("[CACHE] Loading login_user_daily from:", LOGIN_DAILY_CACHE)
    con.execute(f"""
    CREATE OR REPLACE TABLE login_user_daily AS
    SELECT * FROM read_parquet('{LOGIN_DAILY_CACHE}')
    """)
else:
    con.execute(f"""
    CREATE OR REPLACE TABLE login_user_daily AS
    WITH base AS (
        SELECT
            CAST(user_id AS VARCHAR) AS user_id,
            TRY_CAST(date AS DATE) AS active_date,
            CAST(session_id AS VARCHAR) AS session_id,
            CAST(item_id AS VARCHAR) AS item_id,

            LOWER(TRIM(COALESCE(CAST(event_type AS VARCHAR), 'unknown'))) AS event_type_clean,
            COALESCE(TRY_CAST(is_contact AS INTEGER), 0) AS is_contact_int,
            TRY_CAST(event_ts AS TIMESTAMP) AS event_ts

        FROM read_parquet('{EVENTS_GLOB}')
        WHERE LOWER(TRIM(COALESCE(CAST(is_login AS VARCHAR), 'unknown'))) = 'login'
          AND user_id IS NOT NULL
    )

    SELECT
        user_id,
        active_date,
        COUNT(*) AS event_rows,
        COUNT(DISTINCT session_id) AS sessions,
        COUNT(DISTINCT item_id) AS items,
        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS contact_flag_rows,
        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS positive_event_rows,
        MIN(event_ts) AS first_event_ts,
        MAX(event_ts) AS last_event_ts
    FROM base
    WHERE active_date IS NOT NULL
    GROUP BY user_id, active_date
    """)

    con.execute(f"""
    COPY login_user_daily TO '{LOGIN_DAILY_CACHE}' (FORMAT PARQUET)
    """)
    print("[SAVE CACHE]", LOGIN_DAILY_CACHE)

print("Building login_user_lifecycle...")

con.execute(f"""
CREATE OR REPLACE TABLE login_user_lifecycle AS
WITH ordered AS (
    SELECT
        user_id,
        active_date,
        event_rows,
        sessions,
        items,
        contact_flag_rows,
        positive_event_rows,

        LAG(active_date) OVER (
            PARTITION BY user_id
            ORDER BY active_date
        ) AS prev_active_date
    FROM login_user_daily
),

with_gap AS (
    SELECT
        *,
        CASE
            WHEN prev_active_date IS NULL THEN NULL
            ELSE DATE_DIFF('day', prev_active_date, active_date) - 1
        END AS inactive_gap_days
    FROM ordered
),

agg AS (
    SELECT
        user_id,
        COUNT(*) AS active_days,
        MIN(active_date) AS first_active_date,
        MAX(active_date) AS last_active_date,
        DATE_DIFF('day', MIN(active_date), MAX(active_date)) + 1 AS observed_span_days,

        COALESCE(MAX(inactive_gap_days), 0) AS max_inactive_gap_days,
        COALESCE(AVG(inactive_gap_days), 0) AS avg_inactive_gap_days,
        SUM(CASE WHEN inactive_gap_days >= {LONG_OFF_DAYS} THEN 1 ELSE 0 END) AS long_off_gap_count,

        SUM(event_rows) AS total_event_rows,
        SUM(sessions) AS total_sessions_daily_sum,
        SUM(items) AS total_items_daily_sum,
        SUM(contact_flag_rows) AS total_contact_flag_rows,
        SUM(positive_event_rows) AS total_positive_event_rows
    FROM with_gap
    GROUP BY user_id
)

SELECT
    *,
    active_days * 1.0 / NULLIF(observed_span_days, 0) AS active_ratio_inside_observed_span,

    CASE
        WHEN active_days = 1 THEN 'one_day_only'
        WHEN max_inactive_gap_days <= {REGULAR_MAX_OFF_DAYS} THEN 'regular_off_0_to_{REGULAR_MAX_OFF_DAYS}_days'
        WHEN max_inactive_gap_days >= {LONG_OFF_DAYS} THEN 'long_off_{LONG_OFF_DAYS}_plus_days'
        ELSE 'medium_off_between_{REGULAR_MAX_OFF_DAYS + 1}_and_{LONG_OFF_DAYS - 1}_days'
    END AS active_pattern

FROM agg
""")

con.execute(f"""
COPY login_user_lifecycle TO '{LOGIN_LIFECYCLE_CACHE}' (FORMAT PARQUET)
""")
print("[SAVE CACHE]", LOGIN_LIFECYCLE_CACHE)

login_overall_summary = run_query(
    "06_login_user_active_overall_summary",
    f"""
    SELECT
        COUNT(*) AS total_login_users,

        AVG(active_days) AS avg_active_days,
        QUANTILE_CONT(active_days, 0.50) AS median_active_days,
        QUANTILE_CONT(active_days, 0.75) AS p75_active_days,
        QUANTILE_CONT(active_days, 0.90) AS p90_active_days,
        QUANTILE_CONT(active_days, 0.95) AS p95_active_days,
        QUANTILE_CONT(active_days, 0.99) AS p99_active_days,
        MAX(active_days) AS max_active_days,

        AVG(max_inactive_gap_days) AS avg_max_inactive_gap_days,
        QUANTILE_CONT(max_inactive_gap_days, 0.50) AS median_max_inactive_gap_days,
        QUANTILE_CONT(max_inactive_gap_days, 0.90) AS p90_max_inactive_gap_days,
        QUANTILE_CONT(max_inactive_gap_days, 0.95) AS p95_max_inactive_gap_days,
        QUANTILE_CONT(max_inactive_gap_days, 0.99) AS p99_max_inactive_gap_days,
        MAX(max_inactive_gap_days) AS max_inactive_gap_days,

        SUM(CASE WHEN active_days = 1 THEN 1 ELSE 0 END) AS one_day_users,
        SUM(CASE WHEN active_days = 1 THEN 1 ELSE 0 END) * 1.0 / NULLIF(COUNT(*), 0) AS one_day_user_rate,

        SUM(CASE WHEN active_pattern = 'regular_off_0_to_{REGULAR_MAX_OFF_DAYS}_days' THEN 1 ELSE 0 END) AS regular_users,
        SUM(CASE WHEN active_pattern = 'long_off_{LONG_OFF_DAYS}_plus_days' THEN 1 ELSE 0 END) AS long_off_users
    FROM login_user_lifecycle
    """
)

active_pattern_dist = run_query(
    "07_login_user_active_pattern_distribution",
    """
    SELECT
        active_pattern,
        COUNT(*) AS users,
        COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share,
        AVG(active_days) AS avg_active_days,
        AVG(total_event_rows) AS avg_event_rows,
        AVG(total_contact_flag_rows) AS avg_contact_flag_rows,
        AVG(total_positive_event_rows) AS avg_positive_event_rows
    FROM login_user_lifecycle
    GROUP BY active_pattern
    ORDER BY users DESC
    """
)

active_days_dist = run_query(
    "08_login_user_active_days_distribution",
    """
    WITH b AS (
        SELECT
            CASE
                WHEN active_days = 1 THEN '01_1_day'
                WHEN active_days BETWEEN 2 AND 3 THEN '02_2_3_days'
                WHEN active_days BETWEEN 4 AND 7 THEN '03_4_7_days'
                WHEN active_days BETWEEN 8 AND 14 THEN '04_8_14_days'
                WHEN active_days BETWEEN 15 AND 30 THEN '05_15_30_days'
                WHEN active_days BETWEEN 31 AND 60 THEN '06_31_60_days'
                WHEN active_days BETWEEN 61 AND 90 THEN '07_61_90_days'
                ELSE '08_91plus_days'
            END AS active_days_bucket,

            CASE
                WHEN active_days = 1 THEN 1
                WHEN active_days BETWEEN 2 AND 3 THEN 2
                WHEN active_days BETWEEN 4 AND 7 THEN 3
                WHEN active_days BETWEEN 8 AND 14 THEN 4
                WHEN active_days BETWEEN 15 AND 30 THEN 5
                WHEN active_days BETWEEN 31 AND 60 THEN 6
                WHEN active_days BETWEEN 61 AND 90 THEN 7
                ELSE 8
            END AS sort_key,

            active_pattern,
            active_days
        FROM login_user_lifecycle
    )

    SELECT
        sort_key,
        active_days_bucket,
        active_pattern,
        COUNT(*) AS users,
        COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share
    FROM b
    GROUP BY sort_key, active_days_bucket, active_pattern
    ORDER BY sort_key, active_pattern
    """,
    show_head=80
)

off_gap_dist = run_query(
    "09_login_user_off_gap_distribution",
    f"""
    WITH b AS (
        SELECT
            CASE
                WHEN max_inactive_gap_days = 0 THEN '00_no_gap_or_one_day'
                WHEN max_inactive_gap_days BETWEEN 1 AND {REGULAR_MAX_OFF_DAYS} THEN '01_off_1_to_{REGULAR_MAX_OFF_DAYS}_days'
                WHEN max_inactive_gap_days BETWEEN {REGULAR_MAX_OFF_DAYS + 1} AND {LONG_OFF_DAYS - 1} THEN '02_off_{REGULAR_MAX_OFF_DAYS + 1}_to_{LONG_OFF_DAYS - 1}_days'
                WHEN max_inactive_gap_days BETWEEN {LONG_OFF_DAYS} AND 14 THEN '03_off_{LONG_OFF_DAYS}_to_14_days'
                WHEN max_inactive_gap_days BETWEEN 15 AND 30 THEN '04_off_15_to_30_days'
                ELSE '05_off_31plus_days'
            END AS off_gap_bucket,

            CASE
                WHEN max_inactive_gap_days = 0 THEN 0
                WHEN max_inactive_gap_days BETWEEN 1 AND {REGULAR_MAX_OFF_DAYS} THEN 1
                WHEN max_inactive_gap_days BETWEEN {REGULAR_MAX_OFF_DAYS + 1} AND {LONG_OFF_DAYS - 1} THEN 2
                WHEN max_inactive_gap_days BETWEEN {LONG_OFF_DAYS} AND 14 THEN 3
                WHEN max_inactive_gap_days BETWEEN 15 AND 30 THEN 4
                ELSE 5
            END AS sort_key,

            active_pattern
        FROM login_user_lifecycle
    )

    SELECT
        sort_key,
        off_gap_bucket,
        active_pattern,
        COUNT(*) AS users,
        COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share
    FROM b
    GROUP BY sort_key, off_gap_bucket, active_pattern
    ORDER BY sort_key, active_pattern
    """,
    show_head=80
)

daily_login_users = run_query(
    "10_daily_login_active_users",
    """
    SELECT
        active_date,
        COUNT(*) AS login_active_users,
        SUM(event_rows) AS event_rows,
        SUM(contact_flag_rows) AS contact_flag_rows,
        SUM(positive_event_rows) AS positive_event_rows,
        SUM(contact_flag_rows) * 1.0 / NULLIF(SUM(event_rows), 0) AS contact_flag_rate,
        SUM(positive_event_rows) * 1.0 / NULLIF(SUM(event_rows), 0) AS positive_event_rate
    FROM login_user_daily
    GROUP BY active_date
    ORDER BY active_date
    """,
    show_head=20
)

# Chart task 2
plot_bar(
    active_pattern_dist,
    "active_pattern",
    "users",
    "Task 2 — Login users by active pattern",
    "task2_active_pattern_users.png",
    rotate=25
)

tmp = active_days_dist.groupby(["sort_key", "active_days_bucket"], as_index=False)["users"].sum()
tmp = tmp.sort_values("sort_key")
plot_bar(
    tmp,
    "active_days_bucket",
    "users",
    "Task 2 — Login users by number of active days",
    "task2_active_days_distribution.png",
    rotate=30
)

tmp = off_gap_dist.groupby(["sort_key", "off_gap_bucket"], as_index=False)["users"].sum()
tmp = tmp.sort_values("sort_key")
plot_bar(
    tmp,
    "off_gap_bucket",
    "users",
    "Task 2 — Login users by max inactive gap",
    "task2_off_gap_distribution.png",
    rotate=30
)

plot_line(
    daily_login_users,
    "active_date",
    "login_active_users",
    "Task 2 — Daily login active users",
    "task2_daily_login_active_users.png"
)

# ============================================================
# TASK 3 — ONE-DAY LOGIN USERS
# Users who appear only 1 active day then disappear in train
# ============================================================

print("\nBuilding one_day_login_users...")

con.execute("""
CREATE OR REPLACE TABLE one_day_login_users AS
SELECT
    user_id,
    first_active_date AS only_active_date,
    total_event_rows,
    total_contact_flag_rows,
    total_positive_event_rows
FROM login_user_lifecycle
WHERE active_days = 1
""")

one_day_summary = run_query(
    "11_one_day_login_users_summary",
    """
    SELECT
        (SELECT COUNT(*) FROM one_day_login_users) AS one_day_login_users,
        (SELECT COUNT(*) FROM login_user_lifecycle) AS total_login_users,
        (SELECT COUNT(*) FROM one_day_login_users) * 1.0
            / NULLIF((SELECT COUNT(*) FROM login_user_lifecycle), 0) AS one_day_login_user_rate,

        AVG(total_event_rows) AS avg_event_rows_in_that_day,
        QUANTILE_CONT(total_event_rows, 0.50) AS median_event_rows_in_that_day,
        QUANTILE_CONT(total_event_rows, 0.90) AS p90_event_rows_in_that_day,
        MAX(total_event_rows) AS max_event_rows_in_that_day,

        SUM(total_contact_flag_rows) AS total_contact_flag_rows_by_one_day_users,
        SUM(total_positive_event_rows) AS total_positive_event_rows_by_one_day_users,
        SUM(CASE WHEN total_contact_flag_rows > 0 THEN 1 ELSE 0 END) AS one_day_users_with_contact_flag,
        SUM(CASE WHEN total_positive_event_rows > 0 THEN 1 ELSE 0 END) AS one_day_users_with_positive_event
    FROM one_day_login_users
    """
)

one_day_event_type_category = run_query(
    "12_one_day_login_users_event_type_category",
    f"""
    WITH base AS (
        SELECT
            CAST(e.user_id AS VARCHAR) AS user_id,
            LOWER(TRIM(COALESCE(CAST(e.event_type AS VARCHAR), 'unknown'))) AS event_type_clean,
            TRY_CAST(e.category AS INTEGER) AS category,
            {category_name_sql("e.category")} AS category_name,
            COALESCE(TRY_CAST(e.is_contact AS INTEGER), 0) AS is_contact_int,
            CAST(e.item_id AS VARCHAR) AS item_id,
            CAST(e.session_id AS VARCHAR) AS session_id
        FROM read_parquet('{EVENTS_GLOB}') e
        INNER JOIN one_day_login_users u
            ON CAST(e.user_id AS VARCHAR) = u.user_id
        WHERE LOWER(TRIM(COALESCE(CAST(e.is_login AS VARCHAR), 'unknown'))) = 'login'
    )

    SELECT
        event_type_clean,
        category,
        category_name,
        COUNT(*) AS rows,
        COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS row_share,

        COUNT(DISTINCT user_id) AS users,
        COUNT(DISTINCT user_id) * 1.0
            / NULLIF((SELECT COUNT(*) FROM one_day_login_users), 0) AS user_share_among_one_day_users,

        COUNT(DISTINCT session_id) AS sessions,
        COUNT(DISTINCT item_id) AS items,

        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS contact_flag_rows,
        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS positive_event_rows
    FROM base
    GROUP BY event_type_clean, category, category_name
    ORDER BY rows DESC
    """,
    show_head=80
)

one_day_first_action = run_query(
    "13_one_day_login_users_first_action_event_type_category",
    f"""
    WITH base AS (
        SELECT
            CAST(e.user_id AS VARCHAR) AS user_id,
            LOWER(TRIM(COALESCE(CAST(e.event_type AS VARCHAR), 'unknown'))) AS event_type_clean,
            TRY_CAST(e.category AS INTEGER) AS category,
            {category_name_sql("e.category")} AS category_name,
            TRY_CAST(e.event_ts AS TIMESTAMP) AS event_ts,
            CAST(e.event_id AS VARCHAR) AS event_id
        FROM read_parquet('{EVENTS_GLOB}') e
        INNER JOIN one_day_login_users u
            ON CAST(e.user_id AS VARCHAR) = u.user_id
        WHERE LOWER(TRIM(COALESCE(CAST(e.is_login AS VARCHAR), 'unknown'))) = 'login'
    ),

    ranked AS (
        SELECT
            *,
            ROW_NUMBER() OVER (
                PARTITION BY user_id
                ORDER BY event_ts ASC NULLS LAST, event_id ASC NULLS LAST
            ) AS rn
        FROM base
    )

    SELECT
        event_type_clean AS first_event_type,
        category,
        category_name,
        COUNT(*) AS users,
        COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share
    FROM ranked
    WHERE rn = 1
    GROUP BY event_type_clean, category, category_name
    ORDER BY users DESC
    """,
    show_head=80
)

one_day_active_date_dist = run_query(
    "14_one_day_login_users_by_only_active_date",
    """
    SELECT
        only_active_date,
        COUNT(*) AS one_day_users,
        SUM(total_event_rows) AS event_rows,
        SUM(total_contact_flag_rows) AS contact_flag_rows,
        SUM(total_positive_event_rows) AS positive_event_rows
    FROM one_day_login_users
    GROUP BY only_active_date
    ORDER BY only_active_date
    """,
    show_head=30
)

# Chart task 3
tmp = one_day_event_type_category.copy()
tmp["label"] = tmp["event_type_clean"] + " | " + tmp["category_name"]
plot_bar(
    tmp.sort_values("rows", ascending=False),
    "label",
    "rows",
    "Task 3 — What one-day login users did: event_type x category",
    "task3_one_day_event_type_category_rows.png",
    top_n=TOP_N
)

tmp = one_day_first_action.copy()
tmp["label"] = tmp["first_event_type"] + " | " + tmp["category_name"]
plot_bar(
    tmp.sort_values("users", ascending=False),
    "label",
    "users",
    "Task 3 — First action of one-day login users",
    "task3_one_day_first_action.png",
    top_n=TOP_N
)

plot_line(
    one_day_active_date_dist,
    "only_active_date",
    "one_day_users",
    "Task 3 — One-day login users by active date",
    "task3_one_day_users_by_date.png"
)

# ============================================================
# TASK 4 — fact_user_events traffic/contact contribution
# Split by is_login + event_type + category
# ============================================================

print("\nTask 4 reads fact_user_events for traffic/contact contribution.")

if USE_APPROX_DISTINCT_FOR_BIG_EVENTS:
    USER_COUNT_EXPR = "APPROX_COUNT_DISTINCT(user_id)"
    SESSION_COUNT_EXPR = "APPROX_COUNT_DISTINCT(session_id)"
    ITEM_COUNT_EXPR = "APPROX_COUNT_DISTINCT(item_id)"
    COUNT_NOTE = "approx"
else:
    USER_COUNT_EXPR = "COUNT(DISTINCT user_id)"
    SESSION_COUNT_EXPR = "COUNT(DISTINCT session_id)"
    ITEM_COUNT_EXPR = "COUNT(DISTINCT item_id)"
    COUNT_NOTE = "exact"

events_is_login_summary = run_query(
    "15_fact_user_events_is_login_summary",
    f"""
    WITH base AS (
        SELECT
            LOWER(TRIM(COALESCE(CAST(is_login AS VARCHAR), 'unknown'))) AS is_login_clean,
            CAST(user_id AS VARCHAR) AS user_id,
            CAST(session_id AS VARCHAR) AS session_id,
            CAST(item_id AS VARCHAR) AS item_id,
            LOWER(TRIM(COALESCE(CAST(event_type AS VARCHAR), 'unknown'))) AS event_type_clean,
            COALESCE(TRY_CAST(is_contact AS INTEGER), 0) AS is_contact_int
        FROM read_parquet('{EVENTS_GLOB}')
    ),

    total AS (
        SELECT
            COUNT(*) AS total_rows,
            SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS total_contact_flag_rows,
            SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS total_positive_event_rows
        FROM base
    )

    SELECT
        is_login_clean,
        COUNT(*) AS rows,
        COUNT(*) * 1.0 / NULLIF((SELECT total_rows FROM total), 0) AS row_share,

        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS contact_flag_rows,
        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) * 1.0
            / NULLIF((SELECT total_contact_flag_rows FROM total), 0) AS contact_flag_share,

        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS positive_event_rows,
        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) * 1.0
            / NULLIF((SELECT total_positive_event_rows FROM total), 0) AS positive_event_share,

        {USER_COUNT_EXPR} AS users_{COUNT_NOTE},
        {SESSION_COUNT_EXPR} AS sessions_{COUNT_NOTE},
        {ITEM_COUNT_EXPR} AS items_{COUNT_NOTE}
    FROM base
    GROUP BY is_login_clean
    ORDER BY rows DESC
    """
)

events_login_event_type = run_query(
    "16_fact_user_events_by_is_login_event_type",
    f"""
    WITH base AS (
        SELECT
            LOWER(TRIM(COALESCE(CAST(is_login AS VARCHAR), 'unknown'))) AS is_login_clean,
            LOWER(TRIM(COALESCE(CAST(event_type AS VARCHAR), 'unknown'))) AS event_type_clean,

            CAST(user_id AS VARCHAR) AS user_id,
            CAST(session_id AS VARCHAR) AS session_id,
            CAST(item_id AS VARCHAR) AS item_id,

            COALESCE(TRY_CAST(is_contact AS INTEGER), 0) AS is_contact_int
        FROM read_parquet('{EVENTS_GLOB}')
    ),

    total AS (
        SELECT
            COUNT(*) AS total_rows,
            SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS total_contact_flag_rows,
            SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS total_positive_event_rows
        FROM base
    )

    SELECT
        is_login_clean,
        event_type_clean,

        COUNT(*) AS rows,
        COUNT(*) * 1.0 / NULLIF((SELECT total_rows FROM total), 0) AS row_share_of_all_events,

        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS contact_flag_rows,
        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) * 1.0
            / NULLIF((SELECT total_contact_flag_rows FROM total), 0) AS contact_flag_share_of_all_contacts,

        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS positive_event_rows,
        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) * 1.0
            / NULLIF((SELECT total_positive_event_rows FROM total), 0) AS positive_event_share_of_all_positive_events,

        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) * 1.0
            / NULLIF(COUNT(*), 0) AS contact_flag_rate_inside_group,

        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) * 1.0
            / NULLIF(COUNT(*), 0) AS positive_event_rate_inside_group,

        {USER_COUNT_EXPR} AS users_{COUNT_NOTE},
        {SESSION_COUNT_EXPR} AS sessions_{COUNT_NOTE},
        {ITEM_COUNT_EXPR} AS items_{COUNT_NOTE}

    FROM base
    GROUP BY is_login_clean, event_type_clean
    ORDER BY rows DESC
    """,
    show_head=80
)

events_login_event_type_category = run_query(
    "17_fact_user_events_by_is_login_event_type_category",
    f"""
    WITH base AS (
        SELECT
            LOWER(TRIM(COALESCE(CAST(is_login AS VARCHAR), 'unknown'))) AS is_login_clean,
            LOWER(TRIM(COALESCE(CAST(event_type AS VARCHAR), 'unknown'))) AS event_type_clean,

            TRY_CAST(category AS INTEGER) AS category,
            {category_name_sql("category")} AS category_name,

            CAST(user_id AS VARCHAR) AS user_id,
            CAST(session_id AS VARCHAR) AS session_id,
            CAST(item_id AS VARCHAR) AS item_id,

            COALESCE(TRY_CAST(is_contact AS INTEGER), 0) AS is_contact_int
        FROM read_parquet('{EVENTS_GLOB}')
    ),

    total AS (
        SELECT
            COUNT(*) AS total_rows,
            SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS total_contact_flag_rows,
            SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS total_positive_event_rows
        FROM base
    )

    SELECT
        is_login_clean,
        event_type_clean,
        category,
        category_name,

        COUNT(*) AS rows,
        COUNT(*) * 1.0 / NULLIF((SELECT total_rows FROM total), 0) AS row_share_of_all_events,

        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) AS contact_flag_rows,
        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) * 1.0
            / NULLIF((SELECT total_contact_flag_rows FROM total), 0) AS contact_flag_share_of_all_contacts,

        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) AS positive_event_rows,
        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) * 1.0
            / NULLIF((SELECT total_positive_event_rows FROM total), 0) AS positive_event_share_of_all_positive_events,

        SUM(CASE WHEN is_contact_int = 1 THEN 1 ELSE 0 END) * 1.0
            / NULLIF(COUNT(*), 0) AS contact_flag_rate_inside_group,

        SUM(CASE WHEN event_type_clean IN ({POSITIVE_EVENTS_SQL}) THEN 1 ELSE 0 END) * 1.0
            / NULLIF(COUNT(*), 0) AS positive_event_rate_inside_group,

        {USER_COUNT_EXPR} AS users_{COUNT_NOTE},
        {SESSION_COUNT_EXPR} AS sessions_{COUNT_NOTE},
        {ITEM_COUNT_EXPR} AS items_{COUNT_NOTE}

    FROM base
    GROUP BY is_login_clean, event_type_clean, category, category_name
    ORDER BY rows DESC
    """,
    show_head=100
)

# Chart task 4
tmp = events_login_event_type.copy()
tmp["label"] = tmp["is_login_clean"] + " | " + tmp["event_type_clean"]
plot_bar(
    tmp.sort_values("rows", ascending=False),
    "label",
    "rows",
    "Task 4 — fact_user_events row contribution by is_login x event_type",
    "task4_rows_by_is_login_event_type.png",
    top_n=TOP_N
)

tmp = events_login_event_type.copy()
tmp["label"] = tmp["is_login_clean"] + " | " + tmp["event_type_clean"]
plot_bar(
    tmp.sort_values("positive_event_rows", ascending=False),
    "label",
    "positive_event_rows",
    "Task 4 — Positive/contact event contribution by is_login x event_type",
    "task4_positive_events_by_is_login_event_type.png",
    top_n=TOP_N
)

# ============================================================
# FINAL INDEX
# ============================================================

con.close()

print("\nDONE.")
print("Output folder:", OUTPUT_DIR)

print("""
IMPORTANT OUTPUT FILES

TASK 1 — Listing spam / duplicate:
01_dim_listing_primary_key_quality.csv
02_listing_duplicate_by_category_seller_type.csv
03_top_duplicate_listing_groups_to_inspect.csv
04_seller_listing_volume_distribution.csv
05_top_sellers_by_listing_count.csv

TASK 2 — Login active behavior:
06_login_user_active_overall_summary.csv
07_login_user_active_pattern_distribution.csv
08_login_user_active_days_distribution.csv
09_login_user_off_gap_distribution.csv
10_daily_login_active_users.csv

TASK 3 — One-day login users:
11_one_day_login_users_summary.csv
12_one_day_login_users_event_type_category.csv
13_one_day_login_users_first_action_event_type_category.csv
14_one_day_login_users_by_only_active_date.csv

TASK 4 — fact_user_events traffic/contact:
15_fact_user_events_is_login_summary.csv
16_fact_user_events_by_is_login_event_type.csv
17_fact_user_events_by_is_login_event_type_category.csv
""")
```

```python
# ============================================================
# MEMORY-SAFE REPLACEMENT FOR TASK 2 + TASK 3
# Login active days + one-day users
# MapReduce style, avoids DuckDB OOM on raw fact_user_events
# ============================================================

import os
import glob
import gc
import shutil
import subprocess
import sys

import numpy as np
import pandas as pd
import pyarrow.parquet as pq

try:
    import duckdb
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "duckdb"])
    import duckdb

try:
    from IPython.display import display
except Exception:
    def display(x):
        print(x)

# ============================================================
# CONFIG
# ============================================================

BASE_PATH_2 = "/kaggle/input/datasets/kimkhanhnguyen214/datathon2026-2"
EVENTS_PATH = os.path.join(BASE_PATH_2, "fact_user_events")

OUTPUT_DIR = "/kaggle/working/eda_user_lifecycle_light"
os.makedirs(OUTPUT_DIR, exist_ok=True)

PARTIAL_DAILY_DIR = os.path.join(OUTPUT_DIR, "partial_login_daily")
LIFECYCLE_DIR = os.path.join(OUTPUT_DIR, "login_lifecycle_by_bucket")
DAILY_SUMMARY_DIR = os.path.join(OUTPUT_DIR, "daily_summary_by_bucket")

PARTIAL_ONE_DAY_ACTION_DIR = os.path.join(OUTPUT_DIR, "partial_one_day_event_category")
PARTIAL_ONE_DAY_UNIQUE_DIR = os.path.join(OUTPUT_DIR, "partial_one_day_unique_user_event_category")
PARTIAL_ONE_DAY_FIRST_DIR = os.path.join(OUTPUT_DIR, "partial_one_day_first_action")

N_BUCKETS = 64
BATCH_SIZE = 300_000

REGULAR_MAX_OFF_DAYS = 2
LONG_OFF_DAYS = 7

REBUILD_PARTIAL_DAILY = True
REBUILD_ONE_DAY_ACTIONS = True

POSITIVE_EVENTS = {
    "view_phone",
    "contact_chat",
    "other_interaction",
    "contact_zalo",
    "contact_sms",
}

EVENT_COLS_DAILY = [
    "is_login",
    "user_id",
    "date",
    "event_type",
    "is_contact",
]

EVENT_COLS_ACTION = [
    "is_login",
    "user_id",
    "session_id",
    "item_id",
    "event_id",
    "date",
    "event_ts",
    "event_type",
    "category",
    "is_contact",
]

EVENT_FILES = sorted(glob.glob(os.path.join(EVENTS_PATH, "*.parquet")))

print("EVENTS_PATH:", EVENTS_PATH)
print("n event files:", len(EVENT_FILES))
print("OUTPUT_DIR:", OUTPUT_DIR)

# ============================================================
# HELPERS
# ============================================================

def category_name_from_series(s):
    s = pd.to_numeric(s, errors="coerce").astype("Int64")
    mp = {
        1010: "1010_room_rental",
        1020: "1020_apartment",
        1030: "1030_house",
        1040: "1040_land_commercial",
        1050: "1050_new_project",
    }
    return s.map(mp).fillna("unknown_" + s.astype(str).fillna("NULL"))

def clean_event_type(s):
    return s.astype("string").fillna("unknown").str.strip().str.lower()

def clean_is_login(s):
    return s.astype("string").fillna("unknown").str.strip().str.lower()

def make_bucket(user_series, n_buckets=N_BUCKETS):
    return (pd.util.hash_pandas_object(user_series.astype("string"), index=False).values % n_buckets).astype(np.int16)

def save_df(df, name):
    csv_path = os.path.join(OUTPUT_DIR, f"{name}.csv")
    parquet_path = os.path.join(OUTPUT_DIR, f"{name}.parquet")
    df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    df.to_parquet(parquet_path, index=False)
    print("[SAVE]", csv_path)
    display(df.head(30))
    return df

def reset_dir(path):
    if os.path.exists(path):
        shutil.rmtree(path)
    os.makedirs(path, exist_ok=True)

# ============================================================
# STEP A — MAP: raw fact_user_events -> partial user-date aggregates
# ============================================================

if REBUILD_PARTIAL_DAILY:
    reset_dir(PARTIAL_DAILY_DIR)

    part_id = 0

    for file_idx, fp in enumerate(EVENT_FILES):
        print(f"\n[MAP DAILY] file {file_idx + 1}/{len(EVENT_FILES)}: {os.path.basename(fp)}")

        pf = pq.ParquetFile(fp)

        for batch_idx, batch in enumerate(pf.iter_batches(batch_size=BATCH_SIZE, columns=EVENT_COLS_DAILY)):
            df = batch.to_pandas()

            # filter login users only
            df["is_login_clean"] = clean_is_login(df["is_login"])
            df = df[(df["is_login_clean"] == "login") & df["user_id"].notna()]

            if len(df) == 0:
                del df
                gc.collect()
                continue

            df["user_id"] = df["user_id"].astype("string")
            df["active_date"] = pd.to_datetime(df["date"], errors="coerce").dt.floor("D")
            df = df[df["active_date"].notna()]

            if len(df) == 0:
                del df
                gc.collect()
                continue

            df["event_type_clean"] = clean_event_type(df["event_type"])
            df["is_contact_int"] = pd.to_numeric(df["is_contact"], errors="coerce").fillna(0).astype("int8")
            df["positive_int"] = df["event_type_clean"].isin(POSITIVE_EVENTS).astype("int8")
            df["bucket"] = make_bucket(df["user_id"])

            g = (
                df.groupby(["bucket", "user_id", "active_date"], observed=True)
                  .agg(
                      event_rows=("user_id", "size"),
                      contact_flag_rows=("is_contact_int", "sum"),
                      positive_event_rows=("positive_int", "sum"),
                  )
                  .reset_index()
            )

            # save partitioned by bucket to avoid one huge reduce
            for b, gb in g.groupby("bucket", observed=True):
                b = int(b)
                out_dir = os.path.join(PARTIAL_DAILY_DIR, f"bucket={b:03d}")
                os.makedirs(out_dir, exist_ok=True)

                out_path = os.path.join(out_dir, f"part_{part_id:07d}.parquet")
                gb.drop(columns=["bucket"]).to_parquet(out_path, index=False)

            part_id += 1

            del df, g
            gc.collect()

    print("\n[DONE MAP DAILY]")
else:
    print("[SKIP] Using existing partial daily files:", PARTIAL_DAILY_DIR)

# ============================================================
# STEP B — REDUCE BY BUCKET: partial user-date -> lifecycle per user
# ============================================================

reset_dir(LIFECYCLE_DIR)
reset_dir(DAILY_SUMMARY_DIR)

for b in range(N_BUCKETS):
    bucket_glob = os.path.join(PARTIAL_DAILY_DIR, f"bucket={b:03d}", "*.parquet").replace("\\", "/")
    files = glob.glob(bucket_glob)

    if len(files) == 0:
        continue

    print(f"\n[REDUCE BUCKET] {b:03d}, files={len(files)}")

    con = duckdb.connect()
    con.execute("PRAGMA threads=1")
    con.execute("PRAGMA preserve_insertion_order=false")
    con.execute("PRAGMA memory_limit='3GB'")

    daily = con.execute(f"""
        SELECT
            CAST(user_id AS VARCHAR) AS user_id,
            CAST(active_date AS DATE) AS active_date,
            SUM(event_rows) AS event_rows,
            SUM(contact_flag_rows) AS contact_flag_rows,
            SUM(positive_event_rows) AS positive_event_rows
        FROM read_parquet('{bucket_glob}')
        GROUP BY user_id, active_date
        ORDER BY user_id, active_date
    """).df()

    con.close()

    if len(daily) == 0:
        del daily
        gc.collect()
        continue

    daily["active_date"] = pd.to_datetime(daily["active_date"])

    # daily summary partial
    daily_summary = (
        daily.groupby("active_date", as_index=False)
             .agg(
                 login_active_users=("user_id", "nunique"),
                 event_rows=("event_rows", "sum"),
                 contact_flag_rows=("contact_flag_rows", "sum"),
                 positive_event_rows=("positive_event_rows", "sum"),
             )
    )
    daily_summary.to_parquet(
        os.path.join(DAILY_SUMMARY_DIR, f"daily_summary_bucket_{b:03d}.parquet"),
        index=False
    )

    # lifecycle
    daily = daily.sort_values(["user_id", "active_date"])
    daily["prev_active_date"] = daily.groupby("user_id")["active_date"].shift(1)
    daily["inactive_gap_days"] = (daily["active_date"] - daily["prev_active_date"]).dt.days - 1

    lifecycle = (
        daily.groupby("user_id", as_index=False)
             .agg(
                 active_days=("active_date", "nunique"),
                 first_active_date=("active_date", "min"),
                 last_active_date=("active_date", "max"),
                 max_inactive_gap_days=("inactive_gap_days", "max"),
                 avg_inactive_gap_days=("inactive_gap_days", "mean"),
                 long_off_gap_count=("inactive_gap_days", lambda x: int((x >= LONG_OFF_DAYS).sum())),
                 total_event_rows=("event_rows", "sum"),
                 total_contact_flag_rows=("contact_flag_rows", "sum"),
                 total_positive_event_rows=("positive_event_rows", "sum"),
             )
    )

    lifecycle["max_inactive_gap_days"] = lifecycle["max_inactive_gap_days"].fillna(0).astype("int32")
    lifecycle["avg_inactive_gap_days"] = lifecycle["avg_inactive_gap_days"].fillna(0.0)
    lifecycle["observed_span_days"] = (
        lifecycle["last_active_date"] - lifecycle["first_active_date"]
    ).dt.days + 1

    lifecycle["active_ratio_inside_observed_span"] = (
        lifecycle["active_days"] / lifecycle["observed_span_days"].replace(0, np.nan)
    )

    lifecycle["active_pattern"] = np.select(
        [
            lifecycle["active_days"] == 1,
            lifecycle["max_inactive_gap_days"] <= REGULAR_MAX_OFF_DAYS,
            lifecycle["max_inactive_gap_days"] >= LONG_OFF_DAYS,
        ],
        [
            "one_day_only",
            f"regular_off_0_to_{REGULAR_MAX_OFF_DAYS}_days",
            f"long_off_{LONG_OFF_DAYS}_plus_days",
        ],
        default=f"medium_off_between_{REGULAR_MAX_OFF_DAYS + 1}_and_{LONG_OFF_DAYS - 1}_days"
    )

    lifecycle.to_parquet(
        os.path.join(LIFECYCLE_DIR, f"lifecycle_bucket_{b:03d}.parquet"),
        index=False
    )

    print("daily rows:", len(daily), "| lifecycle users:", len(lifecycle))

    del daily, daily_summary, lifecycle
    gc.collect()

print("\n[DONE REDUCE LIFECYCLE]")

# ============================================================
# STEP C — SUMMARY TABLES FOR TASK 2
# ============================================================

con = duckdb.connect()
con.execute("PRAGMA threads=1")
con.execute("PRAGMA preserve_insertion_order=false")
con.execute("PRAGMA memory_limit='3GB'")

LIFECYCLE_GLOB = os.path.join(LIFECYCLE_DIR, "*.parquet").replace("\\", "/")
DAILY_SUMMARY_GLOB = os.path.join(DAILY_SUMMARY_DIR, "*.parquet").replace("\\", "/")

login_overall_summary = con.execute(f"""
SELECT
    COUNT(*) AS total_login_users,

    AVG(active_days) AS avg_active_days,
    QUANTILE_CONT(active_days, 0.50) AS median_active_days,
    QUANTILE_CONT(active_days, 0.75) AS p75_active_days,
    QUANTILE_CONT(active_days, 0.90) AS p90_active_days,
    QUANTILE_CONT(active_days, 0.95) AS p95_active_days,
    QUANTILE_CONT(active_days, 0.99) AS p99_active_days,
    MAX(active_days) AS max_active_days,

    AVG(max_inactive_gap_days) AS avg_max_inactive_gap_days,
    QUANTILE_CONT(max_inactive_gap_days, 0.50) AS median_max_inactive_gap_days,
    QUANTILE_CONT(max_inactive_gap_days, 0.90) AS p90_max_inactive_gap_days,
    QUANTILE_CONT(max_inactive_gap_days, 0.95) AS p95_max_inactive_gap_days,
    QUANTILE_CONT(max_inactive_gap_days, 0.99) AS p99_max_inactive_gap_days,
    MAX(max_inactive_gap_days) AS max_inactive_gap_days,

    SUM(CASE WHEN active_days = 1 THEN 1 ELSE 0 END) AS one_day_users,
    SUM(CASE WHEN active_days = 1 THEN 1 ELSE 0 END) * 1.0 / NULLIF(COUNT(*), 0) AS one_day_user_rate
FROM read_parquet('{LIFECYCLE_GLOB}')
""").df()
save_df(login_overall_summary, "06_login_user_active_overall_summary")

active_pattern_dist = con.execute(f"""
SELECT
    active_pattern,
    COUNT(*) AS users,
    COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share,
    AVG(active_days) AS avg_active_days,
    AVG(total_event_rows) AS avg_event_rows,
    AVG(total_contact_flag_rows) AS avg_contact_flag_rows,
    AVG(total_positive_event_rows) AS avg_positive_event_rows
FROM read_parquet('{LIFECYCLE_GLOB}')
GROUP BY active_pattern
ORDER BY users DESC
""").df()
save_df(active_pattern_dist, "07_login_user_active_pattern_distribution")

active_days_dist = con.execute(f"""
WITH b AS (
    SELECT
        CASE
            WHEN active_days = 1 THEN '01_1_day'
            WHEN active_days BETWEEN 2 AND 3 THEN '02_2_3_days'
            WHEN active_days BETWEEN 4 AND 7 THEN '03_4_7_days'
            WHEN active_days BETWEEN 8 AND 14 THEN '04_8_14_days'
            WHEN active_days BETWEEN 15 AND 30 THEN '05_15_30_days'
            WHEN active_days BETWEEN 31 AND 60 THEN '06_31_60_days'
            WHEN active_days BETWEEN 61 AND 90 THEN '07_61_90_days'
            ELSE '08_91plus_days'
        END AS active_days_bucket,

        CASE
            WHEN active_days = 1 THEN 1
            WHEN active_days BETWEEN 2 AND 3 THEN 2
            WHEN active_days BETWEEN 4 AND 7 THEN 3
            WHEN active_days BETWEEN 8 AND 14 THEN 4
            WHEN active_days BETWEEN 15 AND 30 THEN 5
            WHEN active_days BETWEEN 31 AND 60 THEN 6
            WHEN active_days BETWEEN 61 AND 90 THEN 7
            ELSE 8
        END AS sort_key,

        active_pattern
    FROM read_parquet('{LIFECYCLE_GLOB}')
)

SELECT
    sort_key,
    active_days_bucket,
    active_pattern,
    COUNT(*) AS users,
    COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share
FROM b
GROUP BY sort_key, active_days_bucket, active_pattern
ORDER BY sort_key, active_pattern
""").df()
save_df(active_days_dist, "08_login_user_active_days_distribution")

off_gap_dist = con.execute(f"""
WITH b AS (
    SELECT
        CASE
            WHEN max_inactive_gap_days = 0 THEN '00_no_gap_or_one_day'
            WHEN max_inactive_gap_days BETWEEN 1 AND {REGULAR_MAX_OFF_DAYS} THEN '01_off_1_to_{REGULAR_MAX_OFF_DAYS}_days'
            WHEN max_inactive_gap_days BETWEEN {REGULAR_MAX_OFF_DAYS + 1} AND {LONG_OFF_DAYS - 1} THEN '02_off_{REGULAR_MAX_OFF_DAYS + 1}_to_{LONG_OFF_DAYS - 1}_days'
            WHEN max_inactive_gap_days BETWEEN {LONG_OFF_DAYS} AND 14 THEN '03_off_{LONG_OFF_DAYS}_to_14_days'
            WHEN max_inactive_gap_days BETWEEN 15 AND 30 THEN '04_off_15_to_30_days'
            ELSE '05_off_31plus_days'
        END AS off_gap_bucket,

        CASE
            WHEN max_inactive_gap_days = 0 THEN 0
            WHEN max_inactive_gap_days BETWEEN 1 AND {REGULAR_MAX_OFF_DAYS} THEN 1
            WHEN max_inactive_gap_days BETWEEN {REGULAR_MAX_OFF_DAYS + 1} AND {LONG_OFF_DAYS - 1} THEN 2
            WHEN max_inactive_gap_days BETWEEN {LONG_OFF_DAYS} AND 14 THEN 3
            WHEN max_inactive_gap_days BETWEEN 15 AND 30 THEN 4
            ELSE 5
        END AS sort_key,

        active_pattern
    FROM read_parquet('{LIFECYCLE_GLOB}')
)

SELECT
    sort_key,
    off_gap_bucket,
    active_pattern,
    COUNT(*) AS users,
    COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share
FROM b
GROUP BY sort_key, off_gap_bucket, active_pattern
ORDER BY sort_key, active_pattern
""").df()
save_df(off_gap_dist, "09_login_user_off_gap_distribution")

daily_login_users = con.execute(f"""
SELECT
    active_date,
    SUM(login_active_users) AS login_active_users,
    SUM(event_rows) AS event_rows,
    SUM(contact_flag_rows) AS contact_flag_rows,
    SUM(positive_event_rows) AS positive_event_rows,
    SUM(contact_flag_rows) * 1.0 / NULLIF(SUM(event_rows), 0) AS contact_flag_rate,
    SUM(positive_event_rows) * 1.0 / NULLIF(SUM(event_rows), 0) AS positive_event_rate
FROM read_parquet('{DAILY_SUMMARY_GLOB}')
GROUP BY active_date
ORDER BY active_date
""").df()
save_df(daily_login_users, "10_daily_login_active_users")

# ============================================================
# STEP D — ONE-DAY USERS TABLE
# ============================================================

one_day_users = con.execute(f"""
SELECT
    user_id,
    first_active_date AS only_active_date,
    total_event_rows,
    total_contact_flag_rows,
    total_positive_event_rows
FROM read_parquet('{LIFECYCLE_GLOB}')
WHERE active_days = 1
""").df()

one_day_path = os.path.join(OUTPUT_DIR, "one_day_login_users.parquet")
one_day_users.to_parquet(one_day_path, index=False)

one_day_summary = pd.DataFrame([{
    "one_day_login_users": len(one_day_users),
    "total_login_users": int(login_overall_summary.loc[0, "total_login_users"]),
    "one_day_login_user_rate": len(one_day_users) / max(int(login_overall_summary.loc[0, "total_login_users"]), 1),
    "avg_event_rows_in_that_day": one_day_users["total_event_rows"].mean(),
    "median_event_rows_in_that_day": one_day_users["total_event_rows"].median(),
    "p90_event_rows_in_that_day": one_day_users["total_event_rows"].quantile(0.90),
    "max_event_rows_in_that_day": one_day_users["total_event_rows"].max(),
    "total_contact_flag_rows_by_one_day_users": one_day_users["total_contact_flag_rows"].sum(),
    "total_positive_event_rows_by_one_day_users": one_day_users["total_positive_event_rows"].sum(),
    "one_day_users_with_contact_flag": int((one_day_users["total_contact_flag_rows"] > 0).sum()),
    "one_day_users_with_positive_event": int((one_day_users["total_positive_event_rows"] > 0).sum()),
}])
save_df(one_day_summary, "11_one_day_login_users_summary")

one_day_active_date_dist = (
    one_day_users.groupby("only_active_date", as_index=False)
                 .agg(
                     one_day_users=("user_id", "count"),
                     event_rows=("total_event_rows", "sum"),
                     contact_flag_rows=("total_contact_flag_rows", "sum"),
                     positive_event_rows=("total_positive_event_rows", "sum"),
                 )
                 .sort_values("only_active_date")
)
save_df(one_day_active_date_dist, "14_one_day_login_users_by_only_active_date")

con.close()

# ============================================================
# STEP E — MAP ONE-DAY USER ACTIONS
# This scans raw events again, but only keeps one-day login users.
# ============================================================

if REBUILD_ONE_DAY_ACTIONS:
    reset_dir(PARTIAL_ONE_DAY_ACTION_DIR)
    reset_dir(PARTIAL_ONE_DAY_UNIQUE_DIR)
    reset_dir(PARTIAL_ONE_DAY_FIRST_DIR)

    one_day_set = set(one_day_users["user_id"].astype("string"))
    print("\none_day user set size:", len(one_day_set))

    part_id = 0

    for file_idx, fp in enumerate(EVENT_FILES):
        print(f"\n[MAP ONE-DAY ACTION] file {file_idx + 1}/{len(EVENT_FILES)}: {os.path.basename(fp)}")

        pf = pq.ParquetFile(fp)

        for batch_idx, batch in enumerate(pf.iter_batches(batch_size=BATCH_SIZE, columns=EVENT_COLS_ACTION)):
            df = batch.to_pandas()

            df["is_login_clean"] = clean_is_login(df["is_login"])
            df = df[(df["is_login_clean"] == "login") & df["user_id"].notna()]

            if len(df) == 0:
                del df
                gc.collect()
                continue

            df["user_id"] = df["user_id"].astype("string")
            df = df[df["user_id"].isin(one_day_set)]

            if len(df) == 0:
                del df
                gc.collect()
                continue

            df["event_type_clean"] = clean_event_type(df["event_type"])
            df["category"] = pd.to_numeric(df["category"], errors="coerce").astype("Int64")
            df["category_name"] = category_name_from_series(df["category"])

            df["is_contact_int"] = pd.to_numeric(df["is_contact"], errors="coerce").fillna(0).astype("int8")
            df["positive_int"] = df["event_type_clean"].isin(POSITIVE_EVENTS).astype("int8")

            # rows by event_type x category
            g_rows = (
                df.groupby(["event_type_clean", "category", "category_name"], dropna=False, observed=True)
                  .agg(
                      rows=("user_id", "size"),
                      contact_flag_rows=("is_contact_int", "sum"),
                      positive_event_rows=("positive_int", "sum"),
                  )
                  .reset_index()
            )
            g_rows.to_parquet(
                os.path.join(PARTIAL_ONE_DAY_ACTION_DIR, f"part_{part_id:07d}.parquet"),
                index=False
            )

            # exact user count later: save unique user-event-category pairs
            g_unique = df[["user_id", "event_type_clean", "category", "category_name"]].drop_duplicates()
            g_unique.to_parquet(
                os.path.join(PARTIAL_ONE_DAY_UNIQUE_DIR, f"part_{part_id:07d}.parquet"),
                index=False
            )

            # first action later: keep first row per user inside this batch
            df["event_ts_clean"] = pd.to_datetime(df["event_ts"], errors="coerce")
            df["event_id_clean"] = df["event_id"].astype("string").fillna("")
            first_part = (
                df.sort_values(["user_id", "event_ts_clean", "event_id_clean"])
                  .drop_duplicates("user_id", keep="first")
                  [["user_id", "event_ts_clean", "event_id_clean", "event_type_clean", "category", "category_name"]]
            )
            first_part.to_parquet(
                os.path.join(PARTIAL_ONE_DAY_FIRST_DIR, f"part_{part_id:07d}.parquet"),
                index=False
            )

            part_id += 1

            del df, g_rows, g_unique, first_part
            gc.collect()

    print("\n[DONE MAP ONE-DAY ACTIONS]")

# ============================================================
# STEP F — REDUCE ONE-DAY ACTION TABLES
# ============================================================

con = duckdb.connect()
con.execute("PRAGMA threads=1")
con.execute("PRAGMA preserve_insertion_order=false")
con.execute("PRAGMA memory_limit='3GB'")

ACTION_GLOB = os.path.join(PARTIAL_ONE_DAY_ACTION_DIR, "*.parquet").replace("\\", "/")
UNIQUE_GLOB = os.path.join(PARTIAL_ONE_DAY_UNIQUE_DIR, "*.parquet").replace("\\", "/")
FIRST_GLOB = os.path.join(PARTIAL_ONE_DAY_FIRST_DIR, "*.parquet").replace("\\", "/")

one_day_rows = con.execute(f"""
SELECT
    event_type_clean,
    category,
    category_name,
    SUM(rows) AS rows,
    SUM(contact_flag_rows) AS contact_flag_rows,
    SUM(positive_event_rows) AS positive_event_rows
FROM read_parquet('{ACTION_GLOB}')
GROUP BY event_type_clean, category, category_name
""").df()

one_day_users_by_action = con.execute(f"""
SELECT
    event_type_clean,
    category,
    category_name,
    COUNT(DISTINCT user_id) AS users
FROM read_parquet('{UNIQUE_GLOB}')
GROUP BY event_type_clean, category, category_name
""").df()

one_day_event_type_category = one_day_rows.merge(
    one_day_users_by_action,
    on=["event_type_clean", "category", "category_name"],
    how="left"
)

one_day_event_type_category["row_share"] = (
    one_day_event_type_category["rows"] / one_day_event_type_category["rows"].sum()
)
one_day_event_type_category["user_share_among_one_day_users"] = (
    one_day_event_type_category["users"] / max(len(one_day_users), 1)
)

one_day_event_type_category = one_day_event_type_category.sort_values("rows", ascending=False)
save_df(one_day_event_type_category, "12_one_day_login_users_event_type_category")

one_day_first_action = con.execute(f"""
WITH ranked AS (
    SELECT
        user_id,
        event_type_clean,
        category,
        category_name,
        ROW_NUMBER() OVER (
            PARTITION BY user_id
            ORDER BY event_ts_clean ASC NULLS LAST, event_id_clean ASC NULLS LAST
        ) AS rn
    FROM read_parquet('{FIRST_GLOB}')
)

SELECT
    event_type_clean AS first_event_type,
    category,
    category_name,
    COUNT(*) AS users,
    COUNT(*) * 1.0 / SUM(COUNT(*)) OVER () AS user_share
FROM ranked
WHERE rn = 1
GROUP BY event_type_clean, category, category_name
ORDER BY users DESC
""").df()

save_df(one_day_first_action, "13_one_day_login_users_first_action_event_type_category")

con.close()

print("\nDONE.")
print("Output folder:", OUTPUT_DIR)
print("""
Important outputs:
06_login_user_active_overall_summary.csv
07_login_user_active_pattern_distribution.csv
08_login_user_active_days_distribution.csv
09_login_user_off_gap_distribution.csv
10_daily_login_active_users.csv
11_one_day_login_users_summary.csv
12_one_day_login_users_event_type_category.csv
13_one_day_login_users_first_action_event_type_category.csv
14_one_day_login_users_by_only_active_date.csv
""")
```

```python
# ============================================================
# DATATHON 2026 — VISUALIZE ALL PRECOMPUTED OUTPUTS
# Input: Kaggle dataset chứa output .csv/.parquet từ notebook trước
# Output: /kaggle/working/visual_all_outputs
# ============================================================

import os
import glob
import sys
import subprocess
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

try:
    import plotly.express as px
    import plotly.graph_objects as go
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "plotly"])
    import plotly.express as px
    import plotly.graph_objects as go

try:
    from IPython.display import display
except Exception:
    def display(x):
        print(x)

# ============================================================
# 0. CONFIG
# ============================================================

INPUT_ROOT = "/kaggle/input/notebooks/kimkhanhnguyen214/user_login_n_active"
OUTPUT_DIR = "/kaggle/working/visual_all_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

TOP_N = 30
SAVE_PNG = False   # set True nếu đã cài kaleido
SHOW_FIG = True

print("INPUT_ROOT:", INPUT_ROOT)
print("OUTPUT_DIR:", OUTPUT_DIR)

# ============================================================
# 1. DISCOVER ALL OUTPUT FILES
# ============================================================

all_files = []
for ext in ["*.parquet", "*.csv"]:
    all_files.extend(glob.glob(os.path.join(INPUT_ROOT, "**", ext), recursive=True))

all_files = sorted(all_files)

manifest = pd.DataFrame({
    "path": all_files,
    "filename": [os.path.basename(p) for p in all_files],
    "stem": [os.path.splitext(os.path.basename(p))[0] for p in all_files],
    "ext": [os.path.splitext(p)[1].lower() for p in all_files],
    "size_mb": [os.path.getsize(p) / 1024 / 1024 for p in all_files],
})

manifest_path = os.path.join(OUTPUT_DIR, "00_input_file_manifest.csv")
manifest.to_csv(manifest_path, index=False, encoding="utf-8-sig")

print("Found files:", len(manifest))
display(manifest.head(50))
print("[SAVE]", manifest_path)

# Prefer parquet over csv if both exist
def find_path(stem):
    candidates = manifest[manifest["stem"] == stem].copy()
    if len(candidates) == 0:
        return None

    candidates["priority"] = candidates["ext"].map({".parquet": 0, ".csv": 1}).fillna(9)
    candidates = candidates.sort_values(["priority", "size_mb"], ascending=[True, False])
    return candidates.iloc[0]["path"]

def read_table(stem, required_cols=None, verbose=True):
    path = find_path(stem)

    if path is None:
        if verbose:
            print("[MISSING]", stem)
        return None

    try:
        if path.endswith(".parquet"):
            df = pd.read_parquet(path)
        else:
            df = pd.read_csv(path)

        if required_cols is not None:
            missing = [c for c in required_cols if c not in df.columns]
            if missing:
                print("[SKIP]", stem, "| missing columns:", missing)
                return None

        print("[LOAD]", stem, "| rows:", len(df), "| path:", path)
        return df

    except Exception as e:
        print("[ERROR LOAD]", stem, path, e)
        return None

# ============================================================
# 2. HELPERS
# ============================================================

chart_index = []

def plotly_safe_df(df):
    if df is None:
        return None
    return df.copy().astype(object).where(pd.notna(df), None)

def save_fig(fig, filename, title=None):
    html_path = os.path.join(OUTPUT_DIR, f"{filename}.html")
    fig.write_html(html_path)
    print("[SAVE HTML]", html_path)

    png_path = None
    if SAVE_PNG:
        try:
            png_path = os.path.join(OUTPUT_DIR, f"{filename}.png")
            fig.write_image(png_path, scale=2)
            print("[SAVE PNG]", png_path)
        except Exception as e:
            print("[SKIP PNG]", filename, e)

    chart_index.append({
        "filename": filename,
        "title": title if title else filename,
        "html_path": html_path,
        "png_path": png_path,
    })

    if SHOW_FIG:
        fig.show()

def to_numeric_safe(df, cols):
    df = df.copy()
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def add_pct_cols(df, rate_cols):
    df = df.copy()
    for c in rate_cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
            # Nếu rate dạng 0.04 thì đổi sang %
            if df[c].dropna().max() <= 1.5:
                df[c + "_pct"] = df[c] * 100
            else:
                df[c + "_pct"] = df[c]
    return df

def make_label(df):
    df = df.copy()

    if "seller_type_clean" in df.columns and "category_name" in df.columns:
        df["label"] = df["seller_type_clean"].astype(str) + " | " + df["category_name"].astype(str)
    elif "seller_type_mix" in df.columns and "category_name" in df.columns:
        df["label"] = df["seller_type_mix"].astype(str) + " | " + df["category_name"].astype(str)
    elif "event_type_clean" in df.columns and "category_name" in df.columns:
        df["label"] = df["event_type_clean"].astype(str) + " | " + df["category_name"].astype(str)
    elif "first_event_type" in df.columns and "category_name" in df.columns:
        df["label"] = df["first_event_type"].astype(str) + " | " + df["category_name"].astype(str)
    elif "is_login_clean" in df.columns and "event_type_clean" in df.columns:
        df["label"] = df["is_login_clean"].astype(str) + " | " + df["event_type_clean"].astype(str)
    else:
        df["label"] = np.arange(len(df)).astype(str)

    return df

def save_table_preview(df, name, head=50):
    if df is None:
        return
    path = os.path.join(OUTPUT_DIR, f"{name}.csv")
    df.head(head).to_csv(path, index=False, encoding="utf-8-sig")
    print("[SAVE PREVIEW]", path)

# ============================================================
# 3. LISTING SPAM — RULE V3 IGNORE TITLE
# ============================================================

rule_v3_summary = read_table("40_rule_v3_ignore_title_summary_by_seller_type_category")

if rule_v3_summary is not None:
    df = make_label(rule_v3_summary)
    df = to_numeric_safe(df, [
        "total_items",
        "total_sellers",
        "duplicate_groups",
        "items_in_duplicate_groups",
        "extra_duplicate_items",
        "repost_items",
        "overlap_repost_items_old_not_expired",
        "duplicate_item_rate",
        "overlap_spam_item_rate",
        "overlap_spam_seller_rate",
        "avg_days_between_reposts",
        "median_days_between_reposts",
        "p90_days_between_reposts",
        "avg_overlap_days_when_spam",
        "median_overlap_days_when_spam",
        "max_items_in_one_group",
    ])
    df = add_pct_cols(df, [
        "duplicate_item_rate",
        "overlap_spam_item_rate",
        "overlap_spam_seller_rate",
    ])

    # 3.1 Counts
    plot_df = df.sort_values("items_in_duplicate_groups", ascending=False).head(TOP_N)
    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y=["items_in_duplicate_groups", "overlap_repost_items_old_not_expired"],
        barmode="group",
        hover_data=[
            "total_items", "total_sellers", "duplicate_groups",
            "extra_duplicate_items", "repost_items",
            "duplicate_item_rate_pct", "overlap_spam_item_rate_pct",
            "overlap_spam_seller_rate_pct",
            "avg_days_between_reposts", "median_days_between_reposts",
            "avg_overlap_days_when_spam", "max_items_in_one_group",
        ],
        title="Rule V3 — Duplicate/spam counts by seller type × category",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Items",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "01_rule_v3_duplicate_spam_counts", "Rule V3 counts")

    # 3.2 Rates
    plot_df = df.sort_values("overlap_spam_item_rate_pct", ascending=False).head(TOP_N)
    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y=["duplicate_item_rate_pct", "overlap_spam_item_rate_pct", "overlap_spam_seller_rate_pct"],
        barmode="group",
        hover_data=[
            "total_items", "total_sellers",
            "items_in_duplicate_groups",
            "overlap_repost_items_old_not_expired",
            "avg_days_between_reposts",
            "median_days_between_reposts",
            "avg_overlap_days_when_spam",
            "max_items_in_one_group",
        ],
        title="Rule V3 — Duplicate/spam rates by seller type × category",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Rate (%)",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "02_rule_v3_duplicate_spam_rates", "Rule V3 rates")

    # 3.3 Heatmap overlap spam rate
    if {"seller_type_clean", "category_name", "overlap_spam_item_rate_pct"}.issubset(df.columns):
        pivot = df.pivot_table(
            index="category_name",
            columns="seller_type_clean",
            values="overlap_spam_item_rate_pct",
            aggfunc="mean",
        )

        fig = px.imshow(
            pivot,
            text_auto=".2f",
            aspect="auto",
            title="Rule V3 — Overlap spam item rate heatmap (%)",
        )
        fig.update_layout(height=550)
        save_fig(fig, "03_rule_v3_overlap_spam_rate_heatmap", "Rule V3 heatmap")

    # 3.4 Bubble: duplicate rate vs overlap spam rate
    fig = px.scatter(
        plotly_safe_df(df),
        x="duplicate_item_rate_pct",
        y="overlap_spam_item_rate_pct",
        size="total_items",
        color="seller_type_clean" if "seller_type_clean" in df.columns else None,
        hover_name="label",
        hover_data=[
            "total_items", "total_sellers",
            "items_in_duplicate_groups",
            "overlap_repost_items_old_not_expired",
            "avg_days_between_reposts",
            "median_days_between_reposts",
            "max_items_in_one_group",
        ],
        title="Rule V3 — Duplicate rate vs overlap spam rate",
    )
    fig.update_layout(
        xaxis_title="Duplicate item rate (%)",
        yaxis_title="Overlap spam item rate (%)",
        height=620,
    )
    save_fig(fig, "04_rule_v3_duplicate_vs_overlap_scatter", "Rule V3 scatter")

    # 3.5 Gap between reposts
    plot_df = df.sort_values("median_days_between_reposts", ascending=True).head(TOP_N)
    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="median_days_between_reposts",
        color="seller_type_clean" if "seller_type_clean" in plot_df.columns else None,
        hover_data=[
            "avg_days_between_reposts",
            "p90_days_between_reposts",
            "avg_overlap_days_when_spam",
            "items_in_duplicate_groups",
            "overlap_repost_items_old_not_expired",
        ],
        title="Rule V3 — Median days between same-info reposts",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Median days",
        xaxis_tickangle=-35,
        height=650,
    )
    save_fig(fig, "05_rule_v3_repost_gap_median", "Rule V3 repost gap")

# ============================================================
# 4. RULE V3 TIMELINE
# ============================================================

timeline_v3 = read_table("41_rule_v3_ignore_title_daily_spam_active_rate_overall")

if timeline_v3 is not None:
    df = timeline_v3.copy()
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df = df.sort_values("date")
    df = to_numeric_safe(df, [
        "total_active_items",
        "spam_active_items",
        "spam_active_rate",
        "spam_active_rate_pct",
    ])

    if "spam_active_rate_pct" not in df.columns and "spam_active_rate" in df.columns:
        df["spam_active_rate_pct"] = df["spam_active_rate"] * 100

    df["spam_active_rate_pct_7d_ma"] = df["spam_active_rate_pct"].rolling(7, min_periods=1).mean()
    df["spam_active_rate_pct_30d_ma"] = df["spam_active_rate_pct"].rolling(30, min_periods=1).mean()

    # 4.1 Rate timeline
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df["date"],
        y=df["spam_active_rate_pct"],
        mode="lines",
        name="Daily",
        hovertemplate="Date=%{x}<br>Rate=%{y:.4f}%<extra></extra>",
    ))
    fig.add_trace(go.Scatter(
        x=df["date"],
        y=df["spam_active_rate_pct_7d_ma"],
        mode="lines",
        name="7D moving avg",
        hovertemplate="Date=%{x}<br>7D MA=%{y:.4f}%<extra></extra>",
    ))
    fig.add_trace(go.Scatter(
        x=df["date"],
        y=df["spam_active_rate_pct_30d_ma"],
        mode="lines",
        name="30D moving avg",
        hovertemplate="Date=%{x}<br>30D MA=%{y:.4f}%<extra></extra>",
    ))
    fig.update_layout(
        title="Rule V3 — % spam-active listings among all active listings",
        xaxis_title="Date",
        yaxis_title="Spam-active / total active listings (%)",
        height=620,
    )
    save_fig(fig, "06_rule_v3_timeline_spam_active_rate_overall", "Rule V3 timeline rate")

    # 4.2 Count timeline
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df["date"],
        y=df["total_active_items"],
        mode="lines",
        name="Total active listings",
        hovertemplate="Date=%{x}<br>Total active=%{y:,.0f}<extra></extra>",
    ))
    fig.add_trace(go.Scatter(
        x=df["date"],
        y=df["spam_active_items"],
        mode="lines",
        name="Spam-active listings",
        hovertemplate="Date=%{x}<br>Spam-active=%{y:,.0f}<extra></extra>",
    ))
    fig.update_layout(
        title="Rule V3 — Total active vs spam-active listings",
        xaxis_title="Date",
        yaxis_title="Listings",
        height=620,
    )
    save_fig(fig, "07_rule_v3_timeline_active_vs_spam_counts", "Rule V3 timeline counts")

    # 4.3 Monthly avg
    m = df.dropna(subset=["date"]).copy()
    m["month"] = m["date"].dt.to_period("M").dt.to_timestamp()
    monthly = (
        m.groupby("month", as_index=False)
         .agg(
             avg_total_active_items=("total_active_items", "mean"),
             avg_spam_active_items=("spam_active_items", "mean"),
             avg_spam_active_rate_pct=("spam_active_rate_pct", "mean"),
             max_spam_active_rate_pct=("spam_active_rate_pct", "max"),
         )
    )

    fig = px.bar(
        plotly_safe_df(monthly),
        x="month",
        y="avg_spam_active_rate_pct",
        hover_data=[
            "avg_total_active_items",
            "avg_spam_active_items",
            "max_spam_active_rate_pct",
        ],
        title="Rule V3 — Monthly average spam-active rate",
    )
    fig.update_layout(
        xaxis_title="Month",
        yaxis_title="Avg spam-active rate (%)",
        height=560,
    )
    save_fig(fig, "08_rule_v3_monthly_avg_spam_active_rate", "Rule V3 monthly avg")

timeline_v3_type = read_table("42_rule_v3_ignore_title_daily_spam_active_rate_by_seller_type")

if timeline_v3_type is not None:
    df = timeline_v3_type.copy()
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df = df.sort_values(["seller_type_clean", "date"])
    df = to_numeric_safe(df, [
        "total_active_items",
        "spam_active_items",
        "spam_active_rate",
        "spam_active_rate_pct",
    ])

    if "spam_active_rate_pct" not in df.columns and "spam_active_rate" in df.columns:
        df["spam_active_rate_pct"] = df["spam_active_rate"] * 100

    # 4.4 Timeline by seller type
    fig = px.line(
        plotly_safe_df(df),
        x="date",
        y="spam_active_rate_pct",
        color="seller_type_clean",
        hover_data=["total_active_items", "spam_active_items"],
        title="Rule V3 — % spam-active listings over time by seller type",
    )
    fig.update_layout(
        xaxis_title="Date",
        yaxis_title="Spam-active / total active listings (%)",
        height=650,
    )
    save_fig(fig, "09_rule_v3_timeline_spam_active_rate_by_seller_type", "Rule V3 by seller type")

    # 4.5 Monthly by seller type
    m = df.dropna(subset=["date"]).copy()
    m["month"] = m["date"].dt.to_period("M").dt.to_timestamp()
    monthly = (
        m.groupby(["month", "seller_type_clean"], as_index=False)
         .agg(
             avg_spam_active_rate_pct=("spam_active_rate_pct", "mean"),
             avg_total_active_items=("total_active_items", "mean"),
             avg_spam_active_items=("spam_active_items", "mean"),
         )
    )

    fig = px.line(
        plotly_safe_df(monthly),
        x="month",
        y="avg_spam_active_rate_pct",
        color="seller_type_clean",
        markers=True,
        hover_data=["avg_total_active_items", "avg_spam_active_items"],
        title="Rule V3 — Monthly average spam-active rate by seller type",
    )
    fig.update_layout(
        xaxis_title="Month",
        yaxis_title="Avg spam-active rate (%)",
        height=600,
    )
    save_fig(fig, "10_rule_v3_monthly_spam_active_rate_by_seller_type", "Rule V3 monthly by type")

# ============================================================
# 5. COMPARE RULES: STRICT vs NOIMG vs V3
# ============================================================

rule_sources = [
    ("Strict exact", "01_same_seller_exact_duplicate_summary_by_seller_type_category"),
    ("No image-count rule", "21_noimg_rule_summary_by_seller_type_category"),
    ("No image-count rule 2", "30_noimg_rule_summary_by_seller_type_category"),
    ("Rule V3 ignore title", "40_rule_v3_ignore_title_summary_by_seller_type_category"),
]

compare_parts = []

for rule_name, stem in rule_sources:
    tmp = read_table(stem, verbose=False)
    if tmp is None:
        continue

    tmp = make_label(tmp)
    tmp = to_numeric_safe(tmp, [
        "total_items",
        "items_in_duplicate_groups",
        "items_in_same_seller_exact_duplicate_groups",
        "overlap_repost_items_old_not_expired",
        "duplicate_item_rate",
        "exact_duplicate_item_rate",
        "overlap_spam_item_rate",
        "overlap_spam_seller_rate",
    ])

    if "items_in_duplicate_groups" not in tmp.columns and "items_in_same_seller_exact_duplicate_groups" in tmp.columns:
        tmp["items_in_duplicate_groups"] = tmp["items_in_same_seller_exact_duplicate_groups"]

    if "duplicate_item_rate" not in tmp.columns and "exact_duplicate_item_rate" in tmp.columns:
        tmp["duplicate_item_rate"] = tmp["exact_duplicate_item_rate"]

    tmp["rule"] = rule_name
    tmp = add_pct_cols(tmp, [
        "duplicate_item_rate",
        "overlap_spam_item_rate",
        "overlap_spam_seller_rate",
    ])

    compare_parts.append(tmp)

if len(compare_parts) > 0:
    compare_df = pd.concat(compare_parts, ignore_index=True)
    compare_df.to_csv(os.path.join(OUTPUT_DIR, "01_compare_rules_table.csv"), index=False, encoding="utf-8-sig")

    # 5.1 Compare overlap spam item rate
    top_labels = (
        compare_df.groupby("label", as_index=False)["overlap_spam_item_rate_pct"]
        .max()
        .sort_values("overlap_spam_item_rate_pct", ascending=False)
        .head(TOP_N)["label"]
        .tolist()
    )

    plot_df = compare_df[compare_df["label"].isin(top_labels)].copy()

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="overlap_spam_item_rate_pct",
        color="rule",
        barmode="group",
        hover_data=[
            "total_items",
            "items_in_duplicate_groups",
            "overlap_repost_items_old_not_expired",
            "duplicate_item_rate_pct",
            "overlap_spam_seller_rate_pct",
        ],
        title="Rule comparison — Overlap spam item rate",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Overlap spam item rate (%)",
        xaxis_tickangle=-35,
        height=720,
    )
    save_fig(fig, "11_compare_rules_overlap_spam_item_rate", "Compare rule overlap rate")

    # 5.2 Compare duplicate item rate
    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="duplicate_item_rate_pct",
        color="rule",
        barmode="group",
        hover_data=[
            "total_items",
            "items_in_duplicate_groups",
            "overlap_repost_items_old_not_expired",
            "overlap_spam_item_rate_pct",
            "overlap_spam_seller_rate_pct",
        ],
        title="Rule comparison — Duplicate item rate",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Duplicate item rate (%)",
        xaxis_tickangle=-35,
        height=720,
    )
    save_fig(fig, "12_compare_rules_duplicate_item_rate", "Compare rule duplicate rate")

# ============================================================
# 6. OTHER LISTING SPAM TABLES
# ============================================================

# 6.1 Repost expiry status
repost_status = read_table("02_repost_expiry_status_breakdown")

if repost_status is not None:
    df = make_label(repost_status)
    df = to_numeric_safe(df, [
        "items",
        "sellers",
        "groups",
        "avg_days_between_reposts",
        "median_days_between_reposts",
        "avg_previous_remaining_days_at_new_post",
        "avg_overlap_days_with_previous_item",
        "total_overlap_days_with_previous_item",
    ])

    plot_df = df.sort_values("items", ascending=False).head(TOP_N * 2)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="items",
        color="repost_expiry_status",
        hover_data=[
            "sellers",
            "groups",
            "avg_days_between_reposts",
            "median_days_between_reposts",
            "avg_previous_remaining_days_at_new_post",
            "avg_overlap_days_with_previous_item",
            "total_overlap_days_with_previous_item",
        ],
        title="Strict rule — Repost status breakdown",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Items",
        xaxis_tickangle=-35,
        height=720,
    )
    save_fig(fig, "13_strict_repost_status_breakdown", "Strict repost status")

# 6.2 Top sellers spam pressure
top_sellers = read_table("06_top_sellers_spam_pressure")

if top_sellers is not None:
    df = top_sellers.copy()
    df["seller_short"] = df["seller_id"].astype(str).str.slice(0, 10)
    df = to_numeric_safe(df, [
        "total_items_seen_in_sequence_table",
        "items_in_exact_duplicate_groups",
        "repost_items",
        "overlap_repost_items_old_not_expired",
        "distinct_exact_objects",
        "exact_objects_with_duplicates",
        "avg_days_between_reposts",
        "median_days_between_reposts",
        "avg_overlap_days_when_spam",
        "total_overlap_days",
    ])

    plot_df = df.sort_values("overlap_repost_items_old_not_expired", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="seller_short",
        y="overlap_repost_items_old_not_expired",
        color="seller_type_clean" if "seller_type_clean" in plot_df.columns else None,
        hover_data=[
            "seller_id",
            "items_in_exact_duplicate_groups",
            "repost_items",
            "distinct_exact_objects",
            "exact_objects_with_duplicates",
            "avg_days_between_reposts",
            "median_days_between_reposts",
            "avg_overlap_days_when_spam",
            "total_overlap_days",
        ],
        title="Top sellers by overlap spam repost items",
    )
    fig.update_layout(
        xaxis_title="seller_id short",
        yaxis_title="Overlap repost items",
        xaxis_tickangle=-35,
        height=650,
    )
    save_fig(fig, "14_top_sellers_overlap_spam_pressure", "Top sellers spam pressure")

# 6.3 Cross-seller exact duplicate
cross_type = read_table("08_cross_seller_exact_duplicate_by_type_category")

if cross_type is not None:
    df = make_label(cross_type)
    df = to_numeric_safe(df, [
        "cross_seller_exact_groups",
        "items_in_cross_seller_exact_groups",
        "seller_group_counts_sum",
        "avg_items_per_group",
        "avg_sellers_per_group",
        "max_items_in_group",
        "max_sellers_in_group",
    ])

    plot_df = df.sort_values("items_in_cross_seller_exact_groups", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="items_in_cross_seller_exact_groups",
        color="seller_type_mix" if "seller_type_mix" in plot_df.columns else None,
        hover_data=[
            "cross_seller_exact_groups",
            "seller_group_counts_sum",
            "avg_items_per_group",
            "avg_sellers_per_group",
            "max_items_in_group",
            "max_sellers_in_group",
        ],
        title="Cross-seller exact duplicates — possible shared supply/project",
    )
    fig.update_layout(
        xaxis_title="seller_type_mix | category",
        yaxis_title="Items in cross-seller duplicate groups",
        xaxis_tickangle=-35,
        height=650,
    )
    save_fig(fig, "15_cross_seller_duplicate_by_type_category", "Cross seller duplicate")

# 6.4 Optional snapshot traffic
traffic = read_table("09_optional_snapshot_traffic_for_exact_duplicate_items")

if traffic is not None:
    df = make_label(traffic)
    df = to_numeric_safe(df, [
        "duplicate_items_with_snapshot",
        "sellers",
        "duplicate_groups",
        "total_views_24h_on_duplicate_items",
        "total_contacts_24h_on_duplicate_items",
        "avg_daily_views_per_snapshot_row",
        "avg_daily_contacts_per_snapshot_row",
        "contact_per_view_on_duplicate_items",
    ])
    df = add_pct_cols(df, ["contact_per_view_on_duplicate_items"])

    plot_df = df.sort_values("total_views_24h_on_duplicate_items", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="total_views_24h_on_duplicate_items",
        color="repost_expiry_status" if "repost_expiry_status" in plot_df.columns else None,
        hover_data=[
            "duplicate_items_with_snapshot",
            "sellers",
            "duplicate_groups",
            "total_contacts_24h_on_duplicate_items",
            "contact_per_view_on_duplicate_items_pct",
            "avg_daily_views_per_snapshot_row",
            "avg_daily_contacts_per_snapshot_row",
        ],
        title="Views consumed by exact duplicate listing groups",
    )
    fig.update_layout(
        xaxis_title="seller_type | category",
        yaxis_title="Total views_24h on duplicate items",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "16_duplicate_groups_snapshot_views", "Duplicate snapshot traffic")

# ============================================================
# 7. LOGIN USER LIFECYCLE VISUALS
# ============================================================

login_overall = read_table("06_login_user_active_overall_summary")
active_pattern = read_table("07_login_user_active_pattern_distribution")
active_days = read_table("08_login_user_active_days_distribution")
off_gap = read_table("09_login_user_off_gap_distribution")
daily_login = read_table("10_daily_login_active_users")

if login_overall is not None:
    df = login_overall.copy()
    df = to_numeric_safe(df, df.columns)

    # KPI table
    fig = go.Figure(data=[go.Table(
        header=dict(values=list(df.columns), align="left"),
        cells=dict(values=[df[c] for c in df.columns], align="left")
    )])
    fig.update_layout(
        title="Login user lifecycle — Overall summary",
        height=430,
    )
    save_fig(fig, "17_login_overall_summary_table", "Login overall summary")

if active_pattern is not None:
    df = active_pattern.copy()
    df = to_numeric_safe(df, [
        "users",
        "user_share",
        "avg_active_days",
        "avg_event_rows",
        "avg_contact_flag_rows",
        "avg_positive_event_rows",
    ])
    df = add_pct_cols(df, ["user_share"])

    fig = px.bar(
        plotly_safe_df(df.sort_values("users", ascending=False)),
        x="active_pattern",
        y="users",
        text="user_share_pct",
        hover_data=[
            "user_share_pct",
            "avg_active_days",
            "avg_event_rows",
            "avg_contact_flag_rows",
            "avg_positive_event_rows",
        ],
        title="Login users by active pattern",
    )
    fig.update_traces(texttemplate="%{text:.2f}%", textposition="outside")
    fig.update_layout(
        xaxis_title="Active pattern",
        yaxis_title="Users",
        xaxis_tickangle=-20,
        height=600,
    )
    save_fig(fig, "18_login_active_pattern_distribution", "Active pattern")

if active_days is not None:
    df = active_days.copy()
    df = to_numeric_safe(df, ["sort_key", "users", "user_share"])
    df = add_pct_cols(df, ["user_share"])
    df = df.sort_values(["sort_key", "active_pattern"])

    fig = px.bar(
        plotly_safe_df(df),
        x="active_days_bucket",
        y="users",
        color="active_pattern",
        hover_data=["user_share_pct"],
        title="Login users by active-day bucket",
    )
    fig.update_layout(
        xaxis_title="Active days bucket",
        yaxis_title="Users",
        xaxis_tickangle=-25,
        height=620,
    )
    save_fig(fig, "19_login_active_days_distribution", "Active days distribution")

if off_gap is not None:
    df = off_gap.copy()
    df = to_numeric_safe(df, ["sort_key", "users", "user_share"])
    df = add_pct_cols(df, ["user_share"])
    df = df.sort_values(["sort_key", "active_pattern"])

    fig = px.bar(
        plotly_safe_df(df),
        x="off_gap_bucket",
        y="users",
        color="active_pattern",
        hover_data=["user_share_pct"],
        title="Login users by max inactive gap",
    )
    fig.update_layout(
        xaxis_title="Max inactive gap bucket",
        yaxis_title="Users",
        xaxis_tickangle=-25,
        height=620,
    )
    save_fig(fig, "20_login_off_gap_distribution", "Off gap distribution")

if daily_login is not None:
    df = daily_login.copy()
    df["active_date"] = pd.to_datetime(df["active_date"], errors="coerce")
    df = df.sort_values("active_date")
    df = to_numeric_safe(df, [
        "login_active_users",
        "event_rows",
        "contact_flag_rows",
        "positive_event_rows",
        "contact_flag_rate",
        "positive_event_rate",
    ])
    df = add_pct_cols(df, ["contact_flag_rate", "positive_event_rate"])
    df["login_active_users_7d_ma"] = df["login_active_users"].rolling(7, min_periods=1).mean()

    # 7.1 Daily active users
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df["active_date"],
        y=df["login_active_users"],
        mode="lines",
        name="Daily login active users",
        hovertemplate="Date=%{x}<br>Users=%{y:,.0f}<extra></extra>",
    ))
    fig.add_trace(go.Scatter(
        x=df["active_date"],
        y=df["login_active_users_7d_ma"],
        mode="lines",
        name="7D moving avg",
        hovertemplate="Date=%{x}<br>7D MA=%{y:,.0f}<extra></extra>",
    ))
    fig.update_layout(
        title="Daily login active users",
        xaxis_title="Date",
        yaxis_title="Users",
        height=620,
    )
    save_fig(fig, "21_daily_login_active_users", "Daily login active users")

    # 7.2 Rates
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df["active_date"],
        y=df["contact_flag_rate_pct"],
        mode="lines",
        name="Contact flag rate",
        hovertemplate="Date=%{x}<br>Contact flag rate=%{y:.4f}%<extra></extra>",
    ))
    fig.add_trace(go.Scatter(
        x=df["active_date"],
        y=df["positive_event_rate_pct"],
        mode="lines",
        name="Positive event rate",
        hovertemplate="Date=%{x}<br>Positive event rate=%{y:.4f}%<extra></extra>",
    ))
    fig.update_layout(
        title="Daily login behavior rates",
        xaxis_title="Date",
        yaxis_title="Rate (%)",
        height=620,
    )
    save_fig(fig, "22_daily_login_contact_positive_rates", "Daily login rates")

    # 7.3 Rows
    fig = px.line(
        plotly_safe_df(df),
        x="active_date",
        y=["event_rows", "contact_flag_rows", "positive_event_rows"],
        title="Daily login event/contact/positive rows",
    )
    fig.update_layout(
        xaxis_title="Date",
        yaxis_title="Rows",
        height=620,
    )
    save_fig(fig, "23_daily_login_event_rows", "Daily login rows")

# ============================================================
# 8. ONE-DAY USER VISUALS
# ============================================================

one_day_summary = read_table("11_one_day_login_users_summary")
one_day_event = read_table("12_one_day_login_users_event_type_category")
one_day_first = read_table("13_one_day_login_users_first_action_event_type_category")
one_day_date = read_table("14_one_day_login_users_by_only_active_date")

if one_day_summary is not None:
    df = one_day_summary.copy()

    fig = go.Figure(data=[go.Table(
        header=dict(values=list(df.columns), align="left"),
        cells=dict(values=[df[c] for c in df.columns], align="left")
    )])
    fig.update_layout(
        title="One-day login users — Summary",
        height=430,
    )
    save_fig(fig, "24_one_day_summary_table", "One-day summary")

if one_day_event is not None:
    df = make_label(one_day_event)
    df = to_numeric_safe(df, [
        "rows",
        "row_share",
        "users",
        "user_share_among_one_day_users",
        "sessions",
        "items",
        "contact_flag_rows",
        "positive_event_rows",
    ])
    df = add_pct_cols(df, ["row_share", "user_share_among_one_day_users"])

    plot_df = df.sort_values("rows", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="rows",
        color="event_type_clean" if "event_type_clean" in plot_df.columns else None,
        hover_data=[
            "row_share_pct",
            "users",
            "user_share_among_one_day_users_pct",
            "contact_flag_rows",
            "positive_event_rows",
        ],
        title="One-day users — Event rows by event type × category",
    )
    fig.update_layout(
        xaxis_title="event_type | category",
        yaxis_title="Rows",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "25_one_day_event_type_category_rows", "One-day event rows")

    plot_df = df.sort_values("users", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="users",
        color="event_type_clean" if "event_type_clean" in plot_df.columns else None,
        hover_data=[
            "rows",
            "row_share_pct",
            "user_share_among_one_day_users_pct",
            "contact_flag_rows",
            "positive_event_rows",
        ],
        title="One-day users — Users by event type × category",
    )
    fig.update_layout(
        xaxis_title="event_type | category",
        yaxis_title="Users",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "26_one_day_event_type_category_users", "One-day users by action")

if one_day_first is not None:
    df = make_label(one_day_first)
    df = to_numeric_safe(df, ["users", "user_share"])
    df = add_pct_cols(df, ["user_share"])

    plot_df = df.sort_values("users", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="users",
        color="first_event_type" if "first_event_type" in plot_df.columns else None,
        hover_data=["user_share_pct"],
        title="One-day users — First action by event type × category",
    )
    fig.update_layout(
        xaxis_title="first_event_type | category",
        yaxis_title="Users",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "27_one_day_first_action", "One-day first action")

if one_day_date is not None:
    df = one_day_date.copy()
    df["only_active_date"] = pd.to_datetime(df["only_active_date"], errors="coerce")
    df = df.sort_values("only_active_date")
    df = to_numeric_safe(df, [
        "one_day_users",
        "event_rows",
        "contact_flag_rows",
        "positive_event_rows",
    ])
    df["one_day_users_7d_ma"] = df["one_day_users"].rolling(7, min_periods=1).mean()

    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df["only_active_date"],
        y=df["one_day_users"],
        mode="lines",
        name="One-day users",
        hovertemplate="Date=%{x}<br>Users=%{y:,.0f}<extra></extra>",
    ))
    fig.add_trace(go.Scatter(
        x=df["only_active_date"],
        y=df["one_day_users_7d_ma"],
        mode="lines",
        name="7D moving avg",
        hovertemplate="Date=%{x}<br>7D MA=%{y:,.0f}<extra></extra>",
    ))
    fig.update_layout(
        title="One-day login users by only active date",
        xaxis_title="Date",
        yaxis_title="Users",
        height=620,
    )
    save_fig(fig, "28_one_day_users_by_date", "One-day users timeline")

    fig = px.line(
        plotly_safe_df(df),
        x="only_active_date",
        y=["event_rows", "contact_flag_rows", "positive_event_rows"],
        title="One-day users — Event/contact/positive rows over time",
    )
    fig.update_layout(
        xaxis_title="Date",
        yaxis_title="Rows",
        height=620,
    )
    save_fig(fig, "29_one_day_event_rows_by_date", "One-day event rows timeline")

# ============================================================
# 9. FACT_USER_EVENTS TRAFFIC / CONTACT OUTPUTS
# ============================================================

events_login = read_table("15_fact_user_events_is_login_summary")
events_type = read_table("16_fact_user_events_by_is_login_event_type")
events_type_cat = read_table("17_fact_user_events_by_is_login_event_type_category")

if events_login is not None:
    df = events_login.copy()
    df = to_numeric_safe(df, [
        "rows",
        "row_share",
        "contact_flag_rows",
        "contact_flag_share_of_all_contacts",
        "positive_event_rows",
        "positive_event_share_of_all_positive_events",
        "contact_flag_rate_inside_group",
        "positive_event_rate_inside_group",
    ])
    df = add_pct_cols(df, [
        "row_share",
        "contact_flag_share_of_all_contacts",
        "positive_event_share_of_all_positive_events",
        "contact_flag_rate_inside_group",
        "positive_event_rate_inside_group",
    ])

    fig = px.bar(
        plotly_safe_df(df),
        x="is_login_clean",
        y=["rows", "contact_flag_rows", "positive_event_rows"],
        barmode="group",
        hover_data=[
            "row_share_pct",
            "contact_flag_share_of_all_contacts_pct",
            "positive_event_share_of_all_positive_events_pct",
            "contact_flag_rate_inside_group_pct",
            "positive_event_rate_inside_group_pct",
        ],
        title="fact_user_events — Rows/contact/positive by login status",
    )
    fig.update_layout(
        xaxis_title="Login status",
        yaxis_title="Rows",
        height=600,
    )
    save_fig(fig, "30_events_login_summary", "Events login summary")

if events_type is not None:
    df = make_label(events_type)
    df = to_numeric_safe(df, [
        "rows",
        "row_share",
        "contact_flag_rows",
        "positive_event_rows",
        "contact_flag_rate_inside_group",
        "positive_event_rate_inside_group",
    ])
    df = add_pct_cols(df, [
        "row_share",
        "contact_flag_rate_inside_group",
        "positive_event_rate_inside_group",
    ])

    plot_df = df.sort_values("rows", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="rows",
        color="is_login_clean" if "is_login_clean" in plot_df.columns else None,
        hover_data=[
            "row_share_pct",
            "contact_flag_rows",
            "positive_event_rows",
            "contact_flag_rate_inside_group_pct",
            "positive_event_rate_inside_group_pct",
        ],
        title="fact_user_events — Rows by login status × event type",
    )
    fig.update_layout(
        xaxis_title="is_login | event_type",
        yaxis_title="Rows",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "31_events_rows_by_login_event_type", "Events by login event type")

    plot_df = df.sort_values("positive_event_rows", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label",
        y="positive_event_rows",
        color="is_login_clean" if "is_login_clean" in plot_df.columns else None,
        hover_data=[
            "rows",
            "row_share_pct",
            "contact_flag_rows",
            "positive_event_rate_inside_group_pct",
        ],
        title="fact_user_events — Positive rows by login status × event type",
    )
    fig.update_layout(
        xaxis_title="is_login | event_type",
        yaxis_title="Positive event rows",
        xaxis_tickangle=-35,
        height=680,
    )
    save_fig(fig, "32_events_positive_by_login_event_type", "Events positive by login event type")

if events_type_cat is not None:
    df = make_label(events_type_cat)
    df = to_numeric_safe(df, [
        "rows",
        "row_share",
        "contact_flag_rows",
        "positive_event_rows",
        "contact_flag_rate_inside_group",
        "positive_event_rate_inside_group",
    ])
    df = add_pct_cols(df, [
        "row_share",
        "contact_flag_rate_inside_group",
        "positive_event_rate_inside_group",
    ])

    # label with login too
    if "is_login_clean" in df.columns:
        df["label2"] = df["is_login_clean"].astype(str) + " | " + df["label"].astype(str)
    else:
        df["label2"] = df["label"]

    plot_df = df.sort_values("rows", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label2",
        y="rows",
        color="event_type_clean" if "event_type_clean" in plot_df.columns else None,
        hover_data=[
            "row_share_pct",
            "contact_flag_rows",
            "positive_event_rows",
            "contact_flag_rate_inside_group_pct",
            "positive_event_rate_inside_group_pct",
        ],
        title="fact_user_events — Rows by login × event type × category",
    )
    fig.update_layout(
        xaxis_title="login | event_type | category",
        yaxis_title="Rows",
        xaxis_tickangle=-35,
        height=720,
    )
    save_fig(fig, "33_events_rows_by_login_event_category", "Events login event category")

    plot_df = df.sort_values("positive_event_rows", ascending=False).head(TOP_N)

    fig = px.bar(
        plotly_safe_df(plot_df),
        x="label2",
        y="positive_event_rows",
        color="event_type_clean" if "event_type_clean" in plot_df.columns else None,
        hover_data=[
            "rows",
            "row_share_pct",
            "contact_flag_rows",
            "positive_event_rate_inside_group_pct",
        ],
        title="fact_user_events — Positive rows by login × event type × category",
    )
    fig.update_layout(
        xaxis_title="login | event_type | category",
        yaxis_title="Positive event rows",
        xaxis_tickangle=-35,
        height=720,
    )
    save_fig(fig, "34_events_positive_by_login_event_category", "Events positive login event category")

# ============================================================
# 10. SAVE CHART INDEX
# ============================================================

chart_index_df = pd.DataFrame(chart_index)
chart_index_path = os.path.join(OUTPUT_DIR, "99_chart_index.csv")
chart_index_df.to_csv(chart_index_path, index=False, encoding="utf-8-sig")

print("\nDONE.")
print("Charts created:", len(chart_index_df))
print("Chart index:", chart_index_path)
display(chart_index_df)
print("Output folder:", OUTPUT_DIR)
```

```python
# ============================================================
# USER SEARCH JOURNEY / BURST-THEN-GONE ANALYSIS
# Goal:
# - User vào app thường xuyên trong X ngày rồi biến mất?
# - Estimate "house-hunting/search mission duration"
#
# Input:
# - Kaggle input chứa login_lifecycle_by_bucket/lifecycle_bucket_*.parquet
#   từ output notebook trước
#
# Output:
# - CSV summaries
# - Interactive Plotly HTML charts
# ============================================================

import os
import glob
import sys
import subprocess
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

try:
    import duckdb
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "duckdb"])
    import duckdb

try:
    import plotly.express as px
    import plotly.graph_objects as go
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "plotly"])
    import plotly.express as px
    import plotly.graph_objects as go

try:
    from IPython.display import display
except Exception:
    def display(x):
        print(x)

# ============================================================
# 0. CONFIG
# ============================================================

INPUT_ROOT = "/kaggle/input"
OUTPUT_DIR = "/kaggle/working/user_search_journey_analysis"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# User được xem là "biến mất" nếu sau last_active_date
# không thấy quay lại ít nhất CHURN_GAP_DAYS ngày.
CHURN_GAP_DAYS = 14

# Định nghĩa "vào thường xuyên"
# max gap giữa các active days <= 2 ngày
REGULAR_MAX_GAP_DAYS = 2

TOP_N = 30

print("INPUT_ROOT:", INPUT_ROOT)
print("OUTPUT_DIR:", OUTPUT_DIR)

# ============================================================
# 1. FIND LIFECYCLE FILES
# ============================================================

lifecycle_files = sorted(
    glob.glob(os.path.join(INPUT_ROOT, "**", "lifecycle_bucket_*.parquet"), recursive=True)
)

if len(lifecycle_files) == 0:
    lifecycle_files = sorted(
        glob.glob(os.path.join(INPUT_ROOT, "**", "*lifecycle*.parquet"), recursive=True)
    )

print("Found lifecycle files:", len(lifecycle_files))
for p in lifecycle_files[:10]:
    print(" -", p)

if len(lifecycle_files) == 0:
    raise FileNotFoundError(
        "Không tìm thấy lifecycle parquet. Hãy chắc là Kaggle input có folder "
        "login_lifecycle_by_bucket/lifecycle_bucket_*.parquet"
    )

LIFECYCLE_GLOB = os.path.commonpath(lifecycle_files)
if os.path.isfile(LIFECYCLE_GLOB):
    LIFECYCLE_GLOB = lifecycle_files[0]
else:
    LIFECYCLE_GLOB = os.path.join(LIFECYCLE_GLOB, "*.parquet").replace("\\", "/")

# Nếu commonpath bị quá rộng, dùng list explicit
LIFECYCLE_GLOB = os.path.join(
    os.path.dirname(lifecycle_files[0]), "*.parquet"
).replace("\\", "/")

print("LIFECYCLE_GLOB:", LIFECYCLE_GLOB)

# ============================================================
# 2. HELPERS
# ============================================================

def save_df(df, name, show_head=30):
    csv_path = os.path.join(OUTPUT_DIR, f"{name}.csv")
    parquet_path = os.path.join(OUTPUT_DIR, f"{name}.parquet")

    df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    try:
        df.to_parquet(parquet_path, index=False)
    except Exception as e:
        print("[WARN] Cannot save parquet:", e)

    print("[SAVE]", csv_path)
    display(df.head(show_head))
    return df

def save_fig(fig, name):
    html_path = os.path.join(OUTPUT_DIR, f"{name}.html")
    fig.write_html(html_path)
    print("[SAVE HTML]", html_path)

    try:
        png_path = os.path.join(OUTPUT_DIR, f"{name}.png")
        fig.write_image(png_path, scale=2)
        print("[SAVE PNG]", png_path)
    except Exception:
        print("[SKIP PNG] Nếu muốn save PNG thì cài kaleido.")

    fig.show()

def plotly_safe_df(df):
    return df.copy().astype(object).where(pd.notna(df), None)

# ============================================================
# 3. BUILD JOURNEY TABLE
# ============================================================

con = duckdb.connect()
con.execute("PRAGMA threads=2")
con.execute("PRAGMA memory_limit='6GB'")
con.execute("PRAGMA preserve_insertion_order=false")

# Get data end
data_end = con.execute(f"""
SELECT MAX(CAST(last_active_date AS DATE)) AS data_end
FROM read_parquet('{LIFECYCLE_GLOB}')
""").df()["data_end"].iloc[0]

print("DATA_END:", data_end)
print("CHURN_GAP_DAYS:", CHURN_GAP_DAYS)

journey = con.execute(f"""
WITH base AS (
    SELECT
        CAST(user_id AS VARCHAR) AS user_id,
        CAST(first_active_date AS DATE) AS first_active_date,
        CAST(last_active_date AS DATE) AS last_active_date,

        CAST(active_days AS BIGINT) AS active_days,
        CAST(observed_span_days AS BIGINT) AS observed_span_days,
        CAST(max_inactive_gap_days AS BIGINT) AS max_inactive_gap_days,
        CAST(avg_inactive_gap_days AS DOUBLE) AS avg_inactive_gap_days,
        CAST(long_off_gap_count AS BIGINT) AS long_off_gap_count,

        CAST(total_event_rows AS BIGINT) AS total_event_rows,
        CAST(total_contact_flag_rows AS BIGINT) AS total_contact_flag_rows,
        CAST(total_positive_event_rows AS BIGINT) AS total_positive_event_rows,

        CAST(active_ratio_inside_observed_span AS DOUBLE) AS active_ratio_inside_observed_span,
        CAST(active_pattern AS VARCHAR) AS old_active_pattern
    FROM read_parquet('{LIFECYCLE_GLOB}')
),

scored AS (
    SELECT
        *,

        DATE_DIFF('day', last_active_date, CAST('{data_end}' AS DATE)) AS inactive_after_last_days,

        CASE
            WHEN last_active_date <= CAST('{data_end}' AS DATE) - INTERVAL {CHURN_GAP_DAYS} DAY
                THEN 1
            ELSE 0
        END AS is_observed_gone,

        CASE
            WHEN last_active_date > CAST('{data_end}' AS DATE) - INTERVAL {CHURN_GAP_DAYS} DAY
                THEN 1
            ELSE 0
        END AS is_right_censored_near_data_end,

        CASE
            WHEN observed_span_days = 1 THEN '01_1_day'
            WHEN observed_span_days BETWEEN 2 AND 3 THEN '02_2_3_days'
            WHEN observed_span_days BETWEEN 4 AND 7 THEN '03_4_7_days'
            WHEN observed_span_days BETWEEN 8 AND 14 THEN '04_8_14_days'
            WHEN observed_span_days BETWEEN 15 AND 30 THEN '05_15_30_days'
            WHEN observed_span_days BETWEEN 31 AND 60 THEN '06_31_60_days'
            WHEN observed_span_days BETWEEN 61 AND 90 THEN '07_61_90_days'
            ELSE '08_91plus_days'
        END AS journey_span_bucket,

        CASE
            WHEN observed_span_days = 1 THEN 1
            WHEN observed_span_days BETWEEN 2 AND 3 THEN 2
            WHEN observed_span_days BETWEEN 4 AND 7 THEN 3
            WHEN observed_span_days BETWEEN 8 AND 14 THEN 4
            WHEN observed_span_days BETWEEN 15 AND 30 THEN 5
            WHEN observed_span_days BETWEEN 31 AND 60 THEN 6
            WHEN observed_span_days BETWEEN 61 AND 90 THEN 7
            ELSE 8
        END AS journey_span_sort,

        CASE
            WHEN active_days = 1 THEN '01_one_day_only'

            WHEN last_active_date > CAST('{data_end}' AS DATE) - INTERVAL {CHURN_GAP_DAYS} DAY
                THEN '07_right_censored_still_possible'

            WHEN active_days >= 2
             AND max_inactive_gap_days <= {REGULAR_MAX_GAP_DAYS}
             AND observed_span_days BETWEEN 2 AND 7
                THEN '02_short_frequent_burst_then_gone'

            WHEN active_days >= 2
             AND max_inactive_gap_days <= {REGULAR_MAX_GAP_DAYS}
             AND observed_span_days BETWEEN 8 AND 30
                THEN '03_medium_frequent_burst_then_gone'

            WHEN active_days >= 2
             AND max_inactive_gap_days <= {REGULAR_MAX_GAP_DAYS}
             AND observed_span_days > 30
                THEN '04_long_frequent_search_then_gone'

            WHEN active_days >= 2
             AND max_inactive_gap_days >= 7
                THEN '05_episodic_returner_then_gone'

            ELSE '06_irregular_search_then_gone'
        END AS journey_segment

    FROM base
)

SELECT *
FROM scored
""").df()

con.close()

journey = save_df(journey, "00_user_search_journey_table", show_head=30)

# ============================================================
# 4. SUMMARY TABLES
# ============================================================

overall = pd.DataFrame([{
    "total_login_users": len(journey),
    "data_end": data_end,
    "churn_gap_days": CHURN_GAP_DAYS,

    "observed_gone_users": int(journey["is_observed_gone"].sum()),
    "right_censored_near_data_end_users": int(journey["is_right_censored_near_data_end"].sum()),

    "avg_journey_span_days": journey["observed_span_days"].mean(),
    "median_journey_span_days": journey["observed_span_days"].median(),
    "p75_journey_span_days": journey["observed_span_days"].quantile(0.75),
    "p90_journey_span_days": journey["observed_span_days"].quantile(0.90),
    "p95_journey_span_days": journey["observed_span_days"].quantile(0.95),

    "avg_active_days": journey["active_days"].mean(),
    "median_active_days": journey["active_days"].median(),
    "p90_active_days": journey["active_days"].quantile(0.90),

    "avg_active_density": journey["active_ratio_inside_observed_span"].mean(),
    "median_active_density": journey["active_ratio_inside_observed_span"].median(),
}])

save_df(overall, "01_search_journey_overall_summary")

segment_summary = (
    journey.groupby("journey_segment", as_index=False)
           .agg(
               users=("user_id", "count"),
               avg_journey_span_days=("observed_span_days", "mean"),
               median_journey_span_days=("observed_span_days", "median"),
               p90_journey_span_days=("observed_span_days", lambda x: x.quantile(0.90)),
               avg_active_days=("active_days", "mean"),
               median_active_days=("active_days", "median"),
               avg_active_density=("active_ratio_inside_observed_span", "mean"),
               avg_total_event_rows=("total_event_rows", "mean"),
               avg_total_positive_event_rows=("total_positive_event_rows", "mean"),
               avg_total_contact_flag_rows=("total_contact_flag_rows", "mean"),
           )
)

segment_summary["user_share"] = segment_summary["users"] / segment_summary["users"].sum()
segment_summary["user_share_pct"] = segment_summary["user_share"] * 100
segment_summary = segment_summary.sort_values("journey_segment")

save_df(segment_summary, "02_search_journey_segment_summary")

duration_summary = (
    journey.groupby(["journey_span_sort", "journey_span_bucket", "journey_segment"], as_index=False)
           .agg(users=("user_id", "count"))
           .sort_values(["journey_span_sort", "journey_segment"])
)

duration_summary["user_share"] = duration_summary["users"] / duration_summary["users"].sum()
duration_summary["user_share_pct"] = duration_summary["user_share"] * 100

save_df(duration_summary, "03_search_journey_duration_bucket_summary", show_head=80)

# ============================================================
# 5. CHARTS
# ============================================================

# 5.1 Segment distribution
fig = px.bar(
    plotly_safe_df(segment_summary.sort_values("users", ascending=False)),
    x="journey_segment",
    y="users",
    text="user_share_pct",
    hover_data=[
        "median_journey_span_days",
        "p90_journey_span_days",
        "median_active_days",
        "avg_active_density",
        "avg_total_positive_event_rows",
        "avg_total_contact_flag_rows",
    ],
    title="User search journey segments: who searches for a while then disappears?",
)

fig.update_traces(texttemplate="%{text:.2f}%", textposition="outside")
fig.update_layout(
    xaxis_title="Journey segment",
    yaxis_title="Users",
    xaxis_tickangle=-25,
    height=680,
)
save_fig(fig, "01_journey_segment_distribution")

# 5.2 Journey duration bucket stacked by segment
fig = px.bar(
    plotly_safe_df(duration_summary),
    x="journey_span_bucket",
    y="users",
    color="journey_segment",
    hover_data=["user_share_pct"],
    title="Search journey duration buckets by segment",
)

fig.update_layout(
    xaxis_title="Observed journey span",
    yaxis_title="Users",
    xaxis_tickangle=-25,
    height=680,
)
save_fig(fig, "02_journey_duration_bucket_by_segment")

# 5.3 Only observed-gone users: duration distribution
gone = journey[journey["is_observed_gone"] == 1].copy()

fig = px.histogram(
    plotly_safe_df(gone),
    x="observed_span_days",
    nbins=80,
    marginal="box",
    title=f"Observed-gone users: distribution of search journey span<br><sup>Gone = no activity for at least {CHURN_GAP_DAYS} days after last_active_date</sup>",
)

fig.update_layout(
    xaxis_title="Journey span days",
    yaxis_title="Users",
    height=650,
)
save_fig(fig, "03_gone_users_journey_span_histogram")

# 5.4 Active days vs journey span
sample_n = min(100_000, len(journey))
sample = journey.sample(sample_n, random_state=42) if len(journey) > sample_n else journey.copy()

fig = px.scatter(
    plotly_safe_df(sample),
    x="observed_span_days",
    y="active_days",
    color="journey_segment",
    opacity=0.35,
    hover_data=[
        "active_ratio_inside_observed_span",
        "max_inactive_gap_days",
        "inactive_after_last_days",
        "total_positive_event_rows",
        "total_contact_flag_rows",
    ],
    title=f"Active days vs journey span sampled users n={sample_n:,}",
)

fig.update_layout(
    xaxis_title="Journey span days",
    yaxis_title="Active days",
    height=680,
)
save_fig(fig, "04_active_days_vs_journey_span_scatter")

# 5.5 Active density by duration bucket
density_summary = (
    journey.groupby(["journey_span_sort", "journey_span_bucket"], as_index=False)
           .agg(
               users=("user_id", "count"),
               avg_active_density=("active_ratio_inside_observed_span", "mean"),
               median_active_density=("active_ratio_inside_observed_span", "median"),
               avg_active_days=("active_days", "mean"),
           )
           .sort_values("journey_span_sort")
)

save_df(density_summary, "04_density_by_duration_bucket")

fig = px.bar(
    plotly_safe_df(density_summary),
    x="journey_span_bucket",
    y="median_active_density",
    hover_data=["users", "avg_active_density", "avg_active_days"],
    title="Median active density by journey duration bucket",
)

fig.update_layout(
    xaxis_title="Journey span bucket",
    yaxis_title="Median active density = active_days / span_days",
    xaxis_tickangle=-25,
    height=600,
)
save_fig(fig, "05_active_density_by_duration_bucket")

# 5.6 Survival curve:
# P(user still observed at or after day N since first active)
# Censor-aware approx: only include users whose first_active_date <= data_end - N
max_day = 120
survival_rows = []

journey["first_active_date"] = pd.to_datetime(journey["first_active_date"])
journey["last_active_date"] = pd.to_datetime(journey["last_active_date"])
data_end_ts = pd.to_datetime(data_end)

for n in range(0, max_day + 1):
    eligible = journey[journey["first_active_date"] <= data_end_ts - pd.Timedelta(days=n)]
    if len(eligible) == 0:
        continue

    survived = (
        (eligible["last_active_date"] - eligible["first_active_date"]).dt.days >= n
    ).sum()

    survival_rows.append({
        "days_since_first_active": n,
        "eligible_users": len(eligible),
        "users_still_observed": int(survived),
        "survival_rate": survived / len(eligible),
        "survival_rate_pct": survived / len(eligible) * 100,
    })

survival = pd.DataFrame(survival_rows)
save_df(survival, "05_search_journey_survival_curve")

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=survival["days_since_first_active"],
    y=survival["survival_rate_pct"],
    mode="lines",
    name="Still observed active at/after day N",
    hovertemplate=(
        "Days since first active=%{x}<br>"
        "Survival rate=%{y:.2f}%<br>"
        "<extra></extra>"
    )
))

fig.update_layout(
    title="How long do login users stay in their search journey?",
    xaxis_title="Days since first active",
    yaxis_title="% users still observed at/after this day",
    height=620,
)
save_fig(fig, "06_search_journey_survival_curve")

# 5.7 First-active cohort: average journey length
cohort = journey.copy()
cohort["first_active_month"] = cohort["first_active_date"].dt.to_period("M").dt.to_timestamp()

cohort_summary = (
    cohort.groupby("first_active_month", as_index=False)
          .agg(
              users=("user_id", "count"),
              median_journey_span_days=("observed_span_days", "median"),
              avg_journey_span_days=("observed_span_days", "mean"),
              p90_journey_span_days=("observed_span_days", lambda x: x.quantile(0.90)),
              one_day_rate=("active_days", lambda x: (x == 1).mean()),
              observed_gone_rate=("is_observed_gone", "mean"),
          )
          .sort_values("first_active_month")
)

cohort_summary["one_day_rate_pct"] = cohort_summary["one_day_rate"] * 100
cohort_summary["observed_gone_rate_pct"] = cohort_summary["observed_gone_rate"] * 100

save_df(cohort_summary, "06_first_active_month_cohort_summary")

fig = px.line(
    plotly_safe_df(cohort_summary),
    x="first_active_month",
    y=["median_journey_span_days", "avg_journey_span_days", "p90_journey_span_days"],
    markers=True,
    hover_data=["users", "one_day_rate_pct", "observed_gone_rate_pct"],
    title="Journey duration by first-active month cohort",
)

fig.update_layout(
    xaxis_title="First active month",
    yaxis_title="Journey span days",
    height=620,
)
save_fig(fig, "07_journey_duration_by_first_active_month")

# 5.8 Positive/contact behavior by segment
behavior_summary = (
    journey.groupby("journey_segment", as_index=False)
           .agg(
               users=("user_id", "count"),
               avg_event_rows=("total_event_rows", "mean"),
               avg_positive_event_rows=("total_positive_event_rows", "mean"),
               avg_contact_flag_rows=("total_contact_flag_rows", "mean"),
               users_with_positive=("total_positive_event_rows", lambda x: (x > 0).sum()),
               users_with_contact=("total_contact_flag_rows", lambda x: (x > 0).sum()),
           )
)

behavior_summary["positive_user_rate_pct"] = (
    behavior_summary["users_with_positive"] / behavior_summary["users"] * 100
)
behavior_summary["contact_user_rate_pct"] = (
    behavior_summary["users_with_contact"] / behavior_summary["users"] * 100
)

save_df(behavior_summary, "07_behavior_by_journey_segment")

fig = px.bar(
    plotly_safe_df(behavior_summary.sort_values("users", ascending=False)),
    x="journey_segment",
    y=["positive_user_rate_pct", "contact_user_rate_pct"],
    barmode="group",
    hover_data=[
        "users",
        "avg_event_rows",
        "avg_positive_event_rows",
        "avg_contact_flag_rows",
    ],
    title="Positive/contact user rate by journey segment",
)

fig.update_layout(
    xaxis_title="Journey segment",
    yaxis_title="User rate (%)",
    xaxis_tickangle=-25,
    height=650,
)
save_fig(fig, "08_positive_contact_rate_by_journey_segment")

print("\nDONE.")
print("Output folder:", OUTPUT_DIR)
print("""
Main charts:
01_journey_segment_distribution.html
02_journey_duration_bucket_by_segment.html
03_gone_users_journey_span_histogram.html
04_active_days_vs_journey_span_scatter.html
06_search_journey_survival_curve.html
07_journey_duration_by_first_active_month.html
08_positive_contact_rate_by_journey_segment.html
""")
```
